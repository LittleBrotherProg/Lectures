## 2. Тип 1 — Количество снимков в пакете

<div class="info-box">
<strong style="color:#1b91ff;">Алгоритм:</strong><br>
1. <code>V1 = h · w · i</code> — вес одного снимка<br>
2. <code>Vp = q · t</code> — максимальный вес пакета<br>
3. <code>Q = Vp // V1</code> — количество снимков (округление <strong>вниз</strong>!)
</div>

> ⚠️ Округление **вниз** (`//`): нельзя передать часть снимка — берём меньшее целое.

````python
# Задание 702: 1024×768, 4096 цветов, 1 310 720 бит/с, ≤300 сек
from math import log2

res     = 1024 * 768
colors  = 4096
bitrate = 1_310_720
time    = 300

pixel_size   = log2(colors)        # глубина цвета = 12 бит
img_size     = res * pixel_size    # вес одного снимка
package_size = bitrate * time      # макс. вес пакета

print(package_size // img_size)    # Ответ: 41
````

````python
# Задание 710: 1024×64, 4096 цветов, 655 360 бит/с, ≤70 сек
from math import log2

res = 1024 * 64;  bitrate = 655_360;  colors = 4096;  time = 70

pixel_size        = log2(colors)
img_in_package    = (bitrate * time) // (res * pixel_size)
print(img_in_package)              # Ответ: 58
````

## 3. Тип 2 — Количество цветов в палитре

<div style="display:grid;grid-template-columns:1fr 1fr;gap:12px;margin:16px 0;">
<div style="background:#0d1b2a;border:1.5px solid #1b91ff;border-radius:10px;padding:14px;">
<div style="color:#1b91ff;font-weight:800;margin-bottom:6px;">Максимальное количество</div>
<code>N = 2^i</code> &nbsp; (i округляем <strong>вниз</strong>)
</div>
<div style="background:#0d1b2a;border:1.5px solid #e07b39;border-radius:10px;padding:14px;">
<div style="color:#e07b39;font-weight:800;margin-bottom:6px;">Минимальное количество</div>
<code>N = 2^(i-1) + 1</code> &nbsp; (i округляем <strong>вверх</strong>)
</div>
</div>

> ⚠️ **«более N секунд»** = время строго > N → все промежуточные округления делаем **вверх** (`math.ceil`)

````python
# Задание 711: 2560×5040, 14175 Кбайт → максимальное кол-во цветов?
res         = 2560 * 5040
memory_size = 14_175 * 2**13    # Кбайт → биты (×8×1024 = ×2^13)

pixel_size  = memory_size / res
colors      = 2 ** pixel_size
print(int(colors))               # Ответ: 512
````

````python
# Задание 704: 1280×720, 69 шт, 245760 байт/с, >270 сек → min цветов?
import math

res     = 1280 * 720
bitrate = 245_760 * 8           # байт/с → бит/с
imgs    = 69
time    = 270

Vp = bitrate * time             # минимальный вес пакета
V1 = math.ceil(Vp / imgs)       # вес одного снимка (вверх!)
i  = math.ceil(V1 / res)        # глубина цвета (вверх!)

print(2**(i-1) + 1)             # Ответ: 257
````

## 4. Тип 3 — Экономия трафика при сжатии

<div class="tip-box">
<strong style="color:#2fca72;">Алгоритм:</strong><br>
1. <code>V = h · w · i</code> — вес несжатого<br>
2. <code>Vsj = hsj · wsj · isj</code> — вес сжатого<br>
3. Экономия = <code>(V − Vsj) × N / 2¹³</code> (перевод в Кбайт)
</div>

> Сжатие на **N%** → итоговый файл = **(100 − N)%** от исходного

````python
# Задание 712: 1920×1080, 2^23 цветов → сжатое 1280×1024, 21 бит, 120 фото
res              = 1920 * 1080
bitrate          = 23             # бит/пиксель (2^23 цветов → i=23)
res_compressed   = 1280 * 1024
bitrate_comp     = 21
num_img          = 120

V    = res * bitrate
V_sj = res_compressed * bitrate_comp

ans = (V - V_sj) * num_img / 2**13   # → Кбайт
print(int(ans))                       # Ответ: 295425
````

## 5. Тип 4 — Снимки на карте памяти

<div class="info-box">
<strong style="color:#1b91ff;">Алгоритм:</strong><br>
1. <code>img = h · w · i</code> — вес одного снимка<br>
2. <code>per_card = floor(card_size / img)</code> — снимков на одну карту<br>
3. Снимков на последней карте = <code>total % per_card</code>
</div>

````python
# Задание 706: 7680×4320, 2^16 цветов, карта 9 Гбайт, 4010 снимков
from math import floor

res         = 7680 * 4320
image       = res * 16              # i = 16 (из 2^16)
memory_card = 9 * 2**33             # 9 Гбайт в битах

per_card = floor(memory_card / image)
remind   = 4010 % per_card
print(remind)                       # Ответ: 95
````

## 7. Звук Тип 1 — Время передачи файла

<div class="info-box">
<strong style="color:#1b91ff;">Алгоритм:</strong><br>
1. <code>V = k · t · i · η</code> — вес музыки<br>
2. Добавить заголовки: <code>V += треков × вес_заголовка</code><br>
3. Время = <code>V // скорость_канала</code>
</div>

````python
# Задание 700: стерео 48кГц 34бит, 13 треков 42:20, заголовок 110Кбайт
k       = 2                         # стерео
eta     = 48 * 10**3                # 48 кГц
i       = 34
t       = 42 * 60 + 20             # в секундах
tracks  = 13
header  = 110 * 2**13              # Кбайт → биты
bitrate = 314_572_800

V_tracks = k * t * i * eta
V_album  = V_tracks + header * tracks

print(V_album // bitrate)           # Ответ: 26
````

````python
# Задание 718: стерео, 56кГц, 16 бит, 120 сек, 32000 бит/с
k = 2;  eta = 56 * 10**3;  i = 16;  t = 120;  bitrate = 32_000

V = k * t * i * eta
print(V // bitrate)                 # Ответ: 6720
````

## 8. Звук Тип 2 — Перезапись с новыми параметрами

<div class="tip-box">
<strong style="color:#2fca72;">Идея:</strong> умножаем/делим старый вес на коэффициенты изменений.<br>
<code>V_new = V_old × (k2/k1) × (η2/η1) × (i2/i1)</code>
</div>

````python
# Задание 709: стерео 39 Мбайт → моно, разрешение ×2.5, частота ×4
V_old = 39                          # Мбайт
# стерео(2) → моно(1): ÷2
# разрешение:            ×2.5
# частота:               ×4
V_new = V_old * 2.5 * 4 / 2
print(int(V_new))                   # Ответ: 195

# Задание 716: моно 35 Мбайт → стерео ×2, частота ×3.5
V_old = 35
print(int(V_old * 3.5 * 2))        # Ответ: 245
````
