## Часть 1 — Изображения

### Пиксели и глубина цвета

Каждое изображение состоит из пикселей. Каждый пиксель кодируется определённым количеством бит — это **глубина цвета** (i).

**Формула Хартли:** связывает глубину цвета с количеством цветов в палитре:

```
N = 2^i        — количество цветов по глубине цвета
i = log₂(N)   — глубина цвета по количеству цветов
```

**Формула объёма изображения:**
```
V = h × w × i   (бит)
```
где `h` — высота, `w` — ширина, `i` — глубина цвета (бит/пиксель)

### Передача данных

```
V = q × t        — объём = скорость × время
```

### Перевод единиц (важно!)

```
1 байт  = 8 бит = 2³ бит
1 Кбайт = 2¹³ бит
1 Мбайт = 2²³ бит
1 Гбайт = 2³³ бит
байт/с → бит/с : × 8
```

## Тип 1 — Количество снимков в пакете

**Алгоритм:**
```
1. i = log₂(N)              — глубина цвета
2. V_снимка = h × w × i     — вес одного снимка (бит)
3. V_пакета = q × t         — максимальный вес пакета (бит)
4. Снимков = V_пакета // V_снимка   — округление ВНИЗ (не более)
```

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 702 — Тип 1 (разбор) — снимки с камеры видеофиксации</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Прибор автоматической фиксации нарушений правил дорожного движения делает цветные фотографии размером <strong>1024×768 пикселей</strong>, используя палитру из <strong>4096 цветов</strong>. Снимки сохраняются в памяти камеры, группируются в пакеты по несколько штук, затем передаются в центр обработки информации со скоростью передачи данных <strong>1 310 720 бит/с</strong>. Определите максимально возможное количество снимков в одном пакете, если на передачу одного пакета отводится <strong>не более 300 секунд</strong>.<br>
В ответ запишите целое число.<br><br>
<strong>Разбор:</strong> i = log₂(4096) = 12 бит. V_снимка = 1024×768×12 = 9 437 184 бит. V_пакета = 1 310 720×300 = 393 216 000 бит. Снимков = 393 216 000 // 9 437 184 = 41,6 → <strong>41</strong> (округляем вниз, т.к. «не более»).</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 41</div></div>

````python
from math import log2

h, w = 768, 1024
colors = 4096
speed = 1_310_720      # бит/с
time = 300              # секунд

i = log2(colors)                    # глубина цвета = 12 бит
V_img  = h * w * i                  # вес одного снимка
V_pack = speed * time               # макс. вес пакета

print(int(V_pack // V_img))         # Ответ: 41
````

Размер каждой фотографии <strong>1024×64 пикселя</strong>, при этом используется палитра из <strong>4096 цветов</strong>. Снимки сохраняются в памяти камеры, группируются по несколько штук, а затем передаются в центр охраны природы со скоростью передачи данных <strong>655 360 бит/с</strong>. Каково максимально возможное количество снимков в одном пакете, если на передачу одного пакета отводится <strong>не более 70 секунд</strong>.<br>
В ответе запишите целое число.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 58</div></div>

````python
from math import log2

V_img  = 1024 * 64 * log2(4096)    # вес снимка
V_pack = 655_360 * 70               # вес пакета

print(int(V_pack // V_img))         # Ответ: 58
````

## Тип 2 — Максимальное/минимальное количество цветов

**Максимальное N:** `N = 2^i`, где `i = memory_bits // pixels` (вниз)

**Минимальное N:** `N = 2^(i-1) + 1` — первое значение, которому нужно i бит

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 711 — Тип 2а (разбор) — максимальное кол-во цветов</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для хранения произвольного растрового изображения размером <strong>2560×5040 пикселей</strong> отведено <strong>14 175 Кбайт</strong> памяти без учёта размера заголовка файла. Для кодирования цвета каждого пикселя используется одинаковое количество бит, коды пикселей записываются в файл один за другим без промежутков. Какое <strong>максимальное количество цветов</strong> можно использовать в изображении?<br><br>
<strong>Разбор:</strong> 14175 Кбайт = 14175 × 2¹³ бит. Пикселей = 2560×5040. i = bits // pixels. N = 2^i.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 512</div></div>

````python
res   = 2560 * 5040
bits  = 14_175 * 2**13             # Кбайт → биты

i      = bits // res               # глубина (вниз!)
colors = 2 ** i
print(colors)                       # Ответ: 512
````

Каково <strong>минимальное</strong> возможное количество цветов в палитре изображения, если на передачу одного пакета отводится <strong>более 270 секунд</strong>?<br>

N = 2^(i-1)+1.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 257</div></div>

## Часть 2 — Звук

### Теория: дискретизация

Звук — аналоговый сигнал. Для хранения на компьютере его **дискретизируют**: измеряют уровень сигнала через равные промежутки времени.

- **Частота дискретизации (η)** — количество измерений в секунду (Гц)
- **Глубина кодирования (i)** — количество бит на одно измерение
- **Каналы (k):** моно = 1, стерео = 2, квадро = 4

**Формула объёма звукового файла:**
```
V = k × t × i × η   (бит)
```

| Тип задачи | Алгоритм |
|:---|:---|
| Время передачи | `t = V // скорость` |
| Перезапись с другими параметрами | `V₂ = V₁ × (изм. каналов) × (изм. частоты) × (изм. глубины)` |


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 700 — Звук — Тип 1 (разбор): время скачивания альбома</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Музыкальный альбом записан в формате <strong>стерео</strong> с частотой дискретизации <strong>48 кГц</strong> и разрешением <strong>34 бит</strong> без использования сжатия. В альбоме <strong>13 треков</strong> общей длительностью <strong>42 минуты 20 секунд</strong>. Каждый трек содержит заголовок размером <strong>110 Кбайт</strong>. Сколько секунд потребуется для скачивания альбома по каналу со скоростью передачи данных <strong>314 572 800 бит/с</strong>?<br>
В ответе укажите целую часть числа.<br><br>
<strong>Разбор:</strong> V_музыки = 2×(42×60+20)×34×48000. V_заголовков = 13×110×2¹³. V_total = V_музыки + V_заголовков. t = V_total // скорость.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 26</div></div>

````python
k    = 2                             # стерео
eta  = 48 * 10**3                   # 48 кГц → Гц
i    = 34                           # бит (разрешение)
t    = 42 * 60 + 20                 # секунды

bitrate = 314_572_800
header  = 110 * 2**13              # Кбайт → биты, 13 треков

V_music = k * t * i * eta
V_total = V_music + header * 13

print(V_total // bitrate)           # Ответ: 26
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 718 — Звук — Тип 1 (пример): время передачи голосового</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Голосовое сообщение продолжительностью <strong>120 с</strong> было записано в формате <strong>стерео</strong> и оцифровано с глубиной кодирования <strong>16 бит</strong> и частотой дискретизации <strong>56 000 измерений в секунду</strong>. Сжатие данных не использовалось. Файл был передан по каналу связи с пропускной способностью <strong>32 000 бит/с</strong>. Сколько секунд длилась передача файла?<br>
