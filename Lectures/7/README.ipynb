## Часть 1 — Изображения

**Формулы:**

| Формула | Описание |
|:---|:---|
| `i = log₂(N)` | Глубина цвета (бит/пиксель) по количеству цветов N |
| `N = 2ⁱ` | Количество цветов по глубине цвета |
| `V = h × w × i` | Объём изображения (бит) |
| `V = q × t` | Объём = скорость × время |

```python
from math import log2, floor

# Объём изображения:
V = 1024 * 768 * log2(4096)    # пиксели × глубина цвета

# Количество снимков в пакете:
снимков = floor(V_пакета / V_снимка)

# Мин. кол-во цветов: 2^(i-1)+1
# Макс. кол-во цветов: 2^i
```

<div class="task-box">
<div class="task-title">📋 Задание 702 — Количество снимков в пакете | Ответ: <strong style="color:#2fca72;">41</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Прибор автоматической фиксации нарушений правил дорожного движения делает цветные фотографии размером 1024×768 пикселей, используя палитру из 4096 цветов. Снимки сохраняются в памяти камеры, группируются в пакеты по несколько штук, затем передаются в центр обработки информации со скоростью передачи данных 1&nbsp;310&nbsp;720 бит/с. Определите максимально возможное количество снимков в одном пакете, если на передачу одного пакета отводится <strong>не более 300 секунд</strong>.<br>
В ответ запишите целое число.</div>
</div>

````python
from math import log2

res = 1024 * 768
bitrate = 1_310_720
colors = 4096
time = 300

pixel_size = log2(colors)       # = 12 бит
img_size = res * pixel_size     # вес одного снимка
package_size = bitrate * time   # максимальный вес пакета

print(int(package_size // img_size))  # Ответ: 41
````

<div class="task-title">📋 Задание 704 — Минимальное количество цветов | Ответ: <strong style="color:#2fca72;">257</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Прибор автоматической фиксации нарушений правил дорожного движения делает цветные фотографии размером 1280×720 пикселей. Снимки сохраняются в памяти камеры, группируются в пакеты по 69 штук, затем передаются в центр обработки информации со скоростью передачи данных 245&nbsp;760 байт/с. Каково минимальное возможное количество цветов в палитре изображения, если на передачу одного пакета отводится <strong>более 270 секунд</strong>?<br>
В ответе запишите целое число.</div>
</div>

````python
from math import ceil

speed = 245_760 * 8    # байт/с → бит/с
V_pkt = speed * 270    # минимальный вес пакета (более 270 сек)
V1 = ceil(V_pkt / 69)  # вес одного снимка (вверх — «более»)
i = ceil(V1 / (1280 * 720))  # вес пикселя (вверх)
# Минимальное количество цветов: 2^(i-1)+1
print(2**(i-1) + 1)    # Ответ: 257
````

<div style="color:#c8d8e8;line-height:1.7;">Фотограф делает цветные фотографии размером 7680×4320 пикселей, используя палитру из 2¹⁶ цветов. Для сохранения снимков фотограф использует сменные карты памяти, каждая из которых вмещает не более 9 Гбайт данных. Когда на карте памяти остаётся недостаточно места для записи новой фотографии, фотограф берёт следующую, свободную карту. Известно, что фотограф сделал 4010 снимков. <strong>Сколько снимков оказалось на последней карте памяти из использованных?</strong><br>
В ответе запишите целое число.</div>
</div>

````python
from math import floor

img = 7680 * 4320 * 16           # глубина = 16 бит (2^16 цветов → 16 бит)
card = 9 * 2**33                  # 9 Гбайт → биты

per_card = floor(card / img)
print(4010 % per_card)           # Ответ: 95
````

## Часть 2 — Звук

**Формула объёма звукового файла:**

```
V = k × t × i × η
```

| Параметр | Описание |
|:---|:---|
| k | Количество каналов: моно=1, стерео=2, квадро=4 |
| t | Длительность (секунды) |
| i | Глубина кодирования (бит, разрешение) |
| η | Частота дискретизации (Гц) |

<div class="task-box">
<div class="task-title">📋 Задание 700 — Время скачивания альбома | Ответ: <strong style="color:#2fca72;">26</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Музыкальный альбом записан в формате стерео с частотой дискретизации 48 кГц и разрешением 34 бит без использования сжатия. В альбоме 13 треков общей длительностью 42 минуты 20 секунд. Каждый трек содержит заголовок размером 110 Кбайт. <strong>Сколько секунд потребуется для скачивания альбома по каналу со скоростью передачи данных 314&nbsp;572&nbsp;800 бит/с? В ответе укажите целую часть числа.</strong></div>
</div>

````python
k = 2                           # стерео
eta = 48 * 10**3               # 48 кГц
i = 34                         # бит
t = 42 * 60 + 20               # секунды

bitrate = 314_572_800
header = 110 * 2**13           # Кбайт → биты, 13 треков

V_music = k * t * i * eta
V_total = V_music + header * 13

print(V_total // bitrate)      # Ответ: 26
````
