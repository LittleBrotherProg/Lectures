# –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á —Å —á–µ—Ä–µ–¥–æ–≤–∞–Ω–∏–µ–º —á–µ—Ç–Ω–æ—Å—Ç–∏ —Ü–∏—Ñ—Ä

## –£—Å–ª–æ–≤–∏–µ –∑–∞–¥–∞—á–∏
–°–∫–æ–ª—å–∫–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –¥–µ—Å—è—Ç–∏—á–Ω—ã—Ö —á–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞—á–Ω—ã—Ö —á–∏—Å–µ–ª, –≤ –∫–æ—Ç–æ—Ä—ã—Ö:
- –í—Å–µ —Ü–∏—Ñ—Ä—ã —Ä–∞–∑–ª–∏—á–Ω—ã
- –ù–∏–∫–∞–∫–∏–µ –¥–≤–µ —á—ë—Ç–Ω—ã–µ –∏–ª–∏ –¥–≤–µ –Ω–µ—á—ë—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã –Ω–µ —Å—Ç–æ—è—Ç —Ä—è–¥–æ–º

## –ü–æ—á–µ–º—É –Ω–µ –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ä—É—á–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ?

!> **–í–ê–ñ–ù–û:** –†—É—á–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏ –∫—Ä–∞–π–Ω–µ –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –∏–∑-–∑–∞ —É—Å–ª–æ–≤–∏—è **"–≤—Å–µ —Ü–∏—Ñ—Ä—ã —Ä–∞–∑–ª–∏—á–Ω—ã"**. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –Ω–∞–º –Ω—É–∂–Ω–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –Ω–µ –ø—Ä–æ—Å—Ç–æ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ —Ü–∏—Ñ—Ä, –∞ **–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π**, —á—Ç–æ –≤—Ä—É—á–Ω—É—é —Å—á–∏—Ç–∞—Ç—å –æ—á–µ–Ω—å —Å–ª–æ–∂–Ω–æ –∏ –ª–µ–≥–∫–æ –æ—à–∏–±–∏—Ç—å—Å—è.

!> **–ü–†–ê–ö–¢–ò–ß–ï–°–ö–ò–ô –í–´–í–û–î:** –î–ª—è –∑–∞–¥–∞—á —Å —É—Å–ª–æ–≤–∏–µ–º "–≤—Å–µ —Ü–∏—Ñ—Ä—ã —Ä–∞–∑–ª–∏—á–Ω—ã" –ø—Ä–æ–≥—Ä–∞–º–º–Ω—ã–µ –º–µ—Ç–æ–¥—ã (—Ü–∏–∫–ª—ã –∏–ª–∏ itertools) –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –Ω–∞–¥–µ–∂–Ω–µ–µ —Ä—É—á–Ω—ã—Ö —Ä–∞—Å—á–µ—Ç–æ–≤.

## 1. –†–µ—à–µ–Ω–∏–µ —á–µ—Ä–µ–∑ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã

### –ö–æ–¥ —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è–º–∏

```python
# –°—á–µ—Ç—á–∏–∫ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö —á–∏—Å–µ–ª
count = 0

# –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤—Å–µ —á–µ—Ç—ã—Ä–µ—Ö–∑–Ω–∞—á–Ω—ã–µ –¥–µ—Å—è—Ç–∏—á–Ω—ã–µ —á–∏—Å–ª–∞ (–æ—Ç 1000 –¥–æ 9999)
# –ü–µ—Ä–≤–∞—è —Ü–∏—Ñ—Ä–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å 0, –ø–æ—ç—Ç–æ–º—É '123456789'
for d1 in '123456789':  # –ø–µ—Ä–≤–∞—è —Ü–∏—Ñ—Ä–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å 0
    # –í—Ç–æ—Ä–∞—è —Ü–∏—Ñ—Ä–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –ª—é–±–æ–π, –≤–∫–ª—é—á–∞—è 0
    for d2 in '0123456789':
        # –¢—Ä–µ—Ç—å—è —Ü–∏—Ñ—Ä–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –ª—é–±–æ–π, –≤–∫–ª—é—á–∞—è 0
        for d3 in '0123456789':
            # –ß–µ—Ç–≤–µ—Ä—Ç–∞—è —Ü–∏—Ñ—Ä–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –ª—é–±–æ–π, –≤–∫–ª—é—á–∞—è 0
            for d4 in '0123456789':
                # –°–æ–±–∏—Ä–∞–µ–º —Ü–∏—Ñ—Ä—ã –≤ —Å—Ç—Ä–æ–∫—É (—á–∏—Å–ª–æ)
                number = d1 + d2 + d3 + d4
                # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –≤—Å–µ —Ü–∏—Ñ—Ä—ã —Ä–∞–∑–ª–∏—á–Ω—ã
                # set(number) —Å–æ–∑–¥–∞–µ—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∏–∑ —Ü–∏—Ñ—Ä, —É–¥–∞–ª—è—è –ø–æ–≤—Ç–æ—Ä—ã
                # len(set(number)) == 4 –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤—Å–µ 4 —Ü–∏—Ñ—Ä—ã —É–Ω–∏–∫–∞–ª—å–Ω—ã
                if len(set(number)) == 4:
                    # –°—á–µ—Ç—á–∏–∫ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —á–µ—Ä–µ–¥–æ–≤–∞–Ω–∏—è —á–µ—Ç–Ω–æ—Å—Ç–∏
                    # done –±—É–¥–µ—Ç —Å—á–∏—Ç–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä, –≥–¥–µ —á–µ—Ç–Ω–æ—Å—Ç—å —Ä–∞–∑–Ω–∞—è
                    done = 0 
                    # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ç—Ä–∏ –ø–∞—Ä—ã —Å–æ—Å–µ–¥–Ω–∏—Ö —Ü–∏—Ñ—Ä: (0-1), (1-2), (2-3)
                    for i in range(3):
                        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–∏–º–≤–æ–ª—ã –≤ —á–∏—Å–ª–∞
                        digit1 = int(number[i])
                        digit2 = int(number[i+1])
                        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —Ü–∏—Ñ—Ä—ã –∏–º–µ—é—Ç —Ä–∞–∑–Ω—É—é —á–µ—Ç–Ω–æ—Å—Ç—å
                        # (–æ–¥–Ω–∞ —á–µ—Ç–Ω–∞—è, –¥—Ä—É–≥–∞—è –Ω–µ—á–µ—Ç–Ω–∞—è)
                        if (digit1 % 2 != 0 and digit2 % 2 == 0) or (digit1 % 2 == 0 and digit2 % 2 != 0):
                            # –ï—Å–ª–∏ —á–µ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞–∑–Ω—ã–µ, —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫
                            done += 1
                    # –ï—Å–ª–∏ –≤—Å–µ —Ç—Ä–∏ –ø–∞—Ä—ã –∏–º–µ—é—Ç —Ä–∞–∑–Ω—É—é —á–µ—Ç–Ω–æ—Å—Ç—å (done == 3)
                    if done == 3:
                        # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º –æ–±—â–∏–π —Å—á–µ—Ç—á–∏–∫ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö —á–∏—Å–µ–ª
                        count += 1

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç - –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö —á–∏—Å–µ–ª
print(count)
```

## –û–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è

```python
count = 0

for d1 in '123456789':
    for d2 in '0123456789':
        for d3 in '0123456789':
            for d4 in '0123456789':
                number = d1 + d2 + d3 + d4
                if len(set(number)) == 4:
                    done = 0 
                    for i in range(3):
                        digit1 = int(number[i])
                        digit2 = int(number[i+1])
                        if (digit1 % 2 != 0 and digit2 % 2 == 0) or (digit1 % 2 == 0 and digit2 % 2 != 0):
                            done += 1
                    if done == 3:
                        count += 1

print(count)
```

## –≠–ª–µ–≥–∞–Ω—Ç–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ —á–µ—Ä–µ–∑ itertools

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –≤—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –º–æ–¥—É–ª—è itertools
from itertools import *

# –ó–∞–¥–∞–µ–º –∞–ª—Ñ–∞–≤–∏—Ç - –≤—Å–µ —Ü–∏—Ñ—Ä—ã –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã
alp = '0123456789'

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—Å–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ (permutations) –¥–ª–∏–Ω–æ–π 4 –∏–∑ —Ü–∏—Ñ—Ä –∞–ª—Ñ–∞–≤–∏—Ç–∞
# –ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É—é—Ç, —á—Ç–æ –≤—Å–µ —Ü–∏—Ñ—Ä—ã –≤ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ —Ä–∞–∑–ª–∏—á–Ω—ã
combs = permutations(alp, r=4)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö —á–∏—Å–µ–ª
count = 0

# –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤—Å–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
for comb in combs:
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–æ—Ä—Ç–µ–∂ —Å–∏–º–≤–æ–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫—É
    # –ù–∞–ø—Ä–∏–º–µ—Ä: ('1','2','3','4') ‚Üí "1234"
    comb = ''.join(comb)
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —á–∏—Å–ª–æ –Ω–µ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å 0 (–ø–µ—Ä–≤–∞—è —Ü–∏—Ñ—Ä–∞ –Ω–µ '0')
    if comb[0] not in ['0']:
        # –ó–∞–º–µ–Ω—è–µ–º –≤—Å–µ –Ω–µ—á–µ—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã –Ω–∞ —Å–∏–º–≤–æ–ª '@'
        # –ù–µ—á–µ—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã: 1,3,5,7,9
        for nechet in ['1','3','5','7','9']:
            comb = comb.replace(nechet,'@')
        
        # –ó–∞–º–µ–Ω—è–µ–º –≤—Å–µ —á–µ—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã –Ω–∞ —Å–∏–º–≤–æ–ª '$'
        # –ß–µ—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã: 0,2,4,6,8
        for nechet in ['0','2','4','6','8']:
            comb = comb.replace(nechet,'$')
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –Ω–µ—Ç –¥–≤—É—Ö –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –ø–æ–¥—Ä—è–¥
        # '@@' - –¥–≤–µ –Ω–µ—á–µ—Ç–Ω—ã–µ –ø–æ–¥—Ä—è–¥, '$$' - –¥–≤–µ —á–µ—Ç–Ω—ã–µ –ø–æ–¥—Ä—è–¥
        if comb.count('@@') == 0 and comb.count('$$') == 0:
            # –ï—Å–ª–∏ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫
            count += 1

# –í—ã–≤–æ–¥–∏–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö —á–∏—Å–µ–ª
print(count)
```

## –í–µ—Ä—Å–∏—è –∫–æ–¥–∞ –±–µ–∑ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤

```python
from itertools import *
alp = '0123456789'
combs = permutations(alp,r=4)
count = 0
for comb in combs:
    comb = ''.join(comb)
    if comb[0] not in ['0']:
        for nechet in ['1','3','5','7','9']:
            comb = comb.replace(nechet,'@')
        for nechet in ['0','2','4','6','8']:
            comb = comb.replace(nechet,'$')
        if comb.count('@@') == 0 and comb.count('$$') == 0:
                count += 1
print(count) 
```

## üîç –ê–Ω–∞–ª–∏–∑ —Ä–µ—à–µ–Ω–∏—è:

–í –¥–∞–Ω–Ω–æ–º —Ä–µ—à–µ–Ω–∏–∏ –º—ã –ø—Ä–∏–º–µ–Ω–∏–ª–∏ –¥–≤–∞ –∫–ª—é—á–µ–≤—ã—Ö –ø–æ–¥—Ö–æ–¥–∞:

1.  **–ú–µ—Ç–æ–¥ `permutations()` –∏–∑ –º–æ–¥—É–ª—è `itertools`**: –≠—Ç–æ—Ç –º–µ—Ç–æ–¥ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –∑–∞–¥–∞–Ω–Ω–æ–π –¥–ª–∏–Ω—ã (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ 4 —Ü–∏—Ñ—Ä—ã), –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É—è, —á—Ç–æ **–≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã**. –≠—Ç–æ –∏–¥–µ–∞–ª—å–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —É—Å–ª–æ–≤–∏—é ¬´–≤—Å–µ —Ü–∏—Ñ—Ä—ã —Ä–∞–∑–ª–∏—á–Ω—ã¬ª –∏ –∏–∑–±–∞–≤–ª—è–µ—Ç –Ω–∞—Å –æ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –ø—Ä–æ–≤–µ—Ä—è—Ç—å —ç—Ç–æ —É—Å–ª–æ–≤–∏–µ –≤—Ä—É—á–Ω—É—é.

2.  **–ú–µ—Ç–æ–¥ `replace()` –¥–ª—è —Å—Ç—Ä–æ–∫**: –≠—Ç–æ—Ç –º–µ—Ç–æ–¥ –ø–æ–∑–≤–æ–ª—è–µ—Ç –º–∞—Å—Å–æ–≤–æ –∑–∞–º–µ–Ω—è—Ç—å –æ–¥–Ω–∏ —Å–∏–º–≤–æ–ª—ã –≤ —Å—Ç—Ä–æ–∫–µ –Ω–∞ –¥—Ä—É–≥–∏–µ. –ú—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ –µ–≥–æ –¥–ª—è —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ **¬´–Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏¬ª** –∏–ª–∏ —É–ø—Ä–æ—â–µ–Ω–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —É—Å–ª–æ–≤–∏—è:
    *   –í—Å–µ –Ω–µ—á—ë—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã (`1, 3, 5, 7, 9`) –∑–∞–º–µ–Ω—è—é—Ç—Å—è –Ω–∞ —Å–∏–º–≤–æ–ª `@`
    *   –í—Å–µ —á—ë—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã (`0, 2, 4, 6, 8`) –∑–∞–º–µ–Ω—è—é—Ç—Å—è –Ω–∞ —Å–∏–º–≤–æ–ª `$`

    **–ü—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ:** –ü–æ—Å–ª–µ —Ç–∞–∫–æ–π –∑–∞–º–µ–Ω—ã —Å–ª–æ–∂–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ ¬´–Ω–∏–∫–∞–∫–∏–µ –¥–≤–µ —á—ë—Ç–Ω—ã–µ –∏–ª–∏ –¥–≤–µ –Ω–µ—á—ë—Ç–Ω—ã–µ —Ü–∏—Ñ—Ä—ã –Ω–µ —Å—Ç–æ—è—Ç —Ä—è–¥–æ–º¬ª –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç—Å—è –≤ –ø—Ä–æ—Å—Ç–æ–µ –∏ –Ω–∞–≥–ª—è–¥–Ω–æ–µ: **–ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ –≤ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–µ—Ç –ø–æ–¥—Å—Ç—Ä–æ–∫ `@@` –∏–ª–∏ `$$`**. –≠—Ç–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —É–ø—Ä–æ—â–∞–µ—Ç –ª–æ–≥–∏–∫—É –∫–æ–¥–∞ –∏ —Å–Ω–∏–∂–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—à–∏–±–∫–∏.

–û—Å–≤–æ–∏–ª–∏ –º–æ—â–Ω—É—é —Ç–µ—Ö–Ω–∏–∫—É –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Å–ª–æ–∂–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π –∫–æ–º–±–∏–Ω–∞—Ç–æ—Ä–∏–∫–∏ –≤ –ø—Ä–æ—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –ø—Ä–æ–≤–µ—Ä–∫–∏ —Å –ø–æ–º–æ—â—å—é –∑–∞–º–µ–Ω—ã —Å–∏–º–≤–æ–ª–æ–≤. –°–æ—á–µ—Ç–∞–Ω–∏–µ `permutations()` –¥–ª—è –≥–∞—Ä–∞–Ω—Ç–∏–∏ —É–Ω–∏–∫–∞–ª—å–Ω–æ—Å—Ç–∏ –∏ `replace()` –¥–ª—è –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏ —É—Å–ª–æ–≤–∏—è —á–µ—Ä–µ–¥–æ–≤–∞–Ω–∏—è –¥–µ–ª–∞–µ—Ç –∫–æ–¥ —á–∏—â–µ, –ø–æ–Ω—è—Ç–Ω–µ–µ –∏ –Ω–∞–¥–µ–∂–Ω–µ–µ.
