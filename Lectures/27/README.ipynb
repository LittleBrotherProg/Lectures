## Ключевые термины

### Евклидово расстояние

Расстояние между точками A(x₁, y₁) и B(x₂, y₂) — длина отрезка между ними на плоскости:

```
d(A, B) = √((x₂−x₁)² + (y₂−y₁)²)
```

Python уже умеет считать это расстояние — функция `dist()` из модуля `math`:

```python
from math import dist

point_A = (1, 0)
point_B = (5, 3)
print(dist(point_A, point_B))  # 5.0
```

### Медоида кластера

**Медоида** — реальная точка из набора данных, у которой **минимальна сумма расстояний** до всех остальных точек кластера. В отличие от абстрактного центроида, медоида всегда является одной из исходных точек данных.

**Подробный разбор функции поиска медоиды:**

```python
def get_medoid(cluster):
    # min() находит минимальный элемент в списке cluster
    # key= задаёт правило: по чему именно сравнивать элементы
    # lambda p: — для каждой точки p из кластера
    # sum(dist(p, q) for q in cluster) — считаем сумму расстояний
    #   от этой точки p до каждой другой точки q кластера
    # Та точка p, у которой эта сумма минимальна — и есть медоида
    return min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))
```

### Формирование ответа

```python
# medoids — список медоид всех кластеров, каждая медоида = (x, y)

medoids = [get_medoid(cluster) for cluster in clusters]

# Среднее арифметическое абсцисс (x-координат) всех медоид:
px = sum(x for x, y in medoids) / len(medoids)

# Среднее арифметическое ординат (y-координат) всех медоид:
py = sum(y for x, y in medoids) / len(medoids)

# abs() — берём модуль (условие задачи)
# int() — берём целую часть (условие задачи)
# * 10_000 — умножаем на 10000 (условие задачи)
print(abs(int(px * 10_000)), abs(int(py * 10_000)))
```

## Что именно мы ищем и зачем?

### Зачем делить точки на кластеры?

Представь что у тебя есть карта звёздного неба с тысячами точек.
Невооружённым глазом видно что точки собираются в **группы** (созвездия, галактики).
Задача — **автоматически** определить к какой группе относится каждая точка.

Именно это и называется **кластеризацией** — разбивка множества точек на группы
(кластеры) так чтобы точки внутри одного кластера были близки друг к другу,
а точки из разных кластеров — далеки.

---

### Что конкретно мы ищем — пошагово

**Задача в трёх словах:** найти «центр» каждой группы точек.

Но не просто геометрический центр — а **медоиду**: реальную точку из данных,
которая ближе всего расположена ко всем остальным точкам своей группы.

| Шаг | Что делаем | Зачем |
|:---:|:---|:---|
| 1 | Читаем координаты всех точек из файла | Это наши исходные данные |
| 2 | Делим точки на кластеры (группы) | Отделяем «своих» от «чужих» |
| 3 | Для каждого кластера ищем медоиду | Находим «центр» каждой группы |
| 4 | Считаем среднее координат медоид | Итоговый ответ задачи |

---

### Что такое медоида и почему именно она?

Медоида — это **реальная точка** из набора данных с минимальной суммой расстояний до всех остальных точек своего кластера.

В отличие от обычного центра (среднее x и среднее y) — медоида всегда
**существует в данных**. Это важно когда нет точки точно в середине кластера.

```
Кластер: (1,1), (2,2), (3,3), (10,10)

Центроид (среднее): ((1+2+3+10)/4, (1+2+3+10)/4) = (4, 4) — такой точки нет в данных!

Медоида: проверяем каждую точку:
  (1,1)   → суммарное расстояние до остальных = 2.8 + 5.7 + 12.7 = 21.2
  (2,2)   → суммарное расстояние = 2.8 + 2.8 + 11.3 = 16.9
  (3,3)   → суммарное расстояние = 5.7 + 2.8 + 8.5  = 17.0
  (10,10) → суммарное расстояние = 12.7 + 11.3 + 8.5 = 32.5
  
  Минимум у (2,2) → медоида = (2,2) — реальная точка из данных!
```

Это и есть ответ задачи.</div>

### Почему формат ответа такой странный — умножение на 10000?

Координаты медоид могут быть дробными, например Px = 1.073812...

Чтобы не терять точность при записи ответа, условие задачи просит:
- Умножить на 10000: 1.073812 × 10000 = 10738.12
- Взять целую часть: int(10738.12) = 10738
- Взять модуль: abs(10738) = 10738

Так два числа с 4 знаками после запятой превращаются в два целых числа.

```python
print(abs(int(px * 10_000)), abs(int(py * 10_000)))
```

**Самый простой** алгоритм. Смотрим на точечную диаграмму, находим прямые, разделяющие кластеры, и в коде проверяем, по какую сторону от прямой лежит каждая точка.

**Алгоритм:**
1. Визуализировать точки из файла (в LibreOffice Calc или Excel → Вставка → Диаграмма → Точечная)
2. Найти прямые-разделители (вертикальные x = c, горизонтальные y = c, или наклонные y = ax + b)
3. Для каждой точки проверить условие и отнести в нужный кластер

**Шаг 2 — Вывод:** точки с x < 1 → первый кластер, точки с x > 1 → второй кластер.

````python
from math import dist


def get_medoid(cluster):
    # Ищем точку с минимальной суммой расстояний до всех остальных
    return min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))


# Создаём два пустых списка — по одному на каждый кластер
clusters = [[], []]

# Открываем файл А (with автоматически закрывает файл после блока)
with open('2702_A.txt') as file:
    # next(file) пропускает первую строку заголовка "X Y"
    next(file)
    # Читаем файл построчно
    for line in file:
        # split() разбивает строку "1.5 3.2" по пробелам → ["1.5", "3.2"]
        # map(float, ...) применяет float() к каждому элементу → 1.5, 3.2
        x, y = map(float, line.split())
        # Проверяем условие разделения по прямой x = 1
        if x < 1:
            clusters[0].append((x, y))  # точка левее → первый кластер
        else:
            clusters[1].append((x, y))  # точка правее → второй кластер

# Для каждого кластера находим его медоиду
medoids = [get_medoid(cluster) for cluster in clusters]

# Среднее арифметическое абсцисс медоид
px = sum(x for x, y in medoids) / len(medoids)
# Среднее арифметическое ординат медоид
py = sum(y for x, y in medoids) / len(medoids)

# Выводим ответ: абс. значение целой части произведения на 10000
print(abs(int(px * 10_000)), abs(int(py * 10_000)))  # 10738 30730
````

Точки с y < 3 → нижний кластер, y > 7 → верхний, всё остальное → средний.

````python
from math import dist


def get_medoid(cluster):
    return min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))


# Три пустых списка — по одному на каждый из трёх кластеров
clusters = [[], [], []]

with open('2702_B.txt') as file:
    next(file)  # пропускаем заголовок
    for line in file:
        x, y = map(float, line.split())
        # Первое условие: ниже прямой y = 3
        if y < 3:
            clusters[0].append((x, y))   # нижний кластер
        # Второе условие: выше прямой y = 7
        elif y > 7:
            clusters[1].append((x, y))   # верхний кластер
        # Всё остальное (3 ≤ y ≤ 7) — средний кластер
        else:
            clusters[2].append((x, y))   # средний кластер

medoids = [get_medoid(cluster) for cluster in clusters]

px = sum(x for x, y in medoids) / len(medoids)
py = sum(y for x, y in medoids) / len(medoids)
print(abs(int(px * 10_000)), abs(int(py * 10_000)))  # 37522 51277
````

## Часть 2 — Алгоритм k-средних

**Промежуточный** алгоритм. Не нужно вручную искать разделяющие прямые — алгоритм сам находит медоиды. Нужно только задать **начальные точки** (по одной на каждый кластер).

**Принцип работы:**
1. Выбираем начальные точки (одну в каждом предполагаемом кластере)
2. Каждую точку данных относим к **ближайшей** начальной точке
3. Вычисляем новые **медоиды** для каждого сформированного кластера
4. Повторяем шаги 2–3, пока медоиды **не перестанут двигаться** (стабилизируются)

<div style="background:#1c1700;border-left:3px solid #f59e0b;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#f59e0b;">Важно:</strong> начальные точки нужно выбирать удалёнными друг от друга и близкими к предполагаемым центрам кластеров — иначе алгоритм может разбить данные неправильно.</div>

---

## Подробный разбор трёх функций алгоритма

````python
from math import dist

# ──────────────────────────────────────────────────────────────
# ФУНКЦИЯ 1: assign_clusters
# Назначает каждую точку к ближайшему центру (медоиде)
# ──────────────────────────────────────────────────────────────
def assign_clusters(data, centers):
    # labels — список меток: labels[i] = номер кластера для точки data[i]
    labels = []

    # Перебираем каждую точку p из набора данных
    for p in data:
        # Для каждого центра считаем расстояние от точки p до него
        # Результат: список расстояний, по одному на каждый центр
        # Например: [5.2, 1.3, 8.7] для трёх центров
        distances = [dist(p, center) for center in centers]

        # min(distances) — минимальное расстояние (1.3 из примера)
        # .index(...) — индекс этого минимума (1 — второй центр)
        # Этот индекс и есть номер кластера для точки p
        labels.append(distances.index(min(distances)))

    # Возвращаем список меток для всех точек
    return labels


# ──────────────────────────────────────────────────────────────
# ФУНКЦИЯ 2: update_centers
# Находит новую медоиду для каждого кластера
# ──────────────────────────────────────────────────────────────
def update_centers(data, labels, k):
    # new_centers — список новых медоид (будет заполнен в цикле)
    new_centers = []

    # Перебираем каждый кластер по его номеру (0, 1, ..., k-1)
    for cluster_label in range(k):
        # Собираем все точки, принадлежащие текущему кластеру:
        # zip(data, labels) соединяет точки с их метками попарно
        # if l == cluster_label — берём только точки нужного кластера
        cluster = [p for p, l in zip(data, labels) if l == cluster_label]

        # Находим медоиду: точку с минимальной суммой расстояний до остальных
        # lambda p: — для каждой точки-кандидата p
        # sum(dist(p, q) for q in cluster) — сумма расстояний до всех точек кластера
        medoid = min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))

        # Добавляем найденную медоиду в список новых центров
        new_centers.append(medoid)

    return new_centers


# ──────────────────────────────────────────────────────────────
# ФУНКЦИЯ 3: k_means
# Запускает итерации до стабилизации медоид
# ──────────────────────────────────────────────────────────────
def k_means(data, k):
    # Начальные точки выбираем вручную по диаграмме!
    # Для задания 2703 файл А: два кластера, точки выбраны по диаграмме
    centers = [[-9, -2], [-1, 15]]

    # new_centers изначально пустой — гарантирует что цикл запустится
    # (пустой список не равен centers)
    new_centers = []

    # Цикл продолжается пока new_centers не совпадёт с centers
    # То есть пока медоиды ещё двигаются (алгоритм не сошёлся)
    while new_centers != centers:
        # Шаг 1: назначаем каждую точку к ближайшему центру
        labels = assign_clusters(data, centers)

        # Шаг 2: пересчитываем медоиды для новых кластеров
        new_centers = update_centers(data, labels, k)

        # Обновляем центры для следующей итерации
        centers = new_centers

    # Возвращаем итоговые метки кластеров и медоиды
    return labels, centers
````

### Почему алгоритм находит правильный ответ даже при неточных начальных точках?

Потому что на каждой итерации:
1. Все точки **переназначаются** к ближайшему центру
2. Центры **пересчитываются** как медоиды новых кластеров
3. И так до тех пор, пока центры не стабилизируются

Даже если начальные точки стоят немного не там — после нескольких итераций
центры «переползут» к реальным медоидам кластеров.

````python
from math import dist


def assign_clusters(data, centers):
    labels = []
    for p in data:
        distances = [dist(p, center) for center in centers]
        labels.append(distances.index(min(distances)))
    return labels


def update_centers(data, labels, k):
    new_centers = []
    for cluster_label in range(k):
        cluster = [p for p, l in zip(data, labels) if l == cluster_label]
        medoid = min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))
        new_centers.append(medoid)
    return new_centers


def k_means(data, k):
    # Для файла А: centers = [[-5, -7], [5, 7]]     (два кластера)
    # Для файла Б: centers = [[-11, -5], [3, -11], [3, 3]]  (три кластера)
    centers = [[-5, -7], [5, 7]]      # ← меняем для файла Б
    new_centers = []
    while new_centers != centers:
        labels = assign_clusters(data, centers)
        new_centers = update_centers(data, labels, k)
        centers = new_centers
    return labels, centers


# Для файла Б меняем имя файла и k=3
with open('2704_A.txt') as file:   # ← меняем имя файла
    next(file)
    data = [list(map(float, line.split())) for line in file]

labels, centers = k_means(data, 2)  # ← меняем k для файла Б

px = sum(x for x, y in centers) / len(centers)
py = sum(y for x, y in centers) / len(centers)
print(abs(int(px * 10_000)), abs(int(py * 10_000)))
# Файл А: 10592 6300
# Файл Б: 15981 37287
````

## Часть 3 — Алгоритм DBSCAN

**Самый мощный** алгоритм. Работает с кластерами **любой формы** (включая серповидные) и автоматически определяет **выбросы**. Нужно задать только один параметр — **eps** (радиус соседства точек).

**Принцип работы:**
1. Берём произвольную точку и начинаем новый кластер
2. Находим всех её соседей в радиусе eps (расстояние до них < eps)
3. Для каждого соседа тоже ищем его соседей и расширяем кластер
4. Продолжаем до тех пор, пока кластер нельзя расширить
5. Берём следующую необработанную точку → новый кластер

**Как выбрать eps по диаграмме:**
- eps должен быть **больше** расстояния между самыми дальними соседями внутри кластера
- eps должен быть **меньше** расстояния между точками разных кластеров
- Смотрим на диаграмму и подбираем значение визуально

## Подробный разбор функции dbscan

````python
from math import dist


def dbscan(data, eps):
    # clusters — список всех найденных кластеров
    # Каждый кластер — список точек
    clusters = []

    # Продолжаем пока в data остаются необработанные точки
    while data:
        # pop() берёт последнюю точку из data И удаляет её из списка
        # Эта точка станет первой в новом кластере
        point = data.pop()

        # Новый кластер начинается с одной точки
        cluster = [point]

        # queue — очередь точек, соседей которых ещё нужно проверить
        # Начинаем с той же точки
        queue = [point]

        # Продолжаем пока в очереди есть точки для проверки
        while queue:
            # pop(0) берёт первую точку из очереди (FIFO — как живая очередь)
            current = queue.pop(0)

            # Находим всех соседей current среди необработанных точек:
            # dist(current, p) < eps — расстояние меньше радиуса соседства
            neighbors = [p for p in data if dist(current, p) < eps]

            # Добавляем всех найденных соседей в текущий кластер
            cluster.extend(neighbors)

            # Добавляем соседей в очередь — их тоже нужно проверить
            # (у них тоже могут быть свои соседи, которые расширят кластер)
            queue.extend(neighbors)

            # Удаляем обработанных соседей из списка необработанных точек
            # data[:] = ... обновляет список на месте (не создаёт новый)
            # Оставляем только те точки, которые НЕ вошли в соседи
            data[:] = [p for p in data if p not in neighbors]

        # Кластер полностью сформирован (очередь опустела)
        # Добавляем его в общий список кластеров
        clusters.append(cluster)

    # Возвращаем список всех найденных кластеров
    return clusters
````

### Компактный вариант без комментариев

````python
from math import dist


def dbscan(data, eps):
    clusters = []
    while data:
        point = data.pop()
        cluster, queue = [point], [point]
        while queue:
            current = queue.pop(0)
            neighbors = [p for p in data if dist(current, p) < eps]
            cluster.extend(neighbors)
            queue.extend(neighbors)
            data[:] = [p for p in data if p not in neighbors]
        clusters.append(cluster)
    return clusters
````

````python
from math import dist


def get_medoid(cluster):
    return min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))


def dbscan(data, eps):
    clusters = []
    while data:
        point = data.pop()
        cluster, queue = [point], [point]
        while queue:
            current = queue.pop(0)
            neighbors = [p for p in data if dist(current, p) < eps]
            cluster.extend(neighbors)
            queue.extend(neighbors)
            data[:] = [p for p in data if p not in neighbors]
        clusters.append(cluster)
    return clusters


# ── Файл А: серповидные кластеры, eps = 0.3 ───────────────────
with open('2706_A.txt') as f:
    next(f)
    data = [tuple(map(float, line.split())) for line in f]

# eps = 0.3 подобрано по диаграмме:
# граничные соседи внутри кластера ~0.1, между кластерами ~0.4
clusters = dbscan(data, 0.3)
print([len(c) for c in clusters])       # проверяем количество кластеров

medoids = [get_medoid(cluster) for cluster in clusters]
px = sum(x for x, y in medoids) / len(medoids)
py = sum(y for x, y in medoids) / len(medoids)
print(abs(int(px * 10_000)), abs(int(py * 10_000)))  # 6267 15179

# ── Файл Б: eps = 0.1 ──────────────────────────────────────────
# with open('2706_B.txt') as f:
#     next(f)
#     data = [tuple(map(float, line.split())) for line in f]
# clusters = dbscan(data, 0.1)
# ...
# Ответ: 4927 819
````

````python
from math import dist


def get_medoid(cluster):
    return min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))


def dbscan(data, eps):
    clusters = []
    while data:
        point = data.pop()
        cluster, queue = [point], [point]
        while queue:
            current = queue.pop(0)
            # Ищем всех соседей в радиусе eps
            neighbors = [p for p in data if dist(current, p) < eps]
            cluster.extend(neighbors)
            queue.extend(neighbors)
            data[:] = [p for p in data if p not in neighbors]
        clusters.append(cluster)
    return clusters


with open('2710_A.txt') as f:
    next(f)
    data = [tuple(map(float, line.split())) for line in f]

# eps = 1 — выбросы далеко от кластеров, радиус 1 их не захватит
clusters = dbscan(data, 1)

# Смотрим что получилось ДО фильтрации выбросов
print(f'До фильтрации: {[len(c) for c in clusters]}')   # [2, 2, 48, 48]

# Фильтруем: выбросы — кластеры с менее чем 5 точками
# Оставляем только "настоящие" кластеры
clusters = [c for c in clusters if len(c) > 5]

# Проверяем что после фильтрации осталось верное количество кластеров
print(f'После фильтрации: {[len(c) for c in clusters]}') # [48, 48]

medoids = [get_medoid(cluster) for cluster in clusters]

px = sum(x for x, y in medoids) / len(medoids)
py = sum(y for x, y in medoids) / len(medoids)
print(abs(int(px * 10_000)), abs(int(py * 10_000)))      # 5949 27762
````
