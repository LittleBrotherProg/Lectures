---

### Что такое кластеризация простыми словами?

Представь карту звёздного неба с тысячами точек. Невооружённым глазом видно что точки собираются в отдельные **группы** — созвездия, скопления. Задача — **автоматически** определить к какой группе относится каждая точка, не просматривая их вручную.

Именно это и называется **кластеризацией**: разбивка данных на группы (кластеры) так, чтобы точки внутри одного кластера были **близки** друг к другу, а точки из разных кластеров — **далеки**.

## Зачем нам нужны евклидово расстояние и медоида?

Прежде чем лезть в код — разберём зачем вообще нужны эти два понятия и почему без них задачу не решить.

---

### Что требует задача?

Читаем условие: *«определите координаты центра каждого кластера»*.

Центр кластера — это не просто «середина». В условии написано точнее:

> **«центром кластера» называется точка, сумма расстояний от которой до всех остальных точек кластера минимальна**

Разберём эту фразу по частям:

| Часть фразы | Что это значит |
|:---|:---|
| «точка кластера» | Центр — это одна из **реальных** точек данных, не придуманная |
| «сумма расстояний» | Нужно уметь считать **расстояния** между точками |
| «до всех остальных» | Перебираем каждую пару точек |
| «минимальна» | Ищем точку, которая **ближе всего** ко всем остальным |

Значит нам нужны **два инструмента**:
1. Способ измерить расстояние между двумя точками → **евклидово расстояние**
2. Способ найти точку с минимальной суммой расстояний → **медоида**

---

### Зачем нам евклидово расстояние?

Потому что нам нужно **измерять расстояния** между точками — и это единственный способ понять, какие точки «близкие» (одного кластера), а какие «далёкие» (разных кластеров).

Евклидово расстояние — это просто **длина прямой линии** между двумя точками на плоскости. То самое расстояние которое ты измерил бы линейкой на листе бумаги:

```
d(A, B) = √((x₂−x₁)² + (y₂−y₁)²)
```

Оно используется в задаче **три раза**:

| Где используется | Зачем |
|:---|:---|
| При поиске медоиды | Считаем расстояние от каждой точки до всех остальных |
| В алгоритме k-средних | Определяем к какому центру ближе каждая точка |
| В алгоритме DBSCAN | Определяем являются ли две точки «соседями» (d < eps) |

<div style="background:#1c1700;border-left:3px solid #f59e0b;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#f59e0b;">Без расстояния ничего не работает:</strong> мы не можем определить «близкие» и «далёкие» точки, не можем найти центр кластера, не можем решить задачу.</div>

В Python это функция `dist()` из модуля `math`:

```python
from math import dist

dist((1, 0), (5, 3))   # → 5.0
```

---

### Зачем нам медоида?

Теперь когда мы умеем считать расстояния — ищем центр кластера.

Казалось бы, центр — это среднее. Взять среднее x и среднее y всех точек. Но такой «центр» — это **абстрактная точка**, которой может не быть в данных.

Пример: у тебя есть 4 звезды на карте: (1,2), (3,4), (5,6), (7,8).

```
Среднее: ((1+3+5+7)/4, (2+4+6+8)/4) = (4, 5)
```

Точки (4, 5) в твоих данных **нет**. Это придуманная точка в пустом месте.

**Медоида** — решение этой проблемы. Мы перебираем **каждую реальную точку** и считаем сумму расстояний от неё до всех остальных. Та точка у которой эта сумма **наименьшая** — и есть медоида. Она всегда реальна:

```
(1,2): до (3,4) + до (5,6) + до (7,8) = 2.83 + 5.66 + 8.49 = 16.98
(3,4): до (1,2) + до (5,6) + до (7,8) = 2.83 + 2.83 + 5.66 = 11.32  ← минимум
(5,6): до (1,2) + до (3,4) + до (7,8) = 5.66 + 2.83 + 2.83 = 11.32
(7,8): до (1,2) + до (3,4) + до (5,6) = 8.49 + 5.66 + 2.83 = 16.98

Медоида = (3, 4)  — реальная точка из данных ✓
```

<div style="background:#0d1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#10b981;">Итого:</strong> евклидово расстояние — это инструмент измерения. Медоида — это то что мы ищем с его помощью. Без расстояния не найти медоиду, без медоиды нет ответа на задачу.</div>

---

### Связь между ними — одна формула

```python
def get_medoid(cluster):
    #                       ↓ это медоида — точка с min суммой
    return min(cluster, key=lambda p:
    #                             ↓ это евклидово расстояние
               sum(dist(p, q) for q in cluster))
    #               ↑ считаем для каждой пары (p, q)
```

Читается так: *«найди в кластере такую точку p, для которой сумма евклидовых расстояний до всех остальных точек q — минимальна»*. Это и есть вся логика задания 27 в одной строке.

---

### Итоговая цепочка решения

```
Файл с координатами точек
        ↓
Делим на кластеры (прямые / k-средних / DBSCAN)
        ↓
Для каждого кластера ищем медоиду
    → перебираем каждую точку p
    → считаем сумму dist(p, q) для всех q  ← евклидово расстояние
    → берём точку с минимальной суммой     ← медоида
        ↓
Среднее абсцисс и ординат медоид
        ↓
Ответ: abs(int(px × 10000))  abs(int(py × 10000))
```

## Что делать с выпадающими точками (выбросами)?

Когда ты строишь диаграмму, иногда видишь несколько точек, которые **далеко** от всех остальных — как будто они случайно попали на карту. Это и есть **выбросы** (аномалии, выпадающие точки).

---

### Почему выбросы — проблема?

Если просто включить выброс в ближайший кластер — он **испортит медоиду**. Медоида «притянется» к выбросу и перестанет отражать реальный центр кластера.

Пример: кластер из 50 звёзд расположен около координат (5, 5). Но один выброс стоит в точке (100, 100). Медоида сдвинется туда где меньше суммарное расстояние — ближе к центру, но уже не туда куда нам нужно.

---

### Откуда берутся выбросы в задачах ЕГЭ?

В задании 2710 и похожих задачах выбросы **намеренно добавлены** в данные. Условие прямо говорит:

> *«Выбросами будем называть те звёзды, которые имеют менее 5 соседних звёзд в радиусе единичной окружности. В дальнейших расчётах выбросы не учитываются.»*

То есть выброс — это точка у которой **слишком мало соседей** рядом.

---

### Как DBSCAN обнаруживает выбросы автоматически?

Когда DBSCAN доходит до одиночной точки далеко от всех остальных — вокруг неё в радиусе eps **нет соседей**. Очередь сразу пуста, кластер закрывается. Такая точка образует **кластер из 1–4 точек**.

После работы DBSCAN смотрим на размеры кластеров:

```python
clusters = dbscan(data, 1)
print([len(c) for c in clusters])
# → [2, 2, 48, 48]
#    ↑  ↑  ─────────
#  выбросы   нормальные кластеры
```

Два кластера по 2 точки — это выбросы. Два кластера по 48 точек — настоящие кластеры.

**Как работает:**
1. Строим точечную диаграмму в LibreOffice Calc (Вставка → Диаграмма → Тип XY)
2. Находим прямую-разделитель (x = c, y = c, или y = ax + b)
3. В коде: если x < c → кластер 1, иначе → кластер 2

## Часть 2 — Алгоритм k-средних

Этот метод не требует вручную рисовать разделяющие прямые — он **сам** находит к каким кластерам принадлежат точки. Но ему нужна подсказка: по одной **начальной точке** в каждом предполагаемом кластере.

---

### Как работает алгоритм — пошагово

1. Выбираем начальные точки (одну в каждом кластере) — **вручную по диаграмме**
2. Каждую точку данных относим к **ближайшей** начальной точке → получаем черновые кластеры
3. Для каждого черновика ищем **медоиду** → это новые «центры»
4. Снова относим каждую точку к **ближайшему** новому центру → кластеры обновились
5. Повторяем шаги 3–4, пока **центры не перестанут двигаться** (алгоритм сошёлся)

<div style="background:#0d1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#10b981;">Важно:</strong> алгоритм сходится сам — медоиды постепенно «переползают» к истинным центрам кластеров. Даже если начальные точки выбраны немного неточно — всё будет хорошо.</div>

---

### Откуда берутся начальные точки?

<div style="background:#1c1700;border-left:3px solid #f59e0b;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#f59e0b;">Частый вопрос:</strong> начальные точки НЕ указаны в условии задачи. Их нужно выбирать самостоятельно, глядя на диаграмму.</div>

**Алгоритм выбора:**
1. Строим точечную диаграмму в LibreOffice Calc
2. Визуально находим все кластеры
3. В каждом кластере выбираем **любую точку** — на глаз, примерно. Точность не важна.
4. Главное правило: **каждая начальная точка должна быть ближе к своему кластеру, чем к любому другому**

Если нарушить это правило — алгоритм объединит два кластера в один или разобьёт один кластер на два.

**Пример для трёх кластеров:**
```
Левый нижний  → берём (−4, 1)   — можно было (−5, 0) или (−3, 2)
Правый нижний → берём (4, 4)    — можно было (3, 3) или (5, 5)
Верхний       → берём (5, 10)   — можно было (4, 9) или (6, 11)
```

---

### Три функции алгоритма

```python
# Функция 1: назначить каждую точку к ближайшему центру
def assign_clusters(data, centers):
    labels = []
    for p in data:
        # Считаем расстояние от точки p до каждого центра
        distances = [dist(p, center) for center in centers]
        # Берём индекс ближайшего центра — это номер кластера
        labels.append(distances.index(min(distances)))
    return labels

# Функция 2: пересчитать медоиды для каждого кластера
def update_centers(data, labels, k):
    new_centers = []
    for cluster_label in range(k):
        # Собираем все точки текущего кластера
        cluster = [p for p, l in zip(data, labels) if l == cluster_label]
        # Ищем медоиду этого кластера
        medoid = min(cluster, key=lambda p: sum(dist(p, q) for q in cluster))
        new_centers.append(medoid)
    return new_centers

# Функция 3: запускать итерации до стабилизации
def k_means(data, k):
    centers = [[-9, -2], [-1, 15]]   # начальные точки — по диаграмме!
    new_centers = []
    # Пока медоиды ещё двигаются — продолжаем
    while new_centers != centers:
        labels = assign_clusters(data, centers)
        new_centers = update_centers(data, labels, k)
        centers = new_centers
    return labels, centers
```

## Часть 3 — Алгоритм DBSCAN

Самый **универсальный** алгоритм. Работает с кластерами **любой формы** — прямоугольными, серповидными, кольцами. Автоматически определяет **выбросы** (одиночные точки далеко от всех кластеров).

---

### Главная идея: точки-соседи

DBSCAN строит кластеры по принципу **плотности**: если точки находятся достаточно близко — они в одном кластере. «Достаточно близко» — это расстояние меньше **eps** (эпсилон).

Представь что каждая точка «зажигает» фонарик с радиусом eps. Все точки в радиусе — её соседи. Соседи соседей — тоже в одном кластере. Так кластер постепенно «растёт» пока есть близкие точки.

---

### Как выбрать eps?

Eps подбирается **по диаграмме** визуально. Правило:
- eps > максимального расстояния между соседними точками **внутри** одного кластера
- eps < минимального расстояния между точками **разных** кластеров

Если eps слишком мал → граничные точки кластера становятся выбросами (образуют одиночные кластеры).
Если eps слишком велик → разные кластеры сливаются в один.

<div style="background:#0d1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#10b981;">Совет:</strong> после запуска всегда выводи <code>print([len(c) for c in clusters])</code> — должно получиться ровно столько кластеров нужного размера, сколько указано в условии.</div>

### Функция dbscan — разбор каждой строки

```python
def dbscan(data, eps):
    clusters = []           # список всех найденных кластеров

    while data:             # пока есть необработанные точки
        # pop() берёт ПОСЛЕДНЮЮ точку из data И удаляет её
        # Эта точка станет «зерном» нового кластера
        point = data.pop()

        cluster = [point]   # новый кластер начинается с одной точки
        queue   = [point]   # очередь: точки, соседей которых нужно проверить

        while queue:        # пока есть что проверять
            # pop(0) берёт ПЕРВУЮ точку из очереди (как живая очередь)
            current = queue.pop(0)

            # Ищем всех соседей current среди необработанных точек
            # Сосед = точка, до которой расстояние < eps
            neighbors = [p for p in data if dist(current, p) < eps]

            cluster.extend(neighbors)   # добавляем соседей в кластер
            queue.extend(neighbors)     # и в очередь — у них тоже могут быть соседи!

            # Удаляем обработанных соседей из списка необработанных точек
            # data[:] = ... обновляет список на месте, не создаёт новый
            data[:] = [p for p in data if p not in neighbors]

        # Кластер больше нельзя расширить (очередь пуста) — сохраняем
        clusters.append(cluster)

    return clusters
```
