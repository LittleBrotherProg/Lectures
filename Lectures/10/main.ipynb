## Задача 1

Текст произведения Льва Николаевича Толстого «Севастопольские рассказы» представлен в виде файлов различных форматов. Откройте один из файлов и определите, сколько раз встречается в тексте отдельное слово «солдаты» со строчной буквы. Другие формы этого слова учитывать не следует. В ответе запишите только число.

**Файлы к заданию:** https://kompege.ru/files/YVVDi1KNx.docx

### Алгоритм решения:

1) **Скачиваем файл и открываем его**

2) **Открываем функцию "Найти и заменить"**

![Общий вид файла](img/g1.gif)

3) **Вписываем в строку поиска искомое слово "солдаты"**

![Общий вид файла](img/g2.gif)

4) **Настраиваем поиск с учетом регистра**
   - **Регистр** означает:
     - **Верхний регистр**: слова начинаются с заглавной/прописной буквы (Пример: Арбуз, Виноград)
     - **Нижний регистр**: слова начинаются с маленькой/строчной буквы (Пример: арбуз, виноград)
   - Нам нужны слова только с **нижним регистром**
   - Устанавливаем галочку напротив **"Учитывать регистр"**

![Общий вид файла](img/g3.gif)

5) **Нажимаем "Найти все"** для поиска всех слов "солдаты" со строчной буквы

![Общий вид файла](img/g4.gif)

6) **Найдено 26 значений**

![Общий вид файла](img/img1.png)

7) **Рекомендуется проверить все найденные слова**

![Общий вид файла](img/g5.gif)

8) **Подтверждаем правильность ответа** - 26
9) **Итоговый ответ: 26**
