## Машина Тьюринга: устройство

Машина Тьюринга состоит из четырёх элементов:

1. **Бесконечная лента** — разбита на ячейки. Каждая содержит символ: `0`, `1`, или `λ` (пусто)
2. **Читающая головка** — стоит над одной ячейкой, читает/пишет символ, двигается влево (L) или вправо (R)
3. **Внутренние состояния** — `q0`, `q1`, `q2`… и состояние остановки
4. **Таблица переходов** — набор правил: состояние + символ → записать + сдвинуться + новое состояние

### Разбор задания 1218

**Шаг 1: понять алгоритм по таблице**
- `q0 + λ` → сдвиг влево, переход в `q1`
- `q1 + 1` → записать **0**, сдвиг влево, остаться в `q1`
- `q1 + 0` → записать **1**, сдвиг влево, остаться в `q1`
- `q1 + λ` → **стоп**

Вывод: алгоритм **инвертирует** все цифры (0 ↔ 1), двигаясь справа налево.

**Шаг 2: перевести 88 в двоичную**

88 = 64 + 16 + 8 = 2⁶ + 2⁴ + 2³ → **1011000**

**Шаг 3: применить обратную операцию (инвертировать)**

`1011000` → **0100111**

**Шаг 4: устранить ведущий ноль — добавляем 1 слева**

`0100111` имеет ведущий ноль, что запрещено. Ставим **1** перед ним: **10100111**.

При инвертировании: `10100111` → **01011000** = 1011000₂ = 88 ✓

**Шаг 5: перевести 10100111₂ в десятичную**

10100111₂ = 128 + 32 + 4 + 2 + 1 = **167**

После инвертирования эта 1 превращается в 0 (ведущий ноль в результате — допустимо, значение 88 не меняется).</div>

### Разбор задания 1216

**Шаг 1: понять алгоритм**
- `q1 + 1` → записать **0**, **СТОП** — программа останавливается!
- `q1 + 0` → записать **1**, сдвиг влево, продолжать
- `q1 + λ` → **стоп**

Алгоритм движется справа налево, заменяет `0 → 1` и продолжает, пока не встретит **первую единицу** справа. Эту единицу заменяет на **0** и останавливается.

**Шаг 2: что происходит со счётчиком нулей**

Исходная последовательность (схематично):

<div style="background:#0a1628;border:1px solid #1e2d47;border-radius:8px;padding:12px 16px;margin:10px 0;font-family:monospace;font-size:13px;color:#e2e8f0;display:inline-block;">
← нули (не тронуты) &nbsp;|&nbsp; <span style="color:#f59e0b;">1</span> &nbsp;|&nbsp; нули (станут 1) →<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;k штук &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; стоп &nbsp;&nbsp;&nbsp; 1000−k−1 штук
</div>

В результате:
- `k` нулей слева — **не тронуты** → остаются нулями
- 1 единица → заменяется на **0** (добавляет 1 ноль)
- `1000−k−1` нулей справа → заменяются на 1 → **исчезают**

Итого нулей в результате: `k + 1 = 343` → **k = 342**

**Шаг 3: подсчитать нули в исходной**

- k = 342 нуля слева от единицы
- 1 единица
- (1000 − 342 − 1) = **657 нулей** справа от единицы

Итого нулей в исходной: **342 + 657 = 999**
