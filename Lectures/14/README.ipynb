## Теория

Задание 14 — нужно перевести большое число (выражение типа a^n + b^m − x) в другую систему счисления и посчитать нули, значащие цифры и т.д.

```python
# Перевод любого числа в n-ричную СС:
def to_base(num, base):
    from string import digits, ascii_uppercase
    alph = digits + ascii_uppercase
    result = ''
    while num > 0:
        result += alph[num % base]
        num //= base
    return result[::-1] or '0'

# Количество нулей в n-ричной записи:
def count_zeros(num, base):
    count = 0
    while num > 0:
        if num % base == 0:
            count += 1
        num //= base
    return count

# Перебор x для нахождения нужного числа:
for x in range(1, 2031)[::-1]:     # от большего (ищем наибольший)
    num = 7**170 + 7**100 - x
    if count_zeros(num, 7) == 71:
        print(x)
        break
```

<div class="task-box">
<div class="task-title">📋 Задание Шаблон 1 — Нули в 7-ричной системе | Ответ: <strong style="color:#2fca72;">Перебором</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Вычислите значение выражения 7¹⁷⁰ + 7¹⁰⁰ − x, где x — натуральное число, x ≤ 2030.<br>
<strong>Найдите наибольшее x, при котором результат в семеричной системе счисления содержит ровно 71 незначащий ноль.</strong></div>
</div>

````python
def count_zeros(num, base):
    count = 0
    while num > 0:
        if num % base == 0:
            count += 1
        num //= base
    return count

# Ищем наибольший x ≤ 2030:
for x in range(1, 2031)[::-1]:
    num = 7**170 + 7**100 - x
    if count_zeros(num, 7) == 71:
        print(f"x = {x}")
        break
````

<div class="task-title">📋 Задание Шаблон 2 — Нули в 25-ричной системе | Ответ: <strong style="color:#2fca72;">Подсчётом</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Вычислите значение выражения 3·3125⁸ + 2·625⁷ − 5·625⁶ + 3·125⁵ − 2·25⁴ − 2025.<br>
<strong>Найдите количество значащих нулей в 25-ричной записи этого числа.</strong><br>
(Незначащие нули — ведущие нули, значащие — внутри или в конце числа.)</div>
</div>

````python
num = (3 * 3125**8 + 2 * 625**7 - 5 * 625**6
       + 3 * 125**5 - 2 * 25**4 - 2025)

count = 0
while num > 0:
    if num % 25 == 0:
        count += 1
    num //= 25

print(f"Нулей: {count}")
````

<div style="color:#c8d8e8;line-height:1.7;">В алфавите некоторой системы счисления с основанием N для записи числа 1234X78 (X — неизвестная цифра в позиции десятков) в этой системе счисления первое число кратно второму числу.<br>
<strong>Найдите все допустимые значения цифры X из алфавита системы счисления.</strong></div>
</div>

````python
from string import digits, ascii_uppercase
alph = digits + ascii_uppercase

# Пример: 19-ричная СС, X — какая-то цифра, проверяем кратность
for base in range(10, 37):        # проверяем возможные основания
    for x in alph[:base]:
        try:
            n1 = int(f'123{x}24', base)
            n2 = int(f'135{x}78', base)
            if n2 != 0 and n1 % n2 == 0:
                print(f"base={base}, x={x}, n1={n1}, n2={n2}")
        except:
            pass
````
