## 2. Тип 1 — Выражение с неизвестным числом x

<div class="info-box">
<strong style="color:#1b91ff;">Признак:</strong> в выражении есть переменная x — целое положительное число из заданного диапазона.<br>
Найти: наибольший (или наименьший) x, при котором запись результата содержит ровно N заданных цифр.
</div>

### Шаблон решения

```python
# Перебираем x от наибольшего к наименьшему
for x in range(1, ПРЕДЕЛ + 1)[::-1]:
    # Вычисляем выражение
    num = ВЫРАЖЕНИЕ_С_x
    # Считаем нужные цифры в основании SYS
    count = 0
    while num:
        if num % SYS == ЦИФРА:   # 0 для нулей
            count += 1
        num //= SYS
    # Нашли нужное количество
    if count == N:
        print(x)
        break
```

````python
# Пример: 7^170 + 7^100 – x, x ≤ 2030
# Найти наибольший x с ровно 71 нулём в 7-ричной записи

for x in range(1, 2031)[::-1]:
    num = 7**170 + 7**100 - x
    count = 0
    while num:
        if num % 7 == 0:   # цифра 0 в 7-ричной СС
            count += 1
        num //= 7
    if count == 71:
        print(x)   # Ответ: 2029
        break
````

### Вариант: найти наибольшее количество нулей

````python
# Найти максимальное количество нулей по всем x ≤ 2030

def count_zero(num):
    count = 0
    while num:
        if num % 6 == 0:
            count += 1
        num //= 6
    return count

max_zeros = max(count_zero(6**2030 + 6**100 - x)
                for x in range(1, 2031))
print(max_zeros)   # Ответ: 1934
````

## 3. Тип 2 — Выражение без неизвестных

<div class="info-box">
<strong style="color:#1b91ff;">Признак:</strong> конкретное арифметическое выражение (без x).<br>
Найти: количество заданных цифр в записи результата в нужной СС.<br>
Проще Типа 1 — просто переписываем выражение и считаем цифры.
</div>

| Условие | Код |
|:---|:---|
| Значащие нули | `if num % SYS == 0` |
| Цифры > 10 | `if num % SYS > 10` |
| Конкретная цифра K | `if num % SYS == K` |
| Не нули | `if num % SYS != 0` |

````python
# Пример: 3·3125^8 + 2·625^7 – 5·625^6 + 3·125^5 – 2·25^4 – 2025
# Основание 25. Найти количество значащих нулей.

num = (3 * 3125**8 + 2 * 625**7 - 5 * 625**6
       + 3 * 125**5 - 2 * 25**4 - 2025)
count = 0

while num:
    if num % 25 == 0:   # цифра 0 в 25-ричной СС
        count += 1
    num //= 25

print(count)   # Ответ: 10
````

````python
# Пример: найти количество цифр > 10 в 25-ричной записи

num = (4 * 3125**2019 + 3 * 625**2020 - 2 * 125**2021
       + 25**2022 - 4 * 5**2023 - 2024)
count = 0

while num:
    if num % 25 > 10:   # цифры 11-24 в 25-ричной СС
        count += 1
    num //= 25

print(count)   # Ответ: 3030
````

## 4. Тип 3 — Неизвестная цифра x в записи числа

<div class="info-box">
<strong style="color:#1b91ff;">Признак:</strong> в числах выражения стоит неизвестная цифра x из алфавита N-ричной СС.<br>
Найти: наибольший x, при котором результат кратен (N-1). Вывести частное от деления.
</div>

### Алфавит N-ричной СС

```python
from string import digits, ascii_uppercase

# '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
alph = digits + ascii_uppercase

# Алфавит 19-ричной СС: первые 19 символов
alph19 = alph[:19]   # '0123456789ABCDEFGHI'
```

<div class="warn-box">
<strong style="color:#f0c040;">Зачем f-строки?</strong><br>
<code>int(f'98897{x}21', 19)</code> — вставляем цифру x в нужное место числа,<br>
затем переводим всё из 19-ричной в десятичную через <code>int(строка, основание)</code>.
</div>

````python
from string import digits, ascii_uppercase

alph = digits + ascii_uppercase

# Перебираем x от наибольшей цифры 19-ричной СС
for x in alph[:19][::-1]:
    # Переводим числа из 19-ричной в десятичную
    n1 = int(f'98897{x}21', 19)
    n2 = int(f'2{x}923', 19)
    res = n1 + n2
    # Проверяем кратность 18
    if res % 18 == 0:
        print(res // 18)   # Ответ: 469034148
        break
````

````python
from string import digits, ascii_uppercase

alph = digits + ascii_uppercase

# Пример 2: основание 27, кратность 26
for x in alph[:27][::-1]:
    n1 = int(f'123{x}24', 27)
    n2 = int(f'135{x}78', 27)
    res = n1 + n2
    if res % 26 == 0:
        print(res // 26)   # Ответ: 1213206
        break
````
