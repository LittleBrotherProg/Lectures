## Ключевой алгоритм: подсчёт цифр в числе

Главный инструмент для всего задания 14 — цикл с последовательным делением:

```python
count = 0
while num:
    if num % sys == 0:   # текущий разряд — ноль?
        count += 1
    num //= sys          # переходим к следующему разряду
```

**Почему это работает?** `num % sys` даёт последнюю цифру числа в системе счисления с основанием `sys`. Затем `num //= sys` сдвигает число вправо на один разряд. Повторяя — проходим по всем цифрам.

## Тип 1 — Выражение с неизвестным числом

В заданиях данного типа в выражении один член является неизвестным (`x`) — **целое положительное** число, не превышающее заданного значения.

**Общий алгоритм:**
1. Определить диапазон `x` из условия
2. Ищем **наибольший** `x` → перебирать **сверху вниз**: `range(...)[::-1]`
3. Ищем **наименьший** `x` → перебирать **снизу вверх**: `range(...)`
4. Для каждого `x` — вычислить выражение и подсчитать цифры
5. Условие выполнено → вывести ответ и выйти через `break`

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1400 — Тип 1 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Значение арифметического выражения <strong>7¹⁷⁰ + 7¹⁰⁰ – x</strong>, где x – целое положительное число, не превышающее 2030, записали в 7-ричной системе счисления. Определите наибольшее значение x, при котором в 7-ричной записи числа, являющегося значением данного арифметического выражения, содержится <strong>ровно 71 нуль</strong>.<br>В ответе запишите число в десятичной системе счисления.<br><br><strong>Разбор:</strong> ищем <em>наибольший</em> x → перебираем от 2030 до 1. Как только количество нулей равно 71 — выводим x и выходим.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 2029</div></div>

````python
# Задание 1400
for x in range(1, 2031)[::-1]:
    num = 7 ** 170 + 7 ** 100 - x
    count = 0
    while num:
        if num % 7 == 0:
            count += 1
        num //= 7
    if count == 71:
        print(x)   # Ответ: 2029
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1401 — Тип 1 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Значение арифметического выражения <strong>6²⁰³⁰ + 6¹⁰⁰ – x</strong>, где x – целое положительное число, не превышающее 2030, записали в 6-ричной системе счисления. Определите <strong>наибольшее количество нулей</strong>, которое может содержать число, являющееся 6-ричной записью значения данного арифметического выражения.<br><br><strong>Разбор:</strong> нужно найти не x, а <em>максимальное количество нулей</em> по всем x. Вынесем подсчёт в функцию и используем <code>max()</code>.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 1934</div></div>

````python
# Задание 1401
def count_zero(num):
    count = 0
    while num:
        if num % 6 == 0:
            count += 1
        num //= 6
    return count

max_zeros = max(count_zero(6 ** 2030 + 6 ** 100 - x)
                for x in range(1, 2031))
print(max_zeros)   # Ответ: 1934
````

## Значащие и незначащие нули

**Незначащие нули** — не влияют на величину числа, их можно отбросить. Пример: `1 = 01 = 001` — нули слева незначащие.

**Значащие нули** — несут информацию о величине. Пример: `10`, `100`, `1000` — нули справа значащие.

## Тип 2 — Выражение без неизвестных

В заданиях этого типа нужно определить **количество** заданных **цифр** в числе — результате арифметического выражения. Данный тип несколько проще первого.

<div style="background:#1c1700;border-left:3px solid #f59e0b;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#f59e0b;">Обратите внимание:</strong> если в задаче написано «значащие нули» — это именно то, что считает наш цикл. Ведущие нули он не учитывает.</div>

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1402 — Тип 2 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Значение арифметического выражения записали в системе счисления с основанием 25:<br><br><strong>3 · 3125⁸ + 2 · 625⁷ – 4 · 625⁶ + 3 · 125⁵ – 2 · 25⁴ – 2025</strong><br><br>Сколько <strong>значащих нулей</strong> содержится в этой записи?<br><br><strong>Разбор:</strong> переписываем выражение в Python и считаем нули с основанием 25. Заметим: <code>3125=5⁵</code>, <code>625=5⁴</code>, <code>125=5³</code>, <code>25=5²</code>.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 10</div></div>

````python
# Задание 1402
num = (3 * 3125 ** 8 + 2 * 625 ** 7 - 5 * 625 ** 6
       + 3 * 125 ** 5 - 2 * 25 ** 4 - 2025)

count = 0
while num:
    if num % 25 == 0:
        count += 1
    num //= 25

print(count)   # Ответ: 10
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1403 — Тип 2 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Определите в 25-ричной записи числа <strong>количество цифр с числовым значением, превышающим 10</strong>:<br><br><strong>4 · 3125²⁰¹⁹ + 3 · 625²⁰²⁰ – 2 · 125²⁰²¹ + 25²⁰²² – 4 · 5²⁰²³ – 2024</strong><br><br><strong>Разбор:</strong> принцип тот же, меняется только условие: вместо <code>== 0</code> пишем <code>&gt; 10</code>.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 3030</div></div>

````python
# Задание 1403
num = (4 * 3125 ** 2019 + 3 * 625 ** 2020 - 2 * 125 ** 2021
       + 25 ** 2022 - 4 * 5 ** 2023 - 2024)

count = 0
while num:
    if num % 25 > 10:
        count += 1
    num //= 25

print(count)   # Ответ: 3030
````

## Тип 3 — Выражение с одной неизвестной цифрой

В каждом числе выражения **одна цифра заменена на `x`**. Нужно перебрать все допустимые значения из алфавита системы счисления.

### Формирование алфавита

```python
from string import digits, ascii_uppercase

alph = digits + ascii_uppercase
# '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

alph[:19]          # алфавит 19-ричной СС
alph[:19][::-1]    # от наибольшего к наименьшему
```

### Запись числа с переменной цифрой

```python
# Число 98897x21 в 19-ричной системе:
n1 = int(f'98897{x}21', 19)

# Число 2x923 в 19-ричной системе:
n2 = int(f'2{x}923', 19)
```

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1404 — Тип 3 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Операнды арифметического выражения записаны в системе счисления с основанием <strong>19</strong>:<br><br><strong>98897x21₁₉ + 2x923₁₉</strong><br><br>Переменной <code>x</code> обозначена неизвестная цифра. Определите <strong>наибольшее</strong> значение x, при котором значение выражения <strong>кратно 18</strong>.<br>Для найденного x вычислите <strong>частное от деления</strong> на 18.<br><br><strong>Разбор:</strong> алфавит 19-ричной СС — 19 символов. Перебираем от последнего (наибольшего). Проверяем кратность. Выводим частное.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 469034148</div></div>

````python
# Задание 1404
from string import digits, ascii_uppercase
alph = digits + ascii_uppercase

for x in alph[:19][::-1]:
    n1 = int(f'98897{x}21', 19)
    n2 = int(f'2{x}923', 19)
    res = n1 + n2
    if res % 18 == 0:
        print(res // 18)   # Ответ: 469034148
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1405 — Тип 3 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Операнды арифметического выражения записаны в системе счисления с основанием <strong>27</strong>:<br><br><strong>123x24₂₇ + 135x78₂₇</strong><br><br>Определите <strong>наибольшее</strong> значение x, при котором значение выражения <strong>кратно 26</strong>.<br>Для найденного значения x вычислите <strong>частное от деления</strong> на 26.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 1213206</div></div>

````python
# Задание 1405
from string import digits, ascii_uppercase
alph = digits + ascii_uppercase

for x in alph[:27][::-1]:
    n1 = int(f'123{x}24', 27)
    n2 = int(f'135{x}78', 27)
    res = n1 + n2
    if res % 26 == 0:
        print(res // 26)   # Ответ: 1213206
        break
````
