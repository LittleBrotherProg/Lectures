## 2. Базовая рекурсивная функция

### Задача-введение: из 1 получить 8 операциями +3 и ×2

Рассматриваем дерево **сверху вниз**: из 8 к 1, применяя обратные операции.

**Шаг 1** — разбиваем 8:



**Шаг 2** — ветка числа 5:



**Шаг 3** — первый путь найден:



**Шаг 4** — ветка числа 4:



**Шаг 5** — третий путь:



**Полное дерево рекурсии f(1,8):**



**Итого: 3 пути → f(1, 8) = 3**

````python
def f(x, y):
    # 1. Достигли цели — путь засчитан
    if x == y:
        return 1
    # 2. Вышли за предел — путь не подходит
    if x > y:          # при уменьшении числа: if x < y
        return 0
    # 3. Применяем все операции, суммируем количество путей
    if x < y:
        return f(x + 3, y) + f(x * 2, y)

print(f(1, 8))  # Ответ: 3
````

<div class="warn-box">
<strong style="color:#f0c040;">Почему складываем?</strong><br>
Нас интересует КОЛИЧЕСТВО путей. Каждый успешный путь возвращает 1.<br>
Сумма всех 1-ц = количество всех подходящих программ.
</div>

## 3. Тип 1 — Траектория содержит число A

<div class="info-box">
<strong style="color:#1b91ff;">Идея:</strong> разбиваем задачу на два участка:<br>
<code>f(x, A) * f(A, y)</code><br><br>
Умножаем, потому что каждый путь через A = один вариант до A × один вариант от A до y.<br>
Через две точки: <code>f(x,A) * f(A,B) * f(B,y)</code> — порядок важен!
</div>

````python
# Задание 2300: −2 и //2, из 32 в 1, через 14
def f(x, y):
    if x == y:
        return 1
    if x < y:      # число уменьшается
        return 0
    if x > y:
        return f(x - 2, y) + f(x // 2, y)

print(f(32, 14) * f(14, 1))   # Ответ: 54

# Задание 2301: из 38 в 2, через 16
print(f(38, 16) * f(16, 2))   # Ответ: 36
````

## 4. Тип 2 — Траектория НЕ содержит число B

<div class="tip-box">
<strong style="color:#2fca72;">Идея:</strong> добавляем B в условие выхода:<br>
<code>if x > y or x == B: return 0</code><br><br>
Несколько чисел: <code>if x > y or x == B1 or x == B2: return 0</code>
</div>

````python
# Задание 2302: +1, ×2, x², из 2 в 20, не через 11
def f(x, y):
    if x == y:
        return 1
    if x > y or x == 11:   # ← добавили запрет
        return 0
    if x < y:
        return f(x + 1, y) + f(x * 2, y) + f(x ** 2, y)

print(f(2, 20))   # Ответ: 37

# Задание 2303: −1, −2, //3, из 19 в 3, не через 9 и 16
def f(x, y):
    if x == y:
        return 1
    if x < y or x == 9 or x == 16:
        return 0
    if x > y:
        return f(x - 1, y) + f(x - 2, y) + f(x // 3, y)

print(f(19, 3))   # Ответ: 180
````

## 5. Тип 3 — Содержит A и не содержит B

<div class="info-box">
<strong style="color:#1b91ff;">Комбинируем оба подхода:</strong><br>
1. Число B → в условие выхода: <code>if x < y or x == B: return 0</code><br>
2. Число A → разбиваем через умножение: <code>f(x,A) * f(A,y)</code>
</div>

````python
# Задание 2304: −1, −6, //2, из 34 в 6, через 29 и 19, не через 24
def f(x, y):
    if x == y:
        return 1
    if x < y or x == 24:   # исключаем 24
        return 0
    if x > y:
        return f(x - 1, y) + f(x - 6, y) + f(x // 2, y)

# Разбиваем: 34→29, затем 29→19, затем 19→6
# Порядок важен: сначала большее число!
print(f(34, 29) * f(29, 19) * f(19, 6))   # Ответ: 115
````

## 6. Особый случай — ограничение на повторение

<div class="warn-box">
<strong style="color:#f0c040;">Условие:</strong> никакая команда не повторяется более двух раз подряд.<br><br>
Добавляем параметр <code>path</code> — строку с историей команд.<br>
Проверяем: <code>if 'AAA' in path or 'BBB' in path: return 0</code>
</div>

````python
# Задание 2305: +1, ×2, из 1 в 24, без трёх одинаковых команд подряд
def f(x, y, path=''):
    if 'AAA' in path or 'BBB' in path:
        return 0
    if x == y:
        return 1
    if x > y:
        return 0
    if x < y:
        return (f(x + 1, y, path + 'A') +
                f(x * 2, y, path + 'B'))

print(f(1, 24))   # Ответ: 6
````
