## О задании

Задания 19, 20, 21 — **повышенный** уровень. В каждом задании описана игра двух игроков (обычно Ваня и Петя). Нужно определить, выигрывает ли первый игрок (Ваня) при правильной игре обеих сторон.

| Задание | Что найти |
|:---:|:---|
| 19 | При каких начальных значениях Ваня выигрывает (`f(s, 2) == True`) |
| 20 | При каких значениях Ваня выигрывает за N ходов, но **не** за N-2 |
| 21 | Аналогично 20, но для другого N |

## Ключевые понятия

- **Чётный ход (m % 2 == 0)** — ходит **Ваня** (первый игрок)
- **Нечётный ход (m % 2 != 0)** — ходит **Петя** (второй игрок)
- **any(ходы)** — хотя бы один ход выигрышный → Ваня выбирает лучший ход
- **all(ходы)** — все ходы выигрышные → Петя не может помешать

## Универсальный шаблон функции

```python
def f(s, m):
    # 1. Условие победы (завершение игры)
    if s >= LIMIT:              # победа достигнута
        return m % 2 == 0      # чётный ход = ход Вани = победа Вани

    # 2. Ходы закончились
    if m == 0:
        return False

    # 3. Вычисляем все возможные ходы
    ходы = [f(s + op, m - 1) for op in ОПЕРАЦИИ]

    # 4. Выбор стратегии:
    if m % 2 == 0:             # ход Вани → выбирает лучший
        return any(ходы)
    else:                      # ход Пети → выбирает худший для Вани
        return all(ходы)
```

## Задание 19 — одна куча, условие «через A, не через B»

Основная формулировка: найти все начальные значения `s`, при которых Ваня выигрывает при `m` ходах.

Вызов функции: `f(s, 2)` для задания 19 (2 хода).

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 2300 — Тип 1 (разбор) — исполнитель уменьшает число</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Исполнитель преобразует число на экране.<br>
У исполнителя есть две команды, которые обозначены латинскими буквами:<br>
<strong>A. Вычесть 2<br>
B. Найти целую часть от деления на 2</strong><br><br>
Программа для исполнителя — это последовательность команд. Сколько существует программ, для которых при исходном числе <strong>32</strong> результатом является число <strong>1</strong> и при этом траектория вычислений содержит число <strong>14</strong>?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>ABB</strong> при исходном числе 13 траектория состоит из чисел 11, 5, 2.<br><br>
<strong>Разбор:</strong> траектория должна содержать 14 → разбиваем: из 32 в 14, затем из 14 в 1. Ответ = f(32,14) × f(14,1).</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 54</div></div>

````python
def f(x, y):
    """Количество путей из x в y командами A(−2) и B(//2)"""
    if x == y: return 1       # дошли до цели
    if x < y:  return 0       # перепрыгнули (x убывает)
    return f(x - 2, y) + f(x // 2, y)

# Траектория должна содержать 14: из 32→14, затем 14→1
print(f(32, 14) * f(14, 1))   # Ответ: 54
````

Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>

## Тип 3 — Через A и B, но не через C

Комбинация: траектория должна содержать промежуточные точки A и B, но не C.

```python
print(f(x, B) * f(B, A) * f(A, y))   # через B и A
# + добавить x == C: return 0 в тело функции
```

У исполнителя есть три команды, которые обозначены латинскими буквами:<br>
<strong>A. Вычесть 1<br>
B. Вычесть 6<br>
C. Найти целую часть от деления на 2</strong><br><br>
Первая команда уменьшает число на экране на 1, вторая команда уменьшает это число на 6, третья команда делит число нацело на 2.<br>
Программа для исполнителя — это последовательность команд.<br>
Сколько существует таких программ, которые исходное число <strong>34</strong> преобразуют в число <strong>6</strong>, и при этом траектория вычислений содержит числа <strong>19 и 29</strong> и <strong>не содержит числа 24</strong>?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>АВС</strong> при исходном числе 14 траектория состоит из чисел 13, 7, 3.<br><br>
<strong>Разбор:</strong> разбиваем: 34→29, 29→19, 19→6. Запрет на 24 добавляем в функцию.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 115</div></div>

````python
def f(x, y):
    if x == y: return 1
    if x < y or x == 24: return 0    # не проходить через 24
    # A: -1, B: -6, C: //2
    return f(x - 1, y) + f(x - 6, y) + f(x // 2, y)

# Траектория: 34→29→19→6
print(f(34, 29) * f(29, 19) * f(19, 6))  # Ответ: 115
````

У исполнителя есть две команды, которые обозначены латинскими буквами:<br>
<strong>A. Прибавить 1<br>
B. Умножить на 2</strong><br><br>
Первая команда увеличивает число на экране на 1, вторая умножает его на 2.<br>
Программа для исполнителя — это последовательность команд.<br>
Сколько существует программ, которые преобразуют исходное число <strong>1</strong> в число <strong>24</strong>, и при этом <strong>никакая команда не повторяется более двух раз подряд</strong>?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>BAB</strong> при исходном числе 1 траектория состоит из чисел 2, 3, 6.<br><br>
<strong>Разбор:</strong> добавляем строку пути `path` и проверяем отсутствие «AAA» и «BBB» в пути.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 6</div></div>

````python
def f(x, y, path=""):
    # Запрещаем три подряд одинаковых команды
    if "AAA" in path or "BBB" in path:
        return 0
    if x == y: return 1
    if x > y:  return 0
    # A: +1, B: ×2
    return f(x + 1, y, path + "A") + f(x * 2, y, path + "B")

print(f(1, 24))               # Ответ: 6
````
