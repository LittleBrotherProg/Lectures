## Универсальный шаблон решения

```python
def f(x, y):
    if x == y:         # дошли до цели
        return 1
    if x > y:          # перепрыгнули (для убывания: x < y)
        return 0
    # применяем все операции исполнителя:
    return f(x + OP1, y) + f(x * OP2, y)

print(f(начало, конец))
```

> **Тип 1:** `print(f(x, A) * f(A, y))` — через промежуточную точку A  
> **Тип 2:** добавить `or x == B: return 0`  
> **Тип 3:** комбинация подходов

---
## Как работает рекурсия

Функция `f(x, y)` считает, сколькими способами можно попасть из `x` в `y`:
- Если `x == y` → нашли 1 путь
- Если `x > y` → зашли слишком далеко, путей 0
- Иначе → пробуем все операции и складываем результаты

**Пример для операций +3 и ×2, из 1 в 8:**
```
f(1,8) = f(4,8) + f(2,8)   # +3=4, ×2=2
f(4,8) = f(7,8) + f(8,8)   # f(8,8)=1 ✓
f(2,8) = f(5,8) + f(4,8)   # и т.д.
```
Итого: f(1,8) = 3

## Тип 1 — Количество программ из x в y

<div class="task-box">
<div class="task-title">📋 Задание 2300 — Тип 1 — базовый | Ответ: <strong style="color:#2fca72;">54</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Исполнитель преобразует число на экране.<br>
У исполнителя есть две команды, которые обозначены латинскими буквами:<br>
<strong>A. Вычесть 2<br>
B. Найти целую часть от деления на 2</strong><br><br>
Программа для исполнителя — это последовательность команд.<br>
Сколько существует программ, для которых при исходном числе 32 результатом является число 1 и при этом траектория вычислений содержит число 14?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>ABB</strong> при исходном числе 13 траектория состоит из чисел 11, 5, 2</div>
</div>

````python
def f(x, y):
    if x == y: return 1
    if x < y:  return 0
    # A: вычесть 2, B: делить на 2
    return f(x - 2, y) + f(x // 2, y)

# Траектория должна содержать число 14:
# из 32 в 14, затем из 14 в 1
print(f(32, 14) * f(14, 1))   # Ответ: 54
````

## Тип 2 — Программы, не проходящие через число B

<div class="task-box">
<div class="task-title">📋 Задание 2302 — Тип 2 — не через B | Ответ: <strong style="color:#2fca72;">37</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Исполнитель преобразует число на экране.<br>
У исполнителя есть три команды, которые обозначены латинскими буквами:<br>
<strong>A. Прибавить 1<br>
B. Умножить на 2<br>
C. Возвести в квадрат</strong><br><br>
Программа для исполнителя — это последовательность команд.<br>
Сколько существует программ, для которых при исходном числе 2 результатом является число 20, при этом траектория вычислений не содержит числа 11?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>CBA</strong> при исходном числе 4 траектория будет состоять из чисел 16, 32, 33</div>
</div>

````python
def f(x, y):
    if x == y: return 1
    if x > y or x == 11:  # не проходим через 11
        return 0
    # A: +1, B: ×2, C: ²
    return f(x + 1, y) + f(x * 2, y) + f(x ** 2, y)

print(f(2, 20))               # Ответ: 37
````

<div class="task-box">
<div class="task-title">📋 Задание 2303 — Тип 2 — не через 9 и 16 | Ответ: <strong style="color:#2fca72;">180</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Исполнитель преобразует число на экране.<br>
У исполнителя есть три команды, которые обозначены латинскими буквами:<br>
<strong>А. Вычесть 1<br>
В. Вычесть 2<br>
С. Найти целую часть от деления на 3</strong><br><br>
Программа для исполнителя — это последовательность команд.<br>
Сколько существует программ, для которых при исходном числе 19 результатом является число 3, при этом траектория вычислений не содержит чисел 9 и 16?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>СВА</strong> при исходном числе 13 траектория будет состоять из чисел 4, 2, 1</div>
</div>

````python
def f(x, y):
    if x == y: return 1
    if x < y or x == 9 or x == 16:  # не через 9 и 16
        return 0
    # А: -1, В: -2, С: //3
    return f(x - 1, y) + f(x - 2, y) + f(x // 3, y)

print(f(19, 3))               # Ответ: 180
````

## Тип 3 — Через A, но не через B

<div class="task-box">
<div class="task-title">📋 Задание 2304 — Тип 3 — через 19 и 29, но не через 24 | Ответ: <strong style="color:#2fca72;">115</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Исполнитель преобразует число на экране.<br>
У исполнителя есть три команды, которые обозначены латинскими буквами:<br>
<strong>A. Вычесть 1<br>
B. Вычесть 6<br>
C. Найти целую часть от деления на 2</strong><br><br>
Первая команда уменьшает число на экране на 1, вторая команда уменьшает это число на 6, третья команда делит число нацело на 2. Программа для исполнителя — это последовательность команд.<br>
Сколько существует таких программ, которые исходное число 34 преобразуют в число 6, и при этом траектория вычислений содержит числа 19 и 29 и не содержит числа 24?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>АВС</strong> при исходном числе 14 траектория состоит из чисел 13, 7, 3</div>
</div>

````python
def f(x, y):
    if x == y: return 1
    if x < y or x == 24:    # не через 24
        return 0
    # A: -1, B: -6, C: //2
    return f(x - 1, y) + f(x - 6, y) + f(x // 2, y)

# Должны пройти через 29 и 19:
# из 34 → 29 → 19 → 6
print(f(34, 29) * f(29, 19) * f(19, 6))   # Ответ: 115
````

<div class="task-box">
<div class="task-title">📋 Задание 2305 — Тип 3 — ограничение на повторы команд | Ответ: <strong style="color:#2fca72;">6</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Исполнитель преобразует число на экране.<br>
У исполнителя есть две команды, которые обозначены латинскими буквами:<br>
<strong>A. Прибавить 1<br>
B. Умножить на 2</strong><br><br>
Первая команда увеличивает число на экране на 1, вторая умножает его на 2. Программа для исполнителя — это последовательность команд.<br>
Сколько существует программ, которые преобразуют исходное число 1 в число 24, и при этом никакая команда не повторяется более двух раз подряд?<br><br>
Траектория вычислений программы — это последовательность результатов выполнения всех команд программы.<br>
Например, для программы <strong>BAB</strong> при исходном числе 1 траектория состоит из чисел 2, 3, 6</div>
</div>

````python
def f(x, y, path=""):
    # Запрещаем три подряд одинаковых команды
    if "AAA" in path or "BBB" in path:
        return 0
    if x == y: return 1
    if x > y:  return 0
    # A: +1, B: ×2
    return f(x + 1, y, path + "A") + f(x * 2, y, path + "B")

print(f(1, 24))               # Ответ: 6
````
