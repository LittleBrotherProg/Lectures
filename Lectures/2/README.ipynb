### Алгоритм сопоставления столбцов

1. **Столбец из одних нулей** → переменная, всегда 0 в нужных строках
2. **Столбец из одних единиц** → переменная, всегда 1
3. **Пустой столбец** (не заполнен в условии) → переменная с разными значениями
4. **Паттерн импликации** A→B: строки (0,0), (0,1), (1,1) — но не (1,0)
