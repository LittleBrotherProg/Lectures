## Приоритет операций

> ⚠️ **Главная ловушка:** порядок операций в алгебре логики и в Python **не совпадает!** Операторы `<=` и `==` в Python выполняются **раньше** `not`, `and`, `or`.

### Алгоритм сопоставления столбцов

1. **Столбец из одних нулей** → переменная, которая всегда 0
2. **Столбец из одних единиц** → переменная, которая всегда 1
3. **Паттерн импликации** A→B: строки (0,0), (0,1), (1,1) — но не (1,0)
