## 1. Хранение информации



<div class="warn-box">
<strong style="color:#f0c040;">Единицы измерения:</strong><br>
1 байт = 8 = 2³ бит &nbsp;|&nbsp; 1 Кбайт = 2¹³ бит &nbsp;|&nbsp; 1 Мбайт = 2²³ бит &nbsp;|&nbsp; 1 Гбайт = 2³³ бит<br><br>
Бит — минимальная единица информации (0 или 1).<br>
Байт — минимальная <strong>адресуемая</strong> единица памяти (8 бит).

## 2. Ключевые формулы



| Формула | Что вычисляем |
|:---|:---|
| `N = 2^i` | Мощность алфавита (кол-во символов) |
| `i = ceil(log₂N)` | Вес одного символа в битах |
| `I = L · i` | Объём сообщения в битах |
| `I_байт = ceil(L·i / 8)` | Объём в байтах (округление вверх) |

## 3. Правило округления — КЛЮЧЕВОЕ!



<div style="display:grid;grid-template-columns:1fr 1fr;gap:12px;margin:16px 0;">
<div style="background:#0d1b2a;border:1.5px solid #e05555;border-radius:10px;padding:14px;">
<div style="color:#e05555;font-weight:800;margin-bottom:6px;">«не более N» / «не превышает»</div>
Округляем <strong>вниз</strong> (<code>//</code> или <code>floor</code>)<br>
Нельзя выйти за лимит памяти
</div>
<div style="background:#0d1b2a;border:1.5px solid #2fca72;border-radius:10px;padding:14px;">
<div style="color:#2fca72;font-weight:800;margin-bottom:6px;">«более N» / «не менее N»</div>
Округляем <strong>вверх</strong> (<code>ceil</code>)<br>
Все символы должны уместиться
</div>
</div>

> **Вес символа `i = ceil(log₂N)` — всегда вверх!** Символ либо влезает, либо нет.

## 4. Тип 1 — Длина серийного номера

<div class="info-box">
<strong style="color:#1b91ff;">Дано:</strong> мощность алфавита N, объём памяти I, количество сообщений n<br>
<strong>Найти:</strong> максимальную или минимальную длину L<br>
<strong>Два способа:</strong> по формулам или перебором
</div>

````python
# По формулам (задание 1100)
# 10+52+963 алфавит, 2000 номеров ≤ 693 Кбайт, MAX длина?
from math import ceil, log2

N = 10 + 52 + 963        # мощность = 1025
I_total = 693 * 2**10    # 693 Кбайт → байты
count = 2000

i = ceil(log2(N))        # = 11 бит (всегда ceil!)
I = I_total // count     # вес одного (ВНИЗ! условие ≤)
L = I * 8 // i          # длина (ВНИЗ!)

print(L)                 # Ответ: 257
````

````python
# Через перебор (задание 1113)
# 10+17 алфавит, 7 564 230 > 31 Мбайт, MIN длина?
from math import ceil, log2

alphabet = 17 + 10
nums = 7_564_230
memory_size = 31 * 2**20    # байты

for L in range(1, 100):
    bits  = ceil(log2(alphabet))
    bytes = ceil(L * bits / 8)  # вверх! (условие >)
    if nums * bytes >= memory_size:
        print(L)                # Ответ: 7
        break
````

## 5. Тип 2 — Объём памяти

<div class="info-box">
<strong style="color:#1b91ff;">Дано:</strong> мощность N, длина L, количество n<br>
<strong>Найти:</strong> точный объём памяти<br>
<strong>Только по формулам</strong> — перебор не применяется
</div>

````python
# Задание 1101: 317 символов, 10+4090 алфавит, 262 144 записей
# Объём в Мбайт?
from math import ceil, log2

L = 317;  N = 10 + 4090;  n = 262_144

i     = ceil(log2(N))        # = 13 бит
I     = ceil(L * i / 8)      # вес одного (ВВЕРХ!)
I_all = I * n / 2**20        # в Мбайт

print(int(I_all))             # Ответ: 129
````

````python
# Задание 1110: 7 символов, 26+10, 9 байт доп., 30 пользователей
# Объём в байтах?
from math import ceil, log2

L = 7;  N = 26 + 10;  extra = 9;  users = 30

i      = ceil(log2(N))
I      = ceil(L * i / 8)          # вес пароля
I_all  = (I + extra) * users       # + доп. данные

print(I_all)                       # Ответ: 450
````

## 6. Тип 3 — Мощность алфавита

<div class="info-box">
<strong style="color:#1b91ff;">Дано:</strong> длина L, объём памяти I, количество n<br>
<strong>Найти:</strong> мин. или макс. мощность N<br>
<strong>Мин. N = 2^(i-1)+1  |  Макс. N = 2^i</strong>
</div>

| | Минимальная мощность | Максимальная мощность |
|:---|:---|:---|
| Формула | `2^(i-1) + 1` | `2^i` |
| Смысл | Первое значение N, которому нужно i бит | Последнее значение N при i битах |
| Условие задачи | «более» / «не менее» | «не более» / «не превышает» |

````python
# Задание 1103 (по формулам): 248 символов, 75 600 > 16 Мбайт
# Минимальная мощность?
from math import ceil

L = 248;  n = 75_600;  I_m = 16 * 2**20

I_one = ceil(I_m / n)         # вес одного (ВВЕРХ! — 'более')
i = ceil(I_one * 8 / L)       # вес символа (ВВЕРХ!)

print(2**(i-1) + 1)           # MIN мощность: Ответ 129
print(2**i)                   # MAX мощность: 256
````

````python
# Задание 1114 (перебор): 2783 символа, 3 845 627 ≥ 11 Гбайт
# Минимальная мощность?
from math import ceil, log2

L = 2783;  n = 3_845_627;  I_m = 11 * 2**30

for N in range(1, 1000):
    bits  = ceil(log2(N))
    bytes = ceil(L * bits / 8)
    if n * bytes >= I_m:         # 'не менее' → >=
        print(N)                 # Ответ: 257
        break
````
