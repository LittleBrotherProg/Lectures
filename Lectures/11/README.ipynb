## Ключевые формулы

```
i = ⌈log₂(N)⌉     — вес символа (бит) — ВСЕГДА ВВЕРХ!
I = ⌈L × i / 8⌉   — вес одного номера (байт) — ВВЕРХ (мин. целое число байт)
I_total = I × K    — общий объём для K номеров
```

**Таблица мощностей:**

| Алфавит | Мощность N |
|:---|:---:|
| Десятичные цифры | 10 |
| Латинских букв (с учётом регистра) | 52 |
| Латинских букв (без учёта регистра) | 26 |
| Спецсимволы (указано в условии) | по условию |
| **Итого** | **сумма всех** |

• Вес символа <code>i = ceil(log₂N)</code> — <strong>всегда вверх</strong><br>

• Длина <code>L</code> при «не более» — ВНИЗ (<code>//</code>)<br>
• Длина <code>L</code> при «более»/«не менее» — ВВЕРХ (<code>ceil</code>)</div>

## Тип 1 — Длина серийного номера

**По формулам:**
```python
from math import ceil, log2
i = ceil(log2(N))                    # вес символа
I = I_total // K                     # вес одного (ВНИЗ при «не более»)
L = I * 8 // i                       # длина (ВНИЗ)
```

**Перебором:**
```python
for L in range(1, 1000):
    I = ceil(L * i / 8)
    if K * I > memory:               # «более» → ≥
        print(L); break
```

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1100 — Тип 1 (разбор) — максимальная длина</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">На предприятии каждой изготовленной детали присваивают серийный номер, содержащий <strong>десятичные цифры</strong>, <strong>52 латинские буквы</strong> (с учётом регистра) и символы из <strong>963-символьного специального алфавита</strong>. В базе данных для хранения каждого серийного номера отведено одинаковое и минимально возможное число байт. При этом используется посимвольное кодирование серийных номеров, все символы кодируются одинаковым и минимально возможным числом бит. Известно, что для хранения <strong>2000 серийных номеров</strong> отведено <strong>не более 693 Кбайт</strong> памяти.<br><br>
Определите <strong>максимально возможную длину серийного номера</strong>. В ответе запишите только целое число.<br><br>
<strong>Разбор:</strong> N = 10+52+963 = 1025. i = ceil(log₂1025) = 11 бит. I_total = 693×2¹⁰ байт. I = I_total // 2000. L = I×8 // 11.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 257</div></div>

````python
from math import ceil, log2

N       = 10 + 52 + 963             # мощность алфавита
I_total = 693 * 2**10               # 693 Кбайт → байты
K       = 2000

i = ceil(log2(N))                   # вес символа (всегда вверх!)
I = I_total // K                    # вес одного (ВНИЗ — «не более»)
L = I * 8 // i                     # длина (ВНИЗ)

print(L)                            # Ответ: 257
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1104 — Тип 1 (пример) — минимальная длина</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">При регистрации в компьютерной системе каждому объекту присваивается идентификатор, содержащий <strong>десятичные цифры</strong>, <strong>26 латинских букв</strong> в верхнем регистре и символы из <strong>2013-символьного специального алфавита</strong>. В базе данных для хранения каждого идентификатора отведено одинаковое и минимально возможное число байт. При этом используется посимвольное кодирование идентификаторов, все символы кодируются одинаковым и минимально возможным числом бит. Известно, что для хранения <strong>2050 идентификаторов</strong> отведено <strong>более 709 Кбайт</strong> памяти.<br><br>
Определите <strong>минимально возможную длину идентификатора</strong>. В ответе запишите только целое число.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 237</div></div>

````python
from math import ceil, log2

N  = 10 + 26 + 2013
I_total = 709 * 2**10
K  = 2050

i = ceil(log2(N))
I = ceil(I_total / K)               # ВВЕРХ — «более»
L = ceil(I * 8 / i)                 # ВВЕРХ — минимальная длина

print(L)                            # Ответ: 237
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1113 — Тип 1 (перебор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">На предприятии каждой изготовленной детали присваивают серийный номер, содержащий <strong>десятичные цифры</strong> и символы из <strong>17-символьного специального алфавита</strong>. В базе данных каждый серийный номер занимает одинаковое и минимально возможное число байт. При этом используется посимвольное кодирование серийных номеров, все символы кодируются одинаковым и минимально возможным числом бит. Известно, что для хранения <strong>7 564 230 серийных номеров</strong> требуется <strong>более 31 Мбайт</strong> памяти.<br><br>
Определите <strong>минимально возможную длину серийного номера</strong>.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 7</div></div>

````python
from math import ceil, log2

N       = 10 + 17
K       = 7_564_230
memory  = 31 * 2**20                # байты

i = ceil(log2(N))

for L in range(1, 1000):
    I = ceil(L * i / 8)             # вес одного номера (вверх)
    if K * I > memory:              # «более»
        print(L)                    # Ответ: 7
        break
````

## Тип 2 — Объём памяти для K записей

```python
from math import ceil, log2

i     = ceil(log2(N))
I     = ceil(L * i / 8)             # байт на один номер
I_all = I * K                       # общий объём
print(I_all // 2**10)               # в Кбайт
print(I_all // 2**20)               # в Мбайт
```

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1101 — Тип 2 (разбор) — объём в Мбайт</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">При регистрации в компьютерной системе каждому объекту присваивается идентификатор, состоящий из <strong>317 символов</strong> и содержащий только десятичные цифры и символы из <strong>4090-символьного специального алфавита</strong>. В базе данных для хранения каждого идентификатора отведено одинаковое и минимально возможное целое число байт. При этом используется посимвольное кодирование идентификаторов, все символы кодируются одинаковым и минимально возможным количеством бит.<br><br>
Определите объём памяти (в Мбайт), необходимый для хранения <strong>262 144 идентификаторов</strong>. В ответе запишите только целое число — количество Мбайт.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 129</div></div>

````python
from math import ceil, log2

L = 317
N = 10 + 4090
K = 262_144

i     = ceil(log2(N))               # вес символа = 13 бит
I     = ceil(L * i / 8)            # вес одного (байт)
I_all = I * K / 2**20              # в Мбайт

print(int(I_all))                   # Ответ: 129
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1110 — Тип 2 (пример) — с доп. данными</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">При регистрации в компьютерной системе каждому пользователю выдаётся пароль, состоящий из <strong>7 символов</strong> и содержащий только символы <strong>26-символьного набора строчных латинских букв</strong> и <strong>десяти арабских цифр</strong>. В базе данных для хранения сведений о каждом пользователе отведено одинаковое и минимально возможное целое число байт. При этом используют посимвольное кодирование паролей, все символы кодируют одинаковым и минимально возможным количеством бит. Кроме собственно пароля, для каждого пользователя в системе хранятся дополнительные сведения, для чего отведено <strong>9 байт</strong> на одного пользователя.<br><br>
Определите объём памяти (в байтах), необходимый для хранения сведений о <strong>30 пользователях</strong>.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 450</div></div>

````python
from math import ceil, log2

L     = 7
N     = 26 + 10
extra = 9
users = 30

i     = ceil(log2(N))
I     = ceil(L * i / 8)            # вес пароля (байт)
total = (I + extra) * users        # + доп. данные

print(total)                        # Ответ: 450
````

## Тип 3 — Мощность алфавита

**По формулам:**
```python
I_one = ceil(I_total / K)          # вес одного (вверх при «более»)
i     = ceil(I_one * 8 / L)        # вес символа (вверх)
N_min = 2**(i-1) + 1               # минимальная мощность
N_max = 2**i                       # максимальная мощность
```

**Перебором (для максимальной — от большего):**
```python
for N in range(1, 10000)[::-1]:
    i = ceil(log2(N))
    if K * ceil(L * i / 8) <= memory:   # «не более»
        print(N); break
```

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1103 — Тип 3 (разбор) — минимальная мощность</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">На предприятии каждой изготовленной детали присваивается серийный номер, состоящий из <strong>248 символов</strong>. В базе данных для хранения каждого серийного номера отведено одинаковое и минимально возможное число байт. При этом используется посимвольное кодирование серийных номеров, все символы кодируются одинаковым и минимально возможным числом бит. Известно, что для хранения <strong>75 600 серийных номеров</strong> требуется <strong>более 16 Мбайт</strong> памяти.<br><br>
Определите <strong>минимально возможную мощность алфавита</strong>, используемого для записи серийных номеров.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 129</div></div>

````python
from math import ceil

L = 248
K = 75_600
I_m = 16 * 2**20                   # байты

I_one = ceil(I_m / K)              # ВВЕРХ — «более»
i     = ceil(I_one * 8 / L)        # ВВЕРХ
print(2**(i-1) + 1)                # Мин. мощность: 129
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1107 — Тип 3 (пример) — максимальная мощность (перебор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">На предприятии каждой изготовленной детали присваивают серийный номер, состоящий из <strong>246 символов</strong>. В базе данных для хранения каждого серийного номера отведено одинаковое и минимально возможное число байт. При этом используется посимвольное кодирование серийных номеров, все символы кодируются одинаковым и минимально возможным числом бит. Известно, что для хранения <strong>703 569 серийных номеров</strong> доступно <strong>не более 77 Мбайт</strong> памяти.<br><br>
Определите <strong>максимально возможную мощность алфавита</strong>, используемого для записи серийных номеров.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 8</div></div>

````python
from math import ceil, log2

L      = 246
K      = 703_569
memory = 77 * 2**20                # байты

# Перебираем от большего к меньшему (ищем МАКСИМАЛЬНУЮ)
for N in range(1, 10000)[::-1]:
    i = ceil(log2(N))
    if K * ceil(L * i / 8) <= memory:  # «не более»
        print(N)                    # Ответ: 8
        break
````
