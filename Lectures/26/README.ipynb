Дан набор объектов с числовым параметром. Нужно выбрать максимальное количество объектов, которые можно вложить друг в друга, при условии что **разница параметров соседних объектов не менее заданного порога**.

### Почему работает жадный подход?

После сортировки по убыванию каждый следующий объект ≤ предыдущего. Берём первый. Для каждого следующего: если разница с последним взятым ≥ порог — берём, иначе пропускаем. Пропуск «маленького» объекта не ухудшает результат — следующие объекты будут ещё меньше и тоже могут подойти.

````python
# ─── Решение задания 2603 ───────────────────────────────────────────
data = []
with open("2603.txt") as file:
    N, S = map(int, file.readline().split())  # N кандидатов, S мест

    for line in file:
        parts = line.strip().split()
        candidate = {
            "id"       : int(parts[0]),
            "exam_1"   : int(parts[1]),
            "exam_2"   : int(parts[2]),
            "exam_3"   : int(parts[3]),
            "interview": int(parts[4])
        }
        candidate["result"] = (
            candidate["exam_1"] +
            candidate["exam_2"] +
            candidate["exam_3"]
        )
        data.append(candidate)

# Сортировка: убыв. сумма → убыв. собеседование → возр. ID
data = sorted(data, key=lambda x: (-x["result"], -x["interview"], x["id"]))

# Полупроходной балл = результат кандидата на позиции S (индекс S)
half_pass_mark = data[S]["result"]

answer_1 = {}
answer_2 = 0

for candidate in data:
    if candidate["result"] > half_pass_mark:
        answer_1 = candidate   # перезаписываем — нужен ПОСЛЕДНИЙ
    elif candidate["result"] == half_pass_mark:
        answer_2 += 1
    else:
        break

print(answer_1["id"])
print(answer_2)
````

````python
# ─── Решение задания 2604 ───────────────────────────────────────────
import statistics

data = []
with open("2604.txt") as file:
    N = int(file.readline())

    for line in file:
        parts = line.strip().split()
        student = {
            "id" : int(parts[0]),
            "m_1": int(parts[1]),
            "m_2": int(parts[2]),
            "m_3": int(parts[3]),
            "m_4": int(parts[4])
        }
        marks = (student["m_1"], student["m_2"],
                 student["m_3"], student["m_4"])
        student["twos"] = marks.count(2)
        student["mean"] = statistics.mean(marks)
        data.append(student)

# Сортировка: убыв. средний балл → возр. количество двоек → возр. ID
data = sorted(data, key=lambda x: (-x["mean"], x["twos"], x["id"]))

# Последний в первой четверти (25%)
quart = N // 4
print(data[quart - 1]["id"])

# Первый с более чем двумя двойками
for student in data:
    if student["twos"] > 2:
        print(student["id"])
        break
````

### Суть задачи

Зал M рядов × K мест. Часть мест занята. Найти пару **соседних** мест, перед которыми **нет занятых** мест с теми же номерами. Ряд — **максимально дальний** от сцены.

### Ключевая идея: первый занятый ряд для каждого места

Для места №j: `min_row[j]` — наименьший ряд, где место j занято.

Допустимый ряд для места j = `min_row[j] - 1` (ряд перед первым занятым).

Для **пары** (j, j+1): допустимый ряд = `min(min_row[j], min_row[j+1]) - 1`.

Ищем **максимальный** такой ряд среди всех пар.

````python
# ─── Решение задания 2602 ───────────────────────────────────────────
with open("2602.txt") as file:
    N, M, K = map(int, file.readline().split())
    data = [list(map(int, line.split())) for line in file]

max_row  = 0
min_seat = 0

# Первый занятый ряд для каждого места (индекс = номер места)
# M+1 = фиктивное "нет занятых", K+2 = чтобы seat+1 не вышел за границу
min_row = [M + 1] * (K + 2)

for row, seat in data:
    if row < min_row[seat]:
        min_row[seat] = row   # запоминаем наименьший занятый ряд

for seat in range(1, K):
    accept_row = min(min_row[seat], min_row[seat + 1]) - 1
    if (accept_row > max_row or
            (accept_row == max_row and seat < min_seat)):
        max_row  = accept_row
        min_seat = seat

print(max_row, min_seat)
````

````python
# ─── Решение задания 2606 ───────────────────────────────────────────
# Два отличия от 2602:
# 1. При равном ряду берём БОЛЬШЕЕ место (seat > max_seat)
# 2. Выводим max_seat + 1 (правое место пары)
with open("2606.txt") as file:
    N, M, K = map(int, file.readline().split())
    data = [list(map(int, line.split())) for line in file]

max_row  = 0
max_seat = 0

min_row = [M + 1] * (K + 2)
for row, seat in data:
    if row < min_row[seat]:
        min_row[seat] = row

for seat in range(1, K):
    accept_row = min(min_row[seat], min_row[seat + 1]) - 1
    if (accept_row > max_row or
            (accept_row == max_row and seat > max_seat)):  # ← изменение
        max_row  = accept_row
        max_seat = seat

print(max_row, max_seat + 1)  # ← +1: правое место пары
````

````python
# Создаём типовой пример для 2609
example = """5
10 150
100 120
131 170
150 180
120 130"""
with open("2609_example.txt", "w") as f:
    f.write(example)
print("Файл 2609_example.txt создан. Ожидаемый ответ: 3 20")
````

````python
# ─── Решение задания 2609 ───────────────────────────────────────────
# Данные: время начала и время ОКОНЧАНИЯ
data = []
with open("2609.txt") as file:
    N = int(file.readline())
    for line in file:
        parts = line.strip().split()
        event = {"start": int(parts[0]), "end": int(parts[1])}
        data.append(event)

# Сортируем по возрастанию конца, при равенстве — убывание начала
# (убывание начала = при равном конце берём более короткое мероприятие)
events = sorted(data, key=lambda e: (e["end"], -e["start"]))

# Первое мероприятие всегда берём
approved = [events[0]]

# Жадный выбор: берём если начинается >= конца последнего выбранного
for event in events[1:]:
    if event["start"] >= approved[-1]["end"]:
        approved.append(event)

# Ответ 1: количество мероприятий
print(len(approved))

# Ответ 2: максимальный перерыв между двумя ПОСЛЕДНИМИ
# Ищем среди ВСЕХ событий, начинающихся >= конца предпоследнего
longest_break = 0
penultimate_end = approved[-2]["end"]
for event in events:
    if event["start"] >= penultimate_end:
        gap = event["start"] - penultimate_end
        longest_break = max(longest_break, gap)

print(longest_break)
````
