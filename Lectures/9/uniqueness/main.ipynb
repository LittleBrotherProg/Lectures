## Программное решение

### Подготовка файла

Для работы с данными в Python необходимо преобразовать файл в формат CSV.

**Процесс преобразования:**
![преобразование в CSV](images/g16.gif)

**Объяснение:** Сохраняем файл электронной таблицы в формате CSV для удобства чтения в Python.

### Размещение файла

Перемещаем CSV-файл в директорию с Python-скриптом:
![размещение файла](images/g17.gif)

**Объяснение:** Файл данных должен находиться в той же папке, что и исполняемый скрипт, для простоты обращения к нему.

```python
# Открываем файл '9_21895.csv' и читаем все строки в список
file = open('9_21895.csv').readlines()

# Инициализируем счетчик
c = 0

# Обрабатываем каждую строку из файла
for s in file:
    # Преобразуем строку в список целых чисел, разделенных запятыми
    # Сортируем список по возрастанию
    s = sorted(list(map(int, s.split(','))))
    
    # Создаем множество set() из элементов строки (убираем дубликаты)
    # Проверяем, что все числа в строке уникальны (длина множества равна длине списка)
    if len(set(s)) == len(s):
        
        # Сумма двух наибольших элементов (последние два в отсортированном списке)
        sum_max_s = s[-1] + s[-2]
        
        # Сумма всех остальных элементов (общая сумма минус два наибольших)
        sum_s = sum(s) - sum_max_s
        
        # Проверяем, что сумма двух наибольших не превышает сумму остальных
        if sum_max_s <= sum_s:
            c += 1

# Выводим количество строк, удовлетворяющих условиям
print(c)
```

## Код без коментариев

```python
file = open('9_21895.csv').readlines()
c = 0
for s in file:
    s = sorted(list(map(int, s.split(','))))
    if len(set(s)) == len(s):
        sum_max_s = s[-1] + s[-2]
        sum_s = sum(s)-sum_max_s
        if sum_max_s <= sum_s:
            c+=1
print(c)
```
