## 1. Кодирование информации

<div class="warn-box">
<strong>Равномерное:</strong> все символы кодируются одинаковым числом бит (А=00, Б=01, В=10)<br>
<strong>Неравномерное:</strong> частым символам — короткие коды (А=0, Б=10, В=110). Экономичнее, но нужно условие Фано!
</div>

## 2. Условие Фано

**Прямое условие Фано:** ни одно кодовое слово не является **НАЧАЛОМ** другого.
- А=0, Б=10, В=11 ✓ — каждый код уникален
- А=0, Б=01 ✗ — код «А» (0) является началом «Б» (01)

**Обратное условие Фано:** ни одно слово не является **КОНЦОМ** другого (постфиксный код).

## 3. Двоичное дерево

| Термин | Определение |
|:---|:---|
| Корень | Начальный узел, не содержит кода буквы |
| Лист | Конечный узел — единственное место для буквы |
| Левая ветвь | Добавляет **0** к коду |
| Правая ветвь | Добавляет **1** к коду |

## 4. Блокировка узлов

При размещении кода **«110»** блокируются:
- **Предки:** `1`, `11` — содержат начало нашего кода
- **Потомки:** `1100`, `1101`... — наш код является их началом
- Только незаблокированные **листья** подходят для новых букв

---
## 5. Алгоритм решения

```
1. Построить дерево нужной глубины
2. Отметить все известные коды (занятые листья)
3. Заблокировать предков и потомков занятых листьев
4. Среди свободных листьев найти кратчайший
5. Тип 2: Σ (длина кода × кол-во повторений буквы в слове)
```

> **Ключевое правило Типа 2:** частым буквам — более короткие коды!

## 7. Тип 2 — Минимальная длина кодирования слова

<div class="task-box">
<div class="task-title">📋 Задание 400 — Тип 2 — КОЛОКОЛ | Ответ: <strong style="color:#2fca72;">13</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">По каналу связи передаются сообщения, содержащие только буквы: Б, К, Л, О, Н. Для передачи используется двоичный код, удовлетворяющий условию Фано. Кодовые слова для некоторых букв известны: Б&nbsp;–&nbsp;1001, К&nbsp;–&nbsp;11.<br><br>
Для трёх оставшихся букв Л, Н и О кодовые слова неизвестны.<br><br>
<strong>Какое наименьшее количество двоичных знаков требуется для кодирования слова КОЛОКОЛ?</strong></div>
</div>

````python
# Задание 400: КОЛОКОЛ (К-2, О-3, Л-2), ответ: 13
# Известно: Б=1001, К=11
# Свободные листья: 0 (для О), 101 (для Л), 1000 (для Н)
#
# Распределяем по частоте:
# О (3 повторения) → код 0   (1 знак)  → 1×3 = 3
# К (2 повторения) → код 11  (2 знака) → 2×2 = 4
# Л (2 повторения) → код 101 (3 знака) → 3×2 = 6
# Итого: 3 + 4 + 6 = 13
````

<div class="task-title">📋 Задание 406 — Тип 2 — КОРОБОК | Ответ: <strong style="color:#2fca72;">17</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">По каналу связи передаются сообщения, содержащие только буквы из набора: Б, К, Р, О, Н. Для передачи используется двоичный код, удовлетворяющий условию Фано. Кодовые слова для некоторых букв известны: Б&nbsp;–&nbsp;10, Н&nbsp;–&nbsp;110, Р&nbsp;–&nbsp;000. Для двух оставшихся букв К и О кодовые слова неизвестны.<br><br>
<strong>Какое количество двоичных знаков требуется для кодирования слова КОРОБОК, если известно, что оно закодировано минимально возможным количеством двоичных знаков?</strong></div>
</div>

````python
# Задание 406: КОРОБОК (К-2, О-3, Р-1, Б-1), ответ: 17
# Свободные листья: 01, 001, 111
# О (3 повт.) → 01  (2 знака) → 2×3 = 6
# К (2 повт.) → 111 (3 знака) → 3×2 = 6
# Р (1 повт.) → 000 (3 знака) → 3×1 = 3
# Б (1 повт.) → 10  (2 знака) → 2×1 = 2
# Итого: 6 + 6 + 3 + 2 = 17
````

<div class="task-title">📋 Задание 408 — Тип 2 — ВЕРМЕЕР | Ответ: <strong style="color:#2fca72;">18</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">По каналу связи передаются сообщения, содержащие только буквы из набора: В, Е, М, Н, Р. Для передачи используется двоичный код, удовлетворяющий условию Фано. Это условие обеспечивает возможность однозначной расшифровки закодированных сообщений. Кодовые слова для некоторых букв известны: В&nbsp;–&nbsp;1, М&nbsp;–&nbsp;001. Для трёх оставшихся букв Е, Н и Р кодовые слова неизвестны.<br><br>
<strong>Какое количество двоичных знаков потребуется для кодирования слова ВЕРМЕЕР, если известно, что оно закодировано минимально возможным количеством двоичных знаков?</strong></div>
</div>

````python
# Задание 408: ВЕРМЕЕР (В-1, Е-3, Р-2, М-1), ответ: 18
# В=1 (занято), М=001 (занято)
# Свободные варианты: Е→01(2зн) или Е→3зн
#
# Вариант 1: Е=01(2), тогда Р=4 знака
# В(1)×1=1, Е(2)×3=6, Р(4)×2=8, М(3)×1=3 → итого 18
#
# Вариант 2: все по 3 знака
# В(1)×1=1, Е(3)×3=9, Р(3)×2=6, М(3)×1=3 → итого 19
#
# Выбираем вариант 1: 18
````
