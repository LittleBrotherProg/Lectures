**Два типа задач:**

| Тип | Что дано | Что найти |

| 1 | Коды других букв | Кратчайший код для одной буквы |
| 2 | Некоторые коды + слово | Мин. число знаков для кодирования слова |

## Теория: равномерное и неравномерное кодирование

**Равномерное кодирование** — все символы кодируются одинаковым числом бит.
Пример: А=00, Б=01, В=10, Г=11.

**Неравномерное кодирование** — частым символам назначаются короткие коды, редким — длинные. Экономичнее, но требует условие Фано для однозначного декодирования.

## Условие Фано

**Прямое условие Фано:** ни одно кодовое слово не является **началом** другого.

Примеры:
- А=0, Б=10, В=11 ✅ — ни одно не начало другого
- А=0, Б=01 ❌ — код «А» (0) является началом «Б» (01)

**Обратное условие Фано:** ни одно слово не является **концом** другого.

## Двоичное дерево

Коды строятся по двоичному дереву: левая ветвь добавляет `0`, правая — `1`.
Буква может находиться **только в листе** (конечном узле). Промежуточные узлы заняты.

**Блокировка узлов:** при размещении буквы с кодом `110` автоматически блокируются:
- **Предки** `1`, `11` — содержат начало нашего кода → нельзя
- **Потомки** `1100`, `1101`... — наш код является их началом → нельзя

## Тип 1 — Найти кратчайший код для буквы

**Алгоритм:**
1. Нарисовать дерево нужной глубины
2. Отметить все известные коды (занятые листья)
3. Заблокировать предков и потомков каждого занятого листа
4. Среди свободных листьев найти кратчайший
5. Если несколько — выбрать с наибольшим/наименьшим числовым значением по условию

## Тип 2 — Минимальная длина кодирования слова

**Принцип:** частым буквам назначать более **короткие** коды.

**Алгоритм:**
1. Подсчитать, сколько раз каждая буква встречается в слове
2. Найти все свободные листья в дереве (после блокировки известных кодов)
3. Назначить: самая частая буква → самый короткий свободный лист
4. Ответ = Σ (длина кода × частота буквы)

## Шпаргалка

### Условие Фано
- Ни одно слово не является **началом** другого
- А=0, Б=10, В=11 ✅
- А=0, Б=01 ❌ (0 — начало 01)

### Блокировка при размещении кода «110»
- Предки: `1`, `11` — заблокированы
- Потомки: `1100`, `1101`... — заблокированы

### Тип 1 — кратчайший код
```
1. Отметить занятые листья
2. Заблокировать их предков и потомков
3. Найти кратчайший свободный лист
4. При равенстве: наименьшее/наибольшее числовое значение по условию
```

### Тип 2 — длина слова
```
Ответ = Σ (длина кода × кол-во вхождений буквы)
Частым буквам — короткие коды!
```
