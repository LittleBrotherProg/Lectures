## 1. Кодирование информации

Вся информация в компьютере хранится в виде **0 и 1**. Для человека эти данные неудобны — нужно **декодирование**.

<div style="display:grid;grid-template-columns:1fr 1fr;gap:12px;margin:16px 0;">
<div style="background:#0d1b2a;border:1.5px solid #1b91ff;border-radius:10px;padding:14px;">
<div style="color:#1b91ff;font-weight:800;margin-bottom:6px;">Равномерное кодирование</div>
Все символы кодируются <strong>одинаковым</strong> числом бит.<br>
<code>А=00, Б=01, В=10</code><br>
Просто реализовать, но неэффективно по памяти.
</div>
<div style="background:#0d1b2a;border:1.5px solid #e07b39;border-radius:10px;padding:14px;">
<div style="color:#e07b39;font-weight:800;margin-bottom:6px;">Неравномерное кодирование</div>
Частым символам — <strong>короткие</strong> коды.<br>
<code>А=0, Б=10, В=110</code><br>
Экономичнее, но нужно условие Фано!
</div>
</div>

> **Пример:** ACABAB → А(1)×3 + В(3)×1 + Б(2)×2 = **10 бит** при неравномерном vs **18 бит** при равномерном

---
## 2. Условие Фано



<div class="tip-box">
<strong style="color:#2fca72;">Прямое условие Фано:</strong> ни одно кодовое слово не является <strong>НАЧАЛОМ</strong> другого.<br>
→ Префиксный код. Гарантирует однозначное декодирование.<br><br>
<strong style="color:#8ab4d4;">Обратное условие Фано:</strong> ни одно кодовое слово не является <strong>КОНЦОМ</strong> другого.<br>
→ Постфиксный код.
</div>

**Пример нарушения:** А=0, Б=01 — код А является началом кода Б → 01 можно прочитать как «АВ» или «Б».

---
## 3. Двоичное дерево



| Термин | Определение |
|:---|:---|
| Корень | Начальный узел, без кода буквы |
| Лист | Конечный узел — единственное место для буквы |
| Левая ветвь | Добавляет **0** к коду |
| Правая ветвь | Добавляет **1** к коду |
| Уровень | Длина кода = уровень узла |

---
## 4. Блокировка узлов



<div class="warn-box">
<strong style="color:#f0c040;">При размещении кода «110» блокируются:</strong><br>
• <strong>Предки:</strong> 1, 11 — они содержат начало нашего кода<br>
• <strong>Потомки:</strong> 1100, 1101... — они содержат наш код как начало<br>
• Сам узел 110 — занят<br><br>
Свободными остаются только незаблокированные листья.
</div>

---
## 5. Алгоритм решения



```
1. Построить дерево нужной глубины
2. Отметить все известные коды (занятые листья)
3. Заблокировать предков и потомков занятых листьев
4. Среди свободных листьев найти кратчайший
5. Тип 2: suммировать длина_кода × частота_буквы
```

> **Важно:** если подходящих кодов несколько — внимательно читай условие: «наибольшее» или «наименьшее» числовое значение?

После заполнения — единственный свободный лист: **10**.

## 7. Тип 2 — Длина закодированного слова

<div class="warn-box">
<strong style="color:#f0c040;">Ключевое правило:</strong> частым буквам — более короткие коды!<br>
Ответ = Σ (длина_кода × количество_повторений)
</div>



### Задание 400 — КОЛОКОЛ, ответ: **13**
К=11, Б=1001. Слово КОЛОКОЛ: К-2, О-3, Л-2.

О получает код **0** (1 знак) × 3 = 3  
К получает код **11** (2 знака) × 2 = 4  
Л получает код **101** (3 знака) × 2 = 6  
**Итого: 3+4+6 = 13**

### Задание 406 — КОРОБОК, ответ: **17**
Б=10, Н=110, Р=000. Слово КОРОБОК: К-2, О-3, Р-1, Б-1.

О→01(2)×3=6 + К→111(3)×2=6 + Р→000(3)×1=3 + Б→10(2)×1=2 = **17**

### Задание 408 — ВЕРМЕЕР, ответ: **18**
В=1, М=001. Слово ВЕРМЕЕР: В-1, Е-3, Р-2, М-1.

Рассматриваем варианты: Е→01(2)×3=6 выгоднее чем Е→(3)×3=9.  
В=1(1)×1=1 + Е=01(2)×3=6 + Р=0001(4)×2=8 + М=001(3)×1=3 = **18**
