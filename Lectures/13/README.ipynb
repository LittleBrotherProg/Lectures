## Основные термины

**IP-адрес** (IPv4) — **32-битное число**, записываемое в виде четырёх десятичных чисел (0–255) через точку. Каждое число — **октет**. Например: `192.168.0.1`.

IP-адрес делится на две части: **адрес сети** (общий для всех устройств сети) и **адрес узла** (идентифицирует конкретное устройство). Граница между частями определяется **маской подсети**.

**Маска подсети** — **32-битное число**, в котором:
- биты = **1** → часть IP, относящаяся к **адресу сети**
- биты = **0** → часть IP, относящаяся к **адресу узла**

Единицы в маске всегда **непрерывны** и стоят в начале. Допустимые значения октета маски: `0, 128, 192, 224, 240, 248, 252, 254, 255`.

**CIDR-нотация** — запись вида `IP/префикс`, где **префикс** = количество единиц в маске (0–32):

| Запись | Маска | Кол-во адресов |
|:---:|:---|:---:|
| /24 | 255.255.255.0 | 256 |
| /28 | 255.255.255.240 | 16 |
| /16 | 255.255.0.0 | 65 536 |
| /8 | 255.0.0.0 | 16 777 216 |

**Формула:** количество адресов = 2^(32 − префикс)

**Адрес сети** — IP с нулями во всей части узла. Получается операцией `IP AND маска`.

**Широковещательный адрес** — IP с единицами во всей части узла. Получается операцией `адрес_сети OR NOT(маска)`. Оба адреса **зарезервированы** и не могут назначаться устройствам.

## Модуль `ipaddress`

Модуль встроен в Python и содержит три ключевых класса. Создаются через фабричные функции:

### `ip_address()` — IP-адрес

```python
from ipaddress import ip_address

ip = ip_address("192.168.0.1")
print(int(ip))          # → 3232235521  (целое число)
print(f'{int(ip):032b}') # → двоичная запись (32 бита)

# Первые и последние 2 байта:
bin_ip = f'{int(ip):032b}'
left  = bin_ip[:16]   # левые два байта
right = bin_ip[16:]   # правые два байта
```

### `ip_network()` — IP-сеть

```python
from ipaddress import ip_network

# Создание по адресу сети:
net = ip_network('192.168.1.0/255.255.255.0')

# Создание по адресу узла (strict=False обнуляет биты узла):
net = ip_network('192.168.1.5/255.255.255.0', strict=False)

print(net.network_address)   # → 192.168.1.0
print(net.broadcast_address) # → 192.168.1.255
print(net.prefixlen)         # → 24
print(net.num_addresses)     # → 256
print(net.netmask)           # → 255.255.255.0

# Итерация по всем адресам (включая сеть и broadcast):
for ip in net: ...

# Только хосты (без сети и broadcast):
for ip in net.hosts(): ...

# Доступ по индексу:
net[-2]  # предпоследний = наибольший адрес хоста
net[1]   # первый хост
```

### `ip_interface()` — IP-интерфейс (адрес + маска)

```python
from ipaddress import ip_interface

iface = ip_interface('192.168.1.10/255.255.255.0')

print(iface.ip)                       # → 192.168.1.10  (адрес узла)
print(iface.network.network_address)  # → 192.168.1.0   (адрес сети)
print(iface.network.prefixlen)        # → 24
print(iface.network[-2])             # → 192.168.1.254  (наибольший хост)
```

## Тип 2 — Два адреса узлов → маска или количество адресов

**Идея:** перебирать значения префикса (обычно от 30 вниз), строить сеть по первому адресу и проверять, что второй адрес тоже в неё входит, причём оба — не служебные (через `hosts()`).

```python
for mask in range(30, -1, -1):
    net = ip_network(f'{ip1}/{mask}', strict=False)
    if ip2 in net and ip1 in net.hosts() and ip2 in net.hosts():
        print(mask)         # кол-во единиц в маске = длина префикса
        print(net.num_addresses)  # кол-во адресов в сети
        break
```

> Если узлы должны быть в **разных** сетях — строим обе сети и проверяем `net_1 != net_2`.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1319 — Тип 2 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Два узла, находящиеся в одной сети, имеют IP-адреса <strong>127.0.1.65</strong> и <strong>127.0.1.101</strong>.<br>Укажите наименьшее возможное количество адресов в этой сети.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 64</div></div></div>

````python
# Задание 1319 — наименьшее кол-во адресов => наибольший префикс => перебор от 30 вниз
from ipaddress import ip_network, ip_address

ip1 = ip_address('127.0.1.65')
ip2 = ip_address('127.0.1.101')

for mask in range(30, -1, -1):
    net = ip_network(f'{ip1}/{mask}', strict=False)
    if ip2 in net and ip1 in net.hosts() and ip2 in net.hosts():
        print(net.num_addresses)   # Ответ: 64
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1333 — Тип 2 (пример — разные сети)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Два узла, находящиеся в <strong>разных</strong> подсетях, имеют IP-адреса <strong>192.168.34.17</strong> и <strong>192.168.45.91</strong>. Известно, что в масках обеих подсетей содержится одинаковое количество единиц.<br>Определите наименьшее возможное количество единиц в масках этих подсетей.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 21</div></div></div>

````python
# Задание 1333 — разные сети, ищем наименьший префикс => перебор от 16 вверх
from ipaddress import ip_network, ip_address

ip1 = ip_address('192.168.34.17')
ip2 = ip_address('192.168.45.91')

for mask in range(16, 31):
    net1 = ip_network(f'{ip1}/{mask}', strict=False)
    net2 = ip_network(f'{ip2}/{mask}', strict=False)
    if (net1 != net2 and
            ip1 in net1.hosts() and
            ip2 in net2.hosts()):
        print(mask)   # Ответ: 21
        break
````

## Тип 3 — Адрес узла + адрес сети → маска

**Идея:** перебрать значения префикса, на каждом шаге строить сеть по адресу узла и проверять двойное условие:
1. Адрес узла входит в `net.hosts()` (не служебный)
2. Адрес сети совпадает с `net.network_address`

Затем из подходящих масок взять `min` или `max` по нужному критерию.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1307 — Тип 3 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для узла с IP-адресом <strong>192.168.24.72</strong> адрес сети равен <strong>192.168.24.64</strong>.<br>Определите наименьшее возможное количество единиц в последних двух байтах маски.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 10</div></div></div>

````python
# Задание 1307 — ищем наименьшее кол-во единиц в последних двух байтах маски
from ipaddress import ip_network, ip_address

ip_net  = ip_address('192.168.24.64')
ip_host = ip_address('192.168.24.72')

masks = []
for prefix in range(17, 31):
    net = ip_network(f'{ip_host}/{prefix}', False)
    if ip_host in net.hosts() and net.network_address == ip_net:
        masks.append(f'{int(net.netmask):032b}'[16:])

print(min(masks).count('1'))   # Ответ: 10
````

````python
# Задание 1311 — наибольшее кол-во единиц = наибольший префикс
from ipaddress import ip_network, ip_address

ip_net  = ip_address('44.44.224.0')
ip_host = ip_address('44.44.229.28')

prefixes = []
for prefix in range(17, 32):
    net = ip_network(f'{ip_host}/{prefix}', False)
    if ip_host in net.hosts() and net.network_address == ip_net:
        prefixes.append(net.prefixlen)

print(max(prefixes))   # Ответ: 21
````

## Тип 4 — Адрес сети + маска → количество адресов по условию

**Идея:** создать объект сети, перебрать все адреса в цикле `for ip in net:`, для каждого получить 32-битную двоичную строку и проверить условие.

```python
net = ip_network('172.16.0.0/255.255.0.0')
cnt = 0
for ip in net:
    b = f'{int(ip):032b}'
    if b.count('1') % 5 != 0:   # пример условия
        cnt += 1
print(cnt)
```

> Если адрес сети имеет ненулевые биты узла — передай `False` вторым аргументом.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1303 — Тип 4 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом <strong>172.16.168.0</strong> и маской сети <strong>255.255.248.0</strong>.<br>Сколько в этой сети IP-адресов, для которых количество единиц в двоичной записи IP-адреса <strong>не кратно 5</strong>?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 1663</div></div></div>

````python
# Задание 1303
from ipaddress import ip_network

net = ip_network('172.16.168.0/255.255.248.0')

cnt = 0
for ip in net:
    b = f'{int(ip):032b}'
    if b.count('1') % 5 != 0:
        cnt += 1

print(cnt)   # Ответ: 1663
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1310 — Тип 4 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом <strong>142.108.56.118</strong> и сетевой маской <strong>255.255.255.240</strong>.<br>Сколько в этой сети IP-адресов, для которых суммарное количество единиц в <strong>левых двух байтах</strong> меньше суммарного количества единиц в <strong>правых двух байтах</strong>?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 5</div></div></div>

````python
# Задание 1310 — сеть задана адресом узла, нужен strict=False
from ipaddress import ip_network

net = ip_network('142.108.56.118/255.255.255.240', False)

cnt = 0
for ip in net:
    b = f'{int(ip):032b}'
    if b[:16].count('1') < b[16:].count('1'):
        cnt += 1

print(cnt)   # Ответ: 5
````

## Тип 5 — Неизвестный байт маски

**Идея:** перебрать допустимые значения префикса в диапазоне, соответствующем неизвестному байту. Если неизвестен **третий** байт маски — перебирать префиксы от 16 до 24. Если **четвёртый** — от 24 до 32.

Для каждого значения проверить, что **все** IP-адреса сети удовлетворяют условию (`all(...)`). При первом выполнении вывести значение нужного байта маски и завершить цикл.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1313 — Тип 5 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом <strong>191.239.130.3</strong> и маской сети <strong>255.255.А.0</strong>, где А — допустимое значение байта маски.<br>Определите минимальное значение А, для которого для всех IP-адресов сети количество единиц в <strong>левых двух байтах ≥ количества единиц в правых двух байтах</strong>.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 224</div></div></div>

````python
# Задание 1313 — неизвестен 3-й байт маски => префикс 16..24
from ipaddress import ip_network, ip_address

ip_net = ip_address('191.239.130.3')

for prefix in range(16, 25):
    net = ip_network(f'{ip_net}/{prefix}', False)
    if all(f'{int(ip):032b}'[:16].count('1') >= f'{int(ip):032b}'[16:].count('1')
           for ip in net):
        print(str(net.netmask).split('.')[2])   # Ответ: 224
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1314 — Тип 5 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом <strong>255.215.88.80</strong> и маской сети <strong>255.255.А.0</strong>.<br>Определите минимальное значение А, для которого для всех IP-адресов сети количество нулей в <strong>левых двух байтах ≤ количества нулей в правых двух байтах</strong>.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 224</div></div></div>

````python
# Задание 1314 — условие на нули, знак <=
from ipaddress import ip_network, ip_address

ip_net = ip_address('255.215.88.80')

for prefix in range(16, 25):
    net = ip_network(f'{ip_net}/{prefix}', False)
    if all(f'{int(ip):032b}'[:16].count('0') <= f'{int(ip):032b}'[16:].count('0')
           for ip in net):
        print(str(net.netmask).split('.')[2])   # Ответ: 224
        break
````

## Тип 6 — Неизвестный байт адреса сети

**Идея:** перебирать значения байта от 0 до 255, подставляя в f-строку IP-адреса сети. Для каждого значения строить объект сети и проверять условие на все адреса.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1315 — Тип 6 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом <strong>32.0.А.5</strong> и маской сети <strong>255.255.240.0</strong>.<br>Определите минимальное значение А, для которого для всех IP-адресов сети количество единиц в <strong>левых двух байтах ≤ количества единиц в правых двух байтах</strong>.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 16</div></div></div>

````python
# Задание 1315 — неизвестен 3-й байт адреса сети
from ipaddress import ip_network

for A in range(256):
    net = ip_network(f'32.0.{A}.5/255.255.240.0', False)
    if all(f'{int(ip):032b}'[:16].count('1') <= f'{int(ip):032b}'[16:].count('1')
           for ip in net):
        print(A)   # Ответ: 16
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1316 — Тип 6 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом <strong>127.254.А.10</strong> и маской сети <strong>255.255.224.0</strong>.<br>Определите максимальное значение А, для которого для всех IP-адресов сети количество единиц в <strong>левых двух байтах ≥ количества единиц в правых двух байтах</strong>.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 159</div></div></div>

````python
# Задание 1316 — ищем максимум => перебор от 255 вниз
from ipaddress import ip_network

for A in range(255, -1, -1):
    net = ip_network(f'127.254.{A}.10/255.255.224.0', False)
    if all(f'{int(ip):032b}'[:16].count('1') >= f'{int(ip):032b}'[16:].count('1')
           for ip in net):
        print(A)   # Ответ: 159
        break
````

## Тип 7 — Неизвестный байт адреса узла

**Идея:** перебирать значения байта от 0 до 255, подставляя в IP-адрес **узла**. Для каждого значения строить сеть с `strict=False` и проверять условие.

> **Отличие от типов 5 и 6:** здесь в `ip_network()` передаётся **адрес узла**, а не адрес сети. Параметр `strict=False` обязателен.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1317 — Тип 7 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Сеть задана IP-адресом узла <strong>192.168.А.55</strong> и маской сети <strong>255.255.248.0</strong>.<br>Определите минимальное значение А, для которого для всех IP-адресов сети количество единиц в <strong>левых двух байтах ≤ количества единиц в правых двух байтах</strong>.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 248</div></div></div>

````python
# Задание 1317 — неизвестен 3-й байт адреса УЗЛА
from ipaddress import ip_network

for A in range(256):
    net = ip_network(f'192.168.{A}.55/255.255.248.0', False)
    if all(f'{int(ip):032b}'[:16].count('1') <= f'{int(ip):032b}'[16:].count('1')
           for ip in net):
        print(A)   # Ответ: 248
        break
````
