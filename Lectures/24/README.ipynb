## Теория

Задание 24 — найти в файле (10⁶ символов) максимальную подстроку, удовлетворяющую условию. Ручной анализ исключён.

**Универсальный алгоритм:**

```python
import re

with open('файл.txt') as f:
    data = f.readline()

# Компилируем паттерн и ищем все совпадения:
matches = re.findall(r'ПАТТЕРН', data)
print(len(max(matches, key=len)))
```

**Ключевые конструкции regex:**

| Паттерн | Значение |
|:---|:---|
| `[AE]` | Один из символов A или E |
| `[A-Z]` | Любая заглавная буква |
| `[0-9]` | Цифра |
| `[02468]` | Чётная цифра |
| `[13579]` | Нечётная цифра |
| `+` | Один или более |
| `*` | Ноль или более |
| `(?:...)` | Группировка без захвата |
| `(?=...)` | Lookahead (не захватывает) |
| `\1` | Обратная ссылка на 1-ю группу |

<div class="task-box">
<div class="task-title">📋 Задание Пример 1 — Пары гласная + согласная | Ответ: <strong style="color:#2fca72;">Перебором regex</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Файл состоит из символов: гласные A, E и согласные B, C, D. Найти в файле максимальную непрерывную последовательность, которая является правильной строкой пар «гласная + согласная».<br>
Например, в строке AABECAEABECADA максимальная такая последовательность — ABECABEС (длина 8).<br>
<strong>Определите количество символов в максимальной такой последовательности.</strong></div>
</div>

````python
import re

data = "AABECAEABECADA"  # пример
# Паттерн: одна или более пар (гласная + согласная)
pattern = r'(?:[AE][BCD])+'

matches = re.findall(pattern, data)
answer = max(len(m) for m in matches)
print(f'Символов: {answer}, Пар: {answer // 2}')
````

<div class="task-box">
<div class="task-title">📋 Задание Пример 2 — Максимальное чётное 14-ричное число | Ответ: <strong style="color:#2fca72;">131 (для файла задания 24)</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Текстовый файл состоит из цифр 0–9 и букв A–D. Найти в файле максимальную непрерывную последовательность, которая является корректной записью чётного числа в 14-ричной системе счисления. Число не должно иметь ведущих нулей, и число 0 не имеет знака.<br>
<strong>Определите количество символов в максимальной такой последовательности.</strong></div>
</div>

````python
import re

with open('файл.txt') as f:
    data = f.readline()

# Чётные цифры 14-ричной: 0,2,4,6,8,A(10),C(12)
# Нечётные: 1,3,5,7,9,B(11),D(13)
# Без ведущих нулей: первая цифра — [1-9ABCD], последняя — [02468AC]
pattern = r'[1-9ABCD][0-9ABCD]*[02468AC]|[02468AC]'

matches = re.findall(pattern, data)
print(len(max(matches, key=len)))  # 131
````

<div class="task-box">
<div class="task-title">📋 Задание Пример 3 — Одинаковые буквы между чётными цифрами | Ответ: <strong style="color:#2fca72;">Regex с обратной ссылкой</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Файл состоит из заглавных букв (A–Z) и цифр (0–9). Найти максимальную последовательность, в которой все символы — одна и та же буква, а первый и последний символы — чётные цифры (0,2,4,6,8).<br>
<strong>Определите количество символов в максимальной такой последовательности.</strong></div>
</div>

````python
import re

with open('файл.txt') as f:
    data = f.readline()

# Паттерн: [чётная цифра][одна и та же буква]+[чётная цифра]
pattern = r'[02468]([A-Z])\1+[02468]'

matches = re.findall(pattern, data)
# findall возвращает группы, нужен весь матч — используем finditer:
matches = [m.group() for m in re.finditer(pattern, data)]
print(len(max(matches, key=len)))
````

<div class="task-box">
<div class="task-title">📋 Задание Пример 4 — Арифметическое выражение (задание 24 из условия) | Ответ: <strong style="color:#2fca72;">131</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Текстовый файл состоит из цифр 0, 1, 2, 3, 4 и знаков арифметических операций «−» и «+» (вычитание и сложение). Определите максимальное количество символов в непрерывной последовательности, которая является корректным арифметическим выражением с целыми неотрицательными числами. В этом выражении никакие два знака арифметических операций не стоят рядом, в записи чисел отсутствуют незначащие нули и число 0 не имеет знака. Наличие любого знака перед первым числом (если это не число 0) допускается.<br>
<strong>В ответе укажите количество символов.</strong></div>
</div>

````python
import re

with open('24.txt') as f:
    data = f.readline()

N = r'(?:0|[1-4][0-4]*)'
# Начало: [+-]? перед ненулевым или просто 0
pattern = rf'[+-]?[1-4][0-4]*(?:[+-]{N})*|0(?:[+-]{N})+'

matches = re.findall(pattern, data)
print(len(max(matches, key=len)))  # 131
````
