## О задании

Задание 24 — **высокий** уровень сложности, 2 балла. Даётся текстовый файл объёмом до **10⁶ символов**. Нужно написать программу, которая находит максимальную по длине подстроку, удовлетворяющую условию.

**Ручной анализ исключён** — только программный подход.

**Три основных метода:**

| Метод | Когда использовать |
|:---:|:---|
| `re.findall` / `re.finditer` | Условие описывается регулярным выражением |
| `replace` + `split` | Одно условие — разбивка по запрещённым символам |
| Скользящее окно (два указателя) | Два условия одновременно |

## Метод 1: регулярные выражения

```python
import re

with open('файл.txt') as f:
    data = f.readline()

matches = re.findall(r'ПАТТЕРН', data)
print(len(max(matches, key=len)))
```

### Ключевые конструкции regex

| Паттерн | Значение |
|:---|:---|
| `[AE]` | Один из символов A или E |
| `[A-Z]` | Любая заглавная буква |
| `[0-9]` | Цифра |
| `[02468]` | Чётная цифра |
| `[1-9ABCD]` | Ненулевая цифра 14-ричной СС |
| `[02468AC]` | Чётная цифра 14-ричной СС |
| `+` | Один или более |
| `*` | Ноль или более |
| `(?:...)` | Группировка без захвата |
| `(X)\1+` | Обратная ссылка: повтор той же группы |

## Тип 1 — Паттерн из повторяющихся пар

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Пример 1 — Пары «гласная + согласная»</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Текстовый файл состоит из символов: гласные <strong>A, E</strong> и согласные <strong>B, C, D</strong>.<br>
Найти в файле максимальную непрерывную последовательность, которая является правильной строкой пар «гласная + согласная».<br><br>
Например, в строке <code>AABECAEABECADA</code> максимальная такая последовательность — <code>ABECABEC</code> (длина 8).<br><br>
<strong>Определите количество символов в максимальной такой последовательности.</strong></div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: Длина максимальной последовательности</div></div>

````python
import re

# Тест из условия
data = "AABECAEABECADA"

# Паттерн: одна или более пар (гласная + согласная)
pattern = r'(?:[AE][BCD])+'

matches = re.findall(pattern, data)
answer  = max(len(m) for m in matches)
print(f'Символов: {answer}, Пар: {answer // 2}')

# Для файла:
# with open('файл.txt') as f:
#     data = f.readline()
# matches = re.findall(r'(?:[AE][BCD])+', data)
# print(len(max(matches, key=len)))
````

## Тип 2 — Числа в N-ричной системе счисления

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Пример 2 — Максимальное чётное 14-ричное число</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Текстовый файл состоит из цифр 0–9 и букв A–D.<br>
Найти в файле максимальную непрерывную последовательность, которая является корректной записью <strong>чётного числа в 14-ричной системе счисления</strong>. Число не должно иметь ведущих нулей (кроме числа 0).<br><br>
<strong>Определите количество символов в максимальной такой последовательности.</strong><br><br>
<strong>Разбор:</strong><br>
Чётные цифры 14-ричной СС: 0, 2, 4, 6, 8, A(=10), C(=12)<br>
Нечётные: 1, 3, 5, 7, 9, B(=11), D(=13)<br>
Без ведущих нулей: первая цифра ∈ [1-9ABCD]<br>
Последняя цифра чётная: ∈ [02468AC]</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 131</div></div>

````python
import re

with open('24.txt') as f:
    data = f.readline()

# Чётные цифры 14-ричной: 0,2,4,6,8,A(10),C(12)
# Нечётные:                1,3,5,7,9,B(11),D(13)
# Без ведущих нулей: первая цифра ≠ 0, последняя чётная
pattern = r'[1-9ABCD][0-9ABCD]*[02468AC]|[02468AC]'

matches = re.findall(pattern, data)
print(len(max(matches, key=len)))  # Ответ: 131
````

## Тип 3 — Одинаковые символы между заданными

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Пример 3 — Одинаковые буквы между чётными цифрами</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Файл состоит из заглавных букв (A–Z) и цифр (0–9). Найти максимальную последовательность, в которой первый и последний символы — <strong>чётные цифры</strong> (0,2,4,6,8), а все символы между ними — <strong>одна и та же буква</strong>.<br><br>
<strong>Определите количество символов в максимальной такой последовательности.</strong></div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: Используй обратную ссылку \1</div></div>

````python
import re

with open('файл.txt') as f:
    data = f.readline()

# Паттерн: чётная_цифра + (одна и та же буква)+ + чётная_цифра
# (\1) — обратная ссылка: та же буква, что захвачена в группу 1
pattern = r'[02468]([A-Z])\1+[02468]'

# findall возвращает только группы, нужен весь матч → finditer
matches = [m.group() for m in re.finditer(pattern, data)]
print(len(max(matches, key=len)))
````

## Тип 4 — Два условия одновременно (скользящее окно)

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Пример 4 — Ровно 80 букв Y И не менее 90 вхождений '2025'</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Текстовый файл состоит из символов 0–9 и Y. Найти максимальную подстроку, в которой:<br>
<strong>Условие 1:</strong> ровно 80 букв Y<br>
<strong>Условие 2:</strong> не менее 90 вхождений подстроки '2025'<br><br>
<strong>Метод:</strong> скользящее окно по позициям вхождений 'Y'.<br>
Берём все позиции 'Y', рассматриваем окна из 80 штук подряд и расширяем границы.</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: Используй скользящее окно</div></div>

````python
import re

with open('файл.txt') as f:
    data = f.readline()

# Находим все позиции буквы Y
y_pos = [m.start() for m in re.finditer('Y', data)]

ans = 0

# Скользящее окно: рассматриваем каждое окно из 80 букв Y подряд
for i in range(len(y_pos) - 79):
    left  = y_pos[i]              # начало окна
    right = y_pos[i + 79] + 1    # конец окна (включительно)

    window = data[left:right]

    # Расширяем границы, захватывая весь контекст
    # Подсчёт вхождений '2025' в окне:
    count_2025 = window.count('2025')

    if count_2025 >= 90:
        ans = max(ans, right - left)

print(ans)
````
