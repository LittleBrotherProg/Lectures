## 2. Шпаргалка по синтаксису regexp

| Символ | Название | Простыми словами |
|:---:|:---|:---|
| `[AE]` | Множество | Один символ — A **или** E |
| `[A-Z]` | Диапазон | Любая заглавная буква от A до Z |
| `[^S02468]` | Отрицание | Всё **кроме** S, 0, 2, 4, 6, 8 |
| `*` | Ноль или больше | Предыдущее повторяется 0, 1, 2... раз |
| `+` | Один или больше | Предыдущее повторяется **хотя бы 1** раз |
| `(?:...)` | Группа без захвата | Объединяет блок |
| `(...)` | Группа с захватом | Запоминает найденное в `\1` |
| `\1` | Обратная ссылка | Требует ту же строку что была в группе 1 |
| `{N}` | Ровно N раз | Предыдущее повторяется ровно N раз |
| `{{N}}` | N в f-строке | Экранирование `{N}` внутри `rf'...'` |

```python
import re

with open('файл.txt', 'r') as file:

pattern = r'ВАШ_ШАБЛОН'

matches = re.finditer(pattern, data)
answer = max(len(m.group()) for m in matches)
print(answer)
```

### Пример 1 — Пары «гласная + согласная»

**Условие:** Файл из символов `A, B, C, D, E`. Найдите максимальное количество идущих подряд **пар** вида «гласная + согласная».

**Шаблон:** `(?:[AE][BCD])+`

| Часть | Что означает |
|:---|:---|
| `[AE]` | Одна гласная: A или E |
| `[BCD]` | Одна согласная: B, C или D |
| `(?:...)` | Группируем пару в блок |
| `+` | Пара повторяется 1+ раз |

````python
import re

data = "AABECAEABECADA"
pattern = r'(?:[AE][BCD])+'

matches = re.finditer(pattern, data)
answer = max(len(m.group()) for m in matches)
print(f'Символов: {answer}, Пар: {answer // 2}')
````

### Пример 2 — Максимальное чётное 14-ричное число

**Условие:** Найдите запись максимального чётного числа в 14-ричной СС (цифры 0–9 и A, B, C, D).

> **Чётное в 14-ричной СС** = оканчивается на 0, 2, 4, 6, 8, A или C

**Шаблон:** `[1-9ABCD][0-9ABCD]*[02468AC]`

| Часть | Что означает |
|:---|:---|
| `[1-9ABCD]` | Первый символ — не ноль (нет ведущих нулей) |
| `[0-9ABCD]*` | Средние — любые допустимые, 0 или более |
| `[02468AC]` | Последний — чётная цифра в 14-ричной СС |

````python
import re

with open('файл.txt', 'r') as file:
    data = file.readline()

pattern = r'[1-9ABCD][0-9ABCD]*[02468AC]'

matches = re.finditer(pattern, data)
answer = max(len(m.group()) for m in matches)
print(answer)
````

### Пример 3 — Одинаковые буквы между чётными цифрами

**Условие:** Найдите: чётная цифра + N **одинаковых** букв + чётная цифра.

**Шаблон:** `[02468]([A-Z])\1+[02468]`

| Часть | Что означает |
|:---|:---|
| `[02468]` | Чётная цифра в начале |
| `([A-Z])` | Любая буква — **запоминаем** её в группу 1 |
| `\1+` | Та же буква ещё 1+ раз (обратная ссылка) |
| `[02468]` | Чётная цифра в конце |

````python
import re

with open('файл.txt', 'r') as file:
    data = file.readline()

pattern = r'[02468]([A-Z])\1+[02468]'

matches = re.finditer(pattern, data)
answer = max(len(m.group()) for m in matches)
print(answer)
````

### Шаблон числа

```python
# Ноль ИЛИ число без ведущих нулей
num = r'(?:0|[1-9][0-9]*)'

# Шаблон выражения: Число (операция Число)*
pattern = rf'{num}(?:[-+*]{num})*'
```

### Пример 1 — Базовое выражение

```python
import re
with open('файл.txt') as f:
    data = f.readline()

matches = re.finditer(pattern, data)

```

```python
# Порядок важен: сначала длинные числа, потом однозначные
```

## 5. Тип 3 — Запрещённые соседства

<div style="background:#0d1b2a;border-left:4px solid #2fca72;border-radius:0 8px 8px 0;padding:16px 20px;margin:10px 0;">
<span style="color:#2fca72;font-weight:700;">Признак:</span> <span style="color:#c8d8e8;">Символы A и B не должны стоять рядом. Метод: replace + split.</span>

### Алгоритм

```
1. replace('ab', 'a b')  →  вставить пробел между запрещёнными парами
2. split()               →  разбить строку на фрагменты
3. max(data, key=len)    →  найти самый длинный фрагмент
```

### Пример 1 — Символы a и d не рядом

```python
with open('файл.txt') as f:
    data = f.readline()

data = data.replace('ad', 'a d').replace('da', 'd a').split()
print(max(len(x) for x in data))
```

### Пример 2 — Группа символов Q, R, S не рядом

```python
with open('файл.txt') as f:
    data = f.readline()

# Заменяем все три буквы на точку, потом разбиваем по парам точек
data = data.replace('Q','.').replace('R','.').replace('S','.')
data = data.replace('..', '. .').split()
print(max(len(x) for x in data))
```

## 6. Тип 4 — Заданное количество вхождений

<div style="background:#0d1b2a;border-left:4px solid #f0c040;border-radius:0 8px 8px 0;padding:16px 20px;margin:10px 0;">
<span style="color:#f0c040;font-weight:700;">Признак:</span> <span style="color:#c8d8e8;">Символ / подстрока встречается ровно N раз. Или два условия одновременно.</span>
</div>

### Метод А — replace + split (одно условие)

**Идея:** разбить по вхождению `AB` → получить фрагменты → склеить **N+1 фрагментов** = ровно N вхождений `AB`.

```python
with open('файл.txt') as f:
    data = f.readline()

N = 21  # ровно N вхождений 'AB'
data = data.replace('AB', 'A B').split()

max_len = 0
for i in range(len(data) - N):
    sub = ''.join(data[i:i + N + 1])  # N+1 фрагментов
    max_len = max(max_len, len(sub))

print(max_len)
```

> **Почему N+1?** Между N+1 фрагментами ровно N «разрезов» = N вхождений `AB`.

### Метод Б — скользящее окно (два условия)

**Пример:** ровно 80 букв Y И не менее 90 вхождений '2025'.

```python
import re

with open('файл.txt') as f:
    data = f.readline()

K = 80; letter = 'Y'; need = 90; sub = '2025'

# 1. Индексируем позиции буквы Y
pos = [m.start() for m in re.finditer(letter, data)]

max_len = 0
for i in range(len(pos) - K + 1):
    left  = pos[i]
    right = pos[i + K - 1]

    # 2. Расширяем границы до соседних Y
    l = left
    while l > 0 and data[l - 1] != letter:
        l -= 1
    r = right
    while r + 1 < len(data) and data[r + 1] != letter:
        r += 1

    substr = data[l:r + 1]

    # 3. Проверяем второе условие
    if substr.count(sub) >= need:
        max_len = max(max_len, len(substr))

print(max_len)
```

### Метод В — regexp (два условия через шаблон)

**Пример:** начинается с чётной цифры, содержит ровно 76 букв F, других чётных нет.

```python
import re

with open('файл.txt') as f:
    data = f.readline()

evens       = r'[02468]'
not_F_evens = r'[^F02468]'
group_F     = rf'(?:{not_F_evens}*F)'
pattern     = rf'{evens}{group_F}{{76}}{not_F_evens}*'

matches = re.finditer(pattern, data)
print(max(len(m.group()) for m in matches))
```
