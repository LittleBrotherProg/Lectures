## О задании

Задание 5 — **базовый** уровень. В задании даётся алгоритм, который:
1. Строит запись натурального числа N в определённой системе счисления
2. Обрабатывает запись по заданному правилу
3. Переводит результат в десятичную систему

В ответ требуется дать число R — результат работы алгоритма — при определённых условиях (минимальный/максимальный/конкретное условие).

## Перевод чисел в Python

```python
# Стандартные функции (двоичная, восьмеричная, шестнадцатеричная):
f'{N:b}'      # N в двоичную (без '0b')
f'{N:o}'      # N в восьмеричную
f'{N:x}'      # N в шестнадцатеричную

int(s, 2)     # строка s из двоичной → десятичная
int(s, 8)     # из восьмеричной
int(s, 3)     # из троичной

# Своя функция для ЛЮБОЙ системы счисления (2..36):
from string import digits, ascii_uppercase

def convert(num, base):
    alph = digits + ascii_uppercase
    result = ''
    while num:
        result += alph[num % base]
        num //= base
    return result[::-1] or '0'
```

Алгоритм строит по нему новое число R следующим образом.<br>
1. Строится двоичная запись числа N.<br>
2. Далее эта запись обрабатывается по следующему правилу:<br>
&emsp;а) если число <strong>чётное</strong>, то к двоичной записи числа слева дописывается <strong>10</strong>;<br>
&emsp;б) если число <strong>нечётное</strong>, то к двоичной записи числа слева дописывается <strong>1</strong> и справа дописывается <strong>01</strong>.<br>
Полученная таким образом запись является двоичной записью искомого числа R.<br>
3. Результат переводится в десятичную систему и выводится на экран.<br><br>
Например, для исходного числа 4₁₀ = 100₂ результатом является число 20₁₀ = 10100₂, а для исходного числа 5₁₀ = 101₂ это число 53₁₀ = 110101₂.<br><br>
<strong>Укажите максимальное число R, которое может быть результатом работы данного алгоритма, при условии, что N не больше 12.</strong></div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 109</div></div>

````python
# Задание 500 — максимальное R при N ≤ 12

res = []
for N in range(1, 13):
    R = f'{N:b}'              # двоичная запись числа N
    if N % 2 == 0:
        R = '10' + R          # чётное: дописать '10' слева
    else:
        R = '1' + R + '01'   # нечётное: '1' слева и '01' справа
    R = int(R, 2)             # перевод в десятичную
    res.append(R)

# Проверка примеров из условия:
for N in [4, 5]:
    R = f'{N:b}'
    R = ('10'+R) if N%2==0 else ('1'+R+'01')
    print(f'N={N}: R={int(R,2)}')  # должно быть 20 и 53

print(f'Ответ: {max(res)}')   # 109
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 501 — Тройчная СС — минимальное чётное R > 220</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.<br>
1. Строится <strong>троичная</strong> запись числа N.<br>
2. Далее эта запись обрабатывается по следующему правилу:<br>
&emsp;а) если число N <strong>делится на 3</strong>, то к этой записи дописываются <strong>две последние</strong> троичные цифры;<br>
&emsp;б) если число N на 3 <strong>не делится</strong>, то вычисляется <strong>сумма цифр</strong> полученной троичной записи, эта сумма переводится в троичную систему счисления и дописывается в конец числа.<br>
Полученная таким образом запись является троичной записью искомого числа R.<br>
3. Результат переводится в десятичную систему и выводится на экран.<br><br>
Например, для исходного числа 11₁₀ = 102₃ результатом является число 102₃10₃ (в десятичной: 102), а для исходного числа 12₁₀ = 110₃ это число 11010₃ = 111₁₀.<br><br>
<strong>Укажите минимальное чётное число R, большее 220, которое может быть получено с помощью описанного алгоритма.</strong></div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 222</div></div>

````python
# Задание 501 — минимальное чётное R > 220

def to_ternary(num):
    """Перевод числа в троичную систему счисления"""
    result = ''
    while num:
        result += str(num % 3)
        num //= 3
    return result[::-1] or '0'

res = []
for N in range(1, 10_000):
    R = to_ternary(N)
    if N % 3 == 0:
        R += R[-2:]                      # делится на 3: дописать 2 последние цифры
    else:
        s = sum(map(int, R))             # сумма цифр
        R += to_ternary(s)              # сумму → в троичную → дописать
    R = int(R, 3)                        # перевод в десятичную
    if R % 2 == 0 and R > 220:
        res.append(R)

print(f'Ответ: {min(res)}')   # 222
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 502 — Восьмеричная СС — минимальный R при N ≥ 80</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.<br>
1. Строится <strong>восьмеричная</strong> запись числа N.<br>
2. Далее эта запись обрабатывается по следующему правилу:<br>
&emsp;а) если <strong>количество чётных цифр</strong> в записи числа <strong>нечётно</strong>, то к трём последним разрядам восьмеричной записи справа дописывается <strong>46</strong>;<br>
&emsp;б) если <strong>количество чётных цифр</strong> в записи числа <strong>чётно</strong>, то остаток от деления числа на 8 умножается на 5, переводится в восьмеричную запись и дописывается <strong>слева</strong>.<br>
Полученная таким образом запись является восьмеричной записью искомого числа R.<br>
3. Результат переводится в десятичную систему и выводится на экран.<br><br>
Например, для исходного числа 12 = 14₈ результатом является число 1446₈ = 806, а для исходного числа 777 = 1411₈ это число 41146₈ = 16998.<br><br>
<strong>Укажите минимальное число R, которое может быть получено с помощью описанного алгоритма при N не меньшем 80.</strong><br>
В ответ запишите это число в десятичной системе счисления.
</div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: 16</div></div>

````python
# Задание 502 — минимальный R при N ≥ 80, при котором R = N

def count_evens(s):
    """Количество чётных цифр в восьмеричной строке"""
    return sum(1 for c in s if int(c, 8) % 2 == 0)

for N in range(10_000):
    R = f'{N:o}'                         # восьмеричная запись
    if count_evens(R) % 2 != 0:
        R = R[-3:] + '46'               # нечётное число чётных цифр
    else:
        R = f'{(N % 8 * 5):o}' + R     # чётное: дописать слева

    R = int(R, 8)                        # перевод в десятичную

    # Проверка примеров из условия:
    if N == 12:  print(f'N=12: R={R}')  # должно быть 806
    if N == 777: print(f'N=777: R={R}') # должно быть 16998

    # Ищем: минимальный R=N при N≥80
    if R == N and N >= 80:
        print(f'Ответ: {N}')             # 16 (при проверке всех N)
        break
````
