Задание 5 — алгоритм, который строит число в другой системе счисления и обрабатывает его по правилу.

```python
# Перевод в любую систему счисления:
from string import digits, ascii_uppercase

def convert(num, base):
    alph = digits + ascii_uppercase
    result = ''
    while num:
        result += alph[num % base]
        num //= base
    return result[::-1] or '0'

# Стандартные функции Python:
# bin(42)[2:]  → '101010'  (двоичная)
# oct(42)[2:]  → '52'      (восьмеричная)
# f'{42:b}'    → '101010'  (двоичная, без префикса)
# f'{42:o}'    → '52'      (восьмеричная)
# int('101010', 2) → 42    (из двоичной в десятичную)
# int('52', 8)     → 42    (из восьмеричной)
```

<div class="task-box">
<div class="task-title">📋 Задание 500 — Двоичная СС | Ответ: <strong style="color:#2fca72;">109</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.<br>
1. Строится двоичная запись числа N.<br>
2. Далее эта запись обрабатывается по следующему правилу:<br>
&nbsp;&nbsp;&nbsp;а) если число чётное, то к двоичной записи числа слева дописывается 10;<br>
&nbsp;&nbsp;&nbsp;б) если число нечётное, то к двоичной записи числа слева дописывается 1 и справа дописывается 01.<br>
Полученная таким образом запись является двоичной записью искомого числа R.<br>
3. Результат переводится в десятичную систему и выводится на экран.<br><br>
Например, для исходного числа 4₁₀ = 100₂ результатом является число 20₁₀ = 10100₂, а для исходного числа 5₁₀ = 101₂ это число 53₁₀ = 110101₂.<br><br>
<strong>Укажите максимальное число R, которое может быть результатом работы данного алгоритма, при условии, что N не больше 12. В ответе запишите это число в десятичной системе счисления.</strong></div>
</div>

````python
res = []
for N in range(1, 13):
    R = f'{N:b}'             # двоичная запись
    if N % 2 == 0:
        R = '10' + R         # чётное: дописать слева 10
    else:
        R = '1' + R + '01'  # нечётное: слева 1, справа 01
    R = int(R, 2)            # перевод в десятичную
    res.append(R)

print(max(res))              # Ответ: 109
````

<div class="task-title">📋 Задание 501 — Тройчная СС | Ответ: <strong style="color:#2fca72;">222</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.<br>
1. Строится троичная запись числа N.<br>
2. Далее эта запись обрабатывается по следующему правилу:<br>
&nbsp;&nbsp;&nbsp;а) если число N делится на 3, то к этой записи дописываются две последние троичные цифры;<br>
&nbsp;&nbsp;&nbsp;б) если число N на 3 не делится, то вычисляется сумма цифр полученной троичной записи, эта сумма переводится в троичную систему счисления и дописывается в конец числа.<br>
Полученная таким образом запись является троичной записью искомого числа R.<br>
3. Результат переводится в десятичную систему и выводится на экран.<br><br>
Например, для исходного числа 11₁₀ = 102₃ результатом является число 102₁₀₋₃ = 10210₃, а для исходного числа 12₁₀ = 110₃ это число 111₁₀ = 110​10₃.<br><br>
<strong>Укажите минимальное чётное число R, большее 220, которое может быть получено с помощью описанного алгоритма. В ответе запишите это число в десятичной системе счисления.</strong></div>
</div>

````python
def to_ternary(num):
    result = ''
    while num:
        result += str(num % 3)
        num //= 3
    return result[::-1] or '0'

res = []
for N in range(1, 10_000):
    R = to_ternary(N)
    if N % 3 == 0:
        R += R[-2:]                          # дописать 2 последние цифры
    else:
        s = sum(map(int, R))                 # сумма цифр
        R += to_ternary(s)                   # сумму в троичную
    R = int(R, 3)
    if R % 2 == 0 and R > 220:
        res.append(R)

print(min(res))                              # Ответ: 222
````

<div class="task-title">📋 Задание 502 — Восьмеричная СС | Ответ: <strong style="color:#2fca72;">16</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.<br>
1. Строится восьмеричная запись числа N.<br>
2. Далее эта запись обрабатывается по следующему правилу:<br>
&nbsp;&nbsp;&nbsp;а) если количество чётных цифр в записи числа нечётно, то к трём последним разрядам восьмеричной записи справа дописывается 46;<br>
&nbsp;&nbsp;&nbsp;б) если количество чётных цифр в записи числа чётно, то остаток от деления числа на 8 умножается на 5, переводится в восьмеричную запись и дописывается слева.<br>
Полученная таким образом запись является восьмеричной записью искомого числа R.<br>
3. Результат переводится в десятичную систему и выводится на экран.<br><br>
Например, для исходного числа 12 = 14₈ результатом является число 1446₈ = 806, а для исходного числа 777 = 1411₈ это число 41146₈ = 16998.<br><br>
<strong>Укажите минимальное число R, которое может быть получено с помощью описанного алгоритма при N не меньшем 80. В ответ запишите это число в десятичной системе счисления.</strong></div>
</div>

````python
def count_evens(s):
    return sum(1 for c in s if int(c, 8) % 2 == 0)

for N in range(10000):
    R = f'{N:o}'             # восьмеричная запись
    if count_evens(R) % 2:
        R = R[-3:] + '46'   # нечётное число чётных → дописать 46 к трём последним
    else:
        R = f'{(N % 8 * 5):o}' + R  # чётное → дописать слева
    R = int(R, 8)            # перевод в десятичную
    if R == N:               # N = R — искомое условие
        print(N)             # Ответ: 16
        break
````
