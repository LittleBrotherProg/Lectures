## 2. Ручной метод перевода



<div class="tip-box">
<strong style="color:#2fca72;">Алгоритм:</strong><br>
1. Делим число на основание N<br>
2. Записываем остатки<br>
3. Читаем остатки <strong>снизу вверх</strong><br>
4. Остатки > 9 заменяем на буквы: 10=A, 11=B, ..., 35=Z
</div>

**Пример: 78₁₀ → восьмеричная:**
```
78 ÷ 8 = 9  остаток 6
 9 ÷ 8 = 1  остаток 1
 1 ÷ 8 = 0  остаток 1
Читаем снизу вверх: 116
78₁₀ = 116₈
```

## 3. Python — встроенные функции



<div class="warn-box">
<strong style="color:#f0c040;">Важно:</strong> bin(), oct(), hex() возвращают строку с префиксом (0b, 0o, 0x).<br>
Убрать префикс: <code>bin(42)[2:]</code> или f-строка: <code>f'{42:b}'</code> — без префикса сразу.
</div>

````python
# Перевод числа 42 в различные СС
N = 42
print(f'{N:b}')    # '101010'  двоичная
print(f'{N:o}')    # '52'      восьмеричная
print(f'{N:x}')    # '2a'      16-ричная
print(f'{N:X}')    # '2A'      заглавные

# Перевод строки в десятичное
print(int('101010', 2))   # 42
print(int('ff', 16))      # 255
print(int('116', 8))      # 78

# Полезные операции со строками
R = '101010'
print(sum(map(int, R)))   # 3 — сумма цифр
print(R[-2:])              # '10' — последние 2
print('11' + R)            # дописать слева
print(R + '01')            # дописать справа
````

## 4. Функция convert() для любой СС (2..36)

````python
from string import digits, ascii_uppercase

# Алфавит: '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
# Цифры > 9 обозначаются буквами: 10=A, 11=B, ..., 35=Z

def convert(num, base):
    alph = digits + ascii_uppercase
    result = ''
    while num:
        result += alph[num % base]  # остаток → символ алфавита
        num //= base                # делим на основание
    return result[::-1]             # разворачиваем (снизу вверх)

print(convert(42, 2))    # '101010'
print(convert(78, 8))    # '116'
print(convert(255, 16))  # 'FF'

# Упрощённая версия для тройчной СС
def to_ternary(num):
    result = ''
    while num:
        result += str(num % 3)
        num //= 3
    return result[::-1]

print(to_ternary(11))    # '102'
print(to_ternary(12))    # '110'
````

## 6. Пример 1 — Задание 500 (двоичная СС)

<div class="info-box">
<strong style="color:#1b91ff;">Условие:</strong><br>
N → двоичная запись.<br>
Чётное N → дописать <code>'10'</code> слева.<br>
Нечётное N → дописать <code>'1'</code> слева и <code>'01'</code> справа.<br>
Найти <strong>максимальное R</strong> при N ≤ 12. Ответ: <strong>109</strong>
</div>

````python
# Шаг 1: Проверяем алгоритм на примерах (4→20, 5→53)
for N in range(1, 13):
    R = f'{N:b}'           # двоичная запись N
    if N % 2 == 0:         # чётное
        R = '10' + R       # дописать '10' слева
    else:                  # нечётное
        R = '1' + R + '01' # дописать '1' слева и '01' справа
    R = int(R, 2)          # перевод в десятичную
    if N in (4, 5):
        print(N, R)        # 4 20 ✓  5 53 ✓

# Шаг 2: Формируем ответ
res = []
for N in range(1, 13):
    R = f'{N:b}'
    if N % 2 == 0:
        R = '10' + R
    else:
        R = '1' + R + '01'
    R = int(R, 2)
    res.append(R)

print(max(res))            # Ответ: 109
````

## 7. Пример 2 — Задание 501 (тройчная СС)

<div class="info-box">
<strong style="color:#1b91ff;">Условие:</strong><br>
N → тройчная запись.<br>
N делится на 3 → дописать 2 последние цифры.<br>
N не делится на 3 → сумму цифр перевести в тройчную и дописать.<br>
Найти <strong>минимальное чётное R > 220</strong>. Ответ: <strong>222</strong>
</div>

````python
def to_ternary(num):
    result = ''
    while num:
        result += str(num % 3)
        num //= 3
    return result[::-1]


res = []
for N in range(1, 10_000):
    R = to_ternary(N)
    if N % 3 == 0:
        R += R[-2:]                  # 2 последние цифры
    else:
        sum_d = sum(map(int, R))     # сумма цифр тройчной записи
        R += to_ternary(sum_d)       # сумму перевести в тройчную
    R = int(R, 3)                    # перевод в десятичную
    # Проверка на примерах: N=11→102, N=12→111
    if N in (11, 12):
        print(N, R)                  # 11 102 ✓  12 111 ✓
    if R % 2 == 0 and R > 220:
        res.append(R)

print(min(res))                      # Ответ: 222
````

## 8. Пример 3 — Задание 502 (восьмеричная СС)

<div class="info-box">
<strong style="color:#1b91ff;">Условие:</strong><br>
N → восьмеричная запись.<br>
Количество чётных цифр нечётно → взять 3 последних цифры, дописать <code>'46'</code> справа.<br>
Количество чётных цифр чётно → N%8 × 5 в восьмеричную, дописать слева.<br>
Найти <strong>минимальное R при N ≥ 80</strong>. Ответ: <strong>16</strong>
</div>

````python
# Функция для подсчёта чётных цифр в восьмеричной записи
def count_evens(num):
    return sum(1 for i in num if int(i, 8) % 2 == 0)


for N in range(80, 10_000):     # N не меньше 80
    R = f'{N:o}'                # восьмеричная запись
    if count_evens(R) % 2:      # кол-во чётных нечётно
        R = R[-3:] + '46'
    else:                       # кол-во чётных чётно
        R = f'{(N % 8 * 5):o}' + R
    R = int(R, 8)               # перевод в десятичную
    # Проверка: N=12→806, N=777→16998
    if N in (12, 777):
        print(N, R)             # 12 806 ✓  777 16998 ✓
    if R == N:                  # условие: R равно N
        print(R)                # Ответ: 16
        break
````
