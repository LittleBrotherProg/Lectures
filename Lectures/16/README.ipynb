## О задании

Задание 16 ЕГЭ по информатике проверяет умение работать с **рекурсивными алгоритмами**. Дана функция, заданная соотношениями — нужно вычислить её значение или значение выражения с ней.

> **Важно:** задание **повышенного** уровня сложности. Функция всегда **рекурсивная** — следующее значение опирается на предыдущие.

Доступны **4 метода решения**:

| Метод | Когда применять | Особенность |
|:---:|:---|:---|
| **Ручной** | Небольшое число вызовов, можно упростить | Самый быстрый на экзамене |
| **Итеративный** | Когда рекурсию легко переписать в цикл | Нет ограничений на глубину |
| **setrecursionlimit** | Глубина рекурсии известна заранее | Увеличить лимит чуть больше максимума |
| **@lru_cache** | Большая глубина, много повторных вызовов | Самый надёжный программный метод |

## Рекурсия

**Рекурсия** — процесс, при котором функция вызывает **саму себя**, разбивая задачу на более мелкие подзадачи того же типа.

Любая рекурсивная функция состоит из двух частей:

- **Базовый случай** — условие остановки. Без него рекурсия бесконечна
- **Рекурсивный случай** — шаг, на котором функция вызывает себя, приближаясь к базовому

**Пример — факториал:**

```python
def factorial(n):
    if n == 1:        # базовый случай
        return 1
    return n * factorial(n - 1)   # рекурсивный случай
```

### Стек вызовов

Каждый вызов функции помещает **контекст** (локальные переменные, аргументы) в **стек вызовов**. Стек работает по принципу **LIFO** — последним вошёл, первым вышел.

```
factorial(5)
  └── factorial(4)
        └── factorial(3)
              └── factorial(2)
                    └── factorial(1) → 1  ← базовый случай
```

Результаты возвращаются в обратном порядке: `1 → 2 → 6 → 24 → 120`.

### Ограничение глубины в Python

По умолчанию Python допускает не более **1000 вызовов**:

```python
import sys
print(sys.getrecursionlimit())   # → 1000
```

При превышении: `RecursionError: maximum recursion depth exceeded`

## Мемоизация

**Мемоизация** — техника оптимизации: результаты вычислений сохраняются в **кеш** (словарь), и при повторном вызове с теми же аргументами берутся оттуда, а не вычисляются снова.

**Без мемоизации** — `fib(35)` вычисляет `fib(1)` миллионы раз → ~3 секунды.  
**С мемоизацией** — каждое значение вычисляется **один раз** → <1 мс.

### Ручная мемоизация (для понимания)

```python
cache = {}

def fib(n):
    if n in cache:          # значение уже есть → сразу вернуть
        return cache[n]
    if n < 2:
        return n
    result = fib(n-1) + fib(n-2)
    cache[n] = result       # сохранить в кеш
    return result
```

## Декоратор `@lru_cache`

**LRU (Least Recently Used)** — стратегия кеша: при переполнении удаляется элемент, **не использовавшийся дольше всего**.

Декоратор из стандартного модуля `functools` делает мемоизацию в одну строку:

```python
from functools import lru_cache

@lru_cache()          # или @lru_cache(maxsize=None) — неограниченный кеш
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
```

**Параметры:**
- `maxsize=128` — макс. количество кешируемых результатов (рекомендуется степень 2 или `None`)
- `typed=False` — если `True`, `f(1)` и `f(1.0)` кешируются отдельно

**Дополнительные методы** после применения декоратора:

```python
fib.cache_info()    # → CacheInfo(hits=..., misses=..., maxsize=128, currsize=...)
fib.cache_clear()   # очистить кеш
```

## Метод 1 — Ручное решение

Подходит, когда число рекурсивных вызовов **невелико** или выражение можно **упростить алгебраически** — сократить дробь, вынести общий множитель.

**Алгоритм:**
1. Развернуть рекурсию вручную на 2–3 шага
2. Подставить в выражение
3. Упростить (сократить на общий множитель)


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1600 — Ручной метод (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">F(n) = n, при n &gt; 2024<br>F(n) = n × F(n+1), если n ≤ 2024<br><br>Чему равно значение выражения <strong>F(2022) / F(2024)</strong>?<br><br><strong>Разбор:</strong><br>F(2022) = 2022 × F(2023) = 2022 × 2023 × F(2024)<br><br>F(2022) / F(2024) = <strong>2022 × 2023 × F(2024) / F(2024)</strong> = 2022 × 2023<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 4090506</div></div></div>

````python
# Задание 1600 — проверка ручного решения
print(2022 * 2023)   # Ответ: 4090506
````

## Метод 2 — Итеративный подход

Рекурсию заменяем циклом: сохраняем каждое значение функции в список и вычисляем последовательно от базового случая до нужного.

**Нет ограничений** на глубину стека — подходит для любых значений аргумента.

```python
def f_iter(n):
    cache = [...]       # начальные значения (базовые случаи)
    for i in range(базовый+1, n+1):
        cache.append(формула_через_cache[i-k])
    return cache[n]
```


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1601 — Итеративный метод (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">F(n) = 1, при n ≤ 3<br>F(n) = (n+3) × F(n−2), если n &gt; 3<br><br>Чему равно значение выражения <strong>F(2028) / F(2024)</strong>?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 4120899</div></div></div>

````python
# Задание 1601 — итеративный подход
def f_iter(n):
    if n <= 3:
        return 1
    cache = [1, 1, 1, 1]   # значения F(0)..F(3)
    for i in range(4, n + 1):
        cache.append((i + 3) * cache[i - 2])
    return cache[n]

print(f_iter(2028) // f_iter(2024))   # Ответ: 4120899
````

## Метод 3 — Увеличение глубины рекурсии

Используем `sys.setrecursionlimit(N)`, где **N чуть больше максимального аргумента** функции.

```python
import sys
sys.setrecursionlimit(максимальный_аргумент + 5)
```

> ⚠️ При очень большой глубине (тысячи вызовов) возможно падение из-за переполнения стека ОС. Для таких случаев лучше `@lru_cache`.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1602 — setrecursionlimit (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">F(n) = 1, при n = 1<br>F(n) = (n−1) × F(n−1), если n &gt; 1<br><br>Чему равно значение выражения <strong>(F(2024) + 2×F(2023)) / F(2022)</strong>?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 4094550</div></div></div>

````python
# Задание 1602 — увеличение глубины рекурсии
import sys
sys.setrecursionlimit(2025)   # чуть больше максимального аргумента

def f(n):
    if n == 1:
        return 1
    return (n - 1) * f(n - 1)

print((f(2024) + 2 * f(2023)) // f(2022))   # Ответ: 4094550
````

## Метод 4 — Мемоизация через `@lru_cache` ⭐

**Самый надёжный** программный метод. Работает даже при глубине в десятки тысяч вызовов.

**Шаблон:**

```python
from functools import lru_cache

@lru_cache()
def f(n):
    if n < БАЗА:          # базовый случай
        return ...
    return формула_с_f(n-k)

# Прогрев кеша — от базового случая до максимального аргумента
for n in range(БАЗА, МАКС + 1):
    f(n)

# Вычисление ответа
print(выражение_с_f())
```

**Почему нужен прогрев?** Без него Python всё равно попытается построить стек рекурсии «сверху вниз» и упрётся в лимит глубины. Прогрев заполняет кеш «снизу вверх», обходя стек.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1603 — @lru_cache (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">F(n) = n, если n &lt; 5<br>F(n) = 2n × F(n−4), если n ≥ 5<br><br>Чему равно значение выражения <strong>(F(13766) − 9×F(13762)) / F(13758)</strong>?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 757543052</div></div></div>

````python
# Задание 1603 — @lru_cache с прогревом кеша
from functools import lru_cache

@lru_cache()
def f(n):
    if n < 5:
        return n
    return 2 * n * f(n - 4)

# Прогрев: от базового случая (5) до максимального аргумента (13766)
for n in range(5, 13767):
    f(n)

print((f(13766) - 9 * f(13762)) // f(13758))   # Ответ: 757543052
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1604 — @lru_cache (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">F(n) = 1, при n = 1<br>F(n) = 2 × n × F(n−1), если n &gt; 1<br><br>Чему равно значение выражения <strong>(F(2024)/16 − F(2023)) / F(2022)</strong>?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 1019592</div></div></div>

````python
# Задание 1604 — @lru_cache
from functools import lru_cache

@lru_cache()
def f(n):
    if n == 1:
        return 1
    return 2 * n * f(n - 1)

for n in range(1, 2026):
    f(n)

print((f(2024) // 16 - f(2023)) // f(2022))   # Ответ: 1019592
````

## Шпаргалка

### Выбор метода

| Ситуация | Метод |
|:---|:---:|
| Небольшое число вызовов / можно упростить | Ручной |
| Рекурсия легко переписывается в цикл | Итеративный |
| Глубина рекурсии ≤ нескольких тысяч | setrecursionlimit |
| Глубина > 5000 или много повторных вызовов | **@lru_cache** ⭐ |

### Шаблон @lru_cache (универсальный)

```python
from functools import lru_cache

@lru_cache()
def f(n):
    if n <= БАЗА:           # базовый случай — точное условие из задания
        return БАЗА_ЗНАЧЕНИЕ
    return ФОРМУЛА          # рекурсивная формула из задания

for n in range(БАЗА + 1, МАКС + 1):   # прогрев кеша
    f(n)

print(ВЫРАЖЕНИЕ)            # вычисление ответа
```

### Типичные ошибки

```python
# ❌ Неправильно — целочисленное деление нужно там, где в условии стоит /
print(f(2024) / f(2022))     # float, может быть неточным

# ✅ Правильно — использовать // для целых чисел
print(f(2024) // f(2022))
```
