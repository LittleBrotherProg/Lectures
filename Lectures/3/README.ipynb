## Структура базы данных

База данных состоит из **трёх таблиц**, связанных ключами:

| Таблица | Ключевое поле | Что содержит |
|:---|:---:|:---|
| **Движение товаров** | `Артикул` + `ID магазина` | Дата, тип операции (Поступление/Продажа), кол-во |
| **Товар** | `Артикул` | Наименование, ед. измерения, кол-во в упаковке, цена |
| **Магазин** | `ID магазина` | Район, улица |

> **Ключевое правило:** данные между таблицами передаются ТОЛЬКО через ключи — `Артикул` и `ID магазина`. Никаких других способов!

## Функция ВПР()

ВПР() — «вертикальный поиск» — переносит данные из одной таблицы в другую по ключу.

```
=ВПР( Искомое_значение ; Таблица ; Номер_столбца ; 0 )
```

- **Искомое_значение** — ключ из текущей строки (Артикул или ID магазина)
- **Таблица** — диапазон поиска с `$` (нажать **F4** после выделения!)
- **Номер_столбца** — номер нужного столбца в таблице
- **0** — точный поиск (всегда 0!)

**Примеры:**
```
# Название товара по артикулу (3-й столбец таблицы Товар):
=ВПР(D2; Товар.$A$2:$F$61; 3; 0)

# Район магазина по ID (2-й столбец таблицы Магазин):
=ВПР(C2; Магазин.$A$2:$C$19; 2; 0)

# Улица магазина (3-й столбец):
=ВПР(C2; Магазин.$A$2:$C$19; 3; 0)

# Цена товара (6-й столбец):
=ВПР(D2; Товар.$A$2:$F$61; 6; 0)
```

## Функция АГРЕГАТ()

Автофильтр **скрывает** строки, но не удаляет — обычные `СУММ()`, `СЧЁТ()` всё равно их считают!

```
=АГРЕГАТ( Номер_функции ; Параметры ; Диапазон )
```

- `9` — функция СУММ
- `1` — игнорировать скрытые строки
- `Диапазон` — столбец с числами

**Пример:**
```
=АГРЕГАТ(9; 1; J2:J5000)
```
