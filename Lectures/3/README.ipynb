## 1. Реляционные базы данных



<div class="warn-box">
<strong style="color:#f0c040;">Ключевой момент:</strong><br>
Связь между таблицами осуществляется ТОЛЬКО через ключи:<br>
• <code>Артикул</code> → таблица <strong>Товар</strong><br>
• <code>ID магазина</code> → таблица <strong>Магазин</strong><br><br>
Перемещать данные о цене, районе и т.д. можно ТОЛЬКО по этим ключам!
</div>

| Термин | Определение |
|:---|:---|
| Таблица | Набор записей (строк) с одинаковыми полями (столбцами) |
| Запись | Строка таблицы — одна сущность (операция, товар, магазин) |
| Ключ (ID) | Уникальный идентификатор записи — основа связей |
| ER-диаграмма | Схема связей между таблицами БД |

## 2. Функция ВПР()



<div class="tip-box">
<strong style="color:#2fca72;">Совет:</strong> после выделения диапазона нажми <strong>F4</strong> — он закрепится знаками $.<br>
Иначе при протягивании формулы вниз диапазон будет смещаться!
</div>

```
=ВПР( Искомое_значение ; Таблица ; Номер_столбца ; 0 )
         ↑ ключ из         ↑ $A$2:$F$61    ↑ 3=название   ↑ точный поиск
         текущей строки    (нажми F4!)      6=цена
```

**Примеры:**
```
# Название товара:
=ВПР(D2; Товар.$A$2:$F$61; 3; 0)

# Район магазина:
=ВПР(C2; Магазин.$A$2:$C$19; 2; 0)

# Цена товара:
=ВПР(D2; Товар.$A$2:$F$61; 6; 0)

# Стоимость строки:
=E2 * I2    (Количество × Цена)
```

## 3. Функции LibreOffice Calc



<div class="warn-box">
<strong style="color:#f0c040;">КРИТИЧЕСКИ ВАЖНО:</strong><br>
Автофильтр только <strong>скрывает</strong> строки, не удаляет их!<br>
Функции <code>СУММ()</code>, <code>СЧЁТ()</code>, <code>СРЗНАЧ()</code> учитывают скрытые строки.<br>
Для подсчёта только видимых строк ВСЕГДА используй <code>АГРЕГАТ(9;1;...)</code>!
</div>

```
# АГРЕГАТ — сумма только видимых строк:
=АГРЕГАТ(9; 1; J2:J5000)    # 9=СУММ, 1=игнорировать скрытые

# ЕСЛИ — разделить поступление и продажи:
=ЕСЛИ(F2="Поступление"; E2; 0)   # столбец Поступило
=ЕСЛИ(F2="Продажа"; E2; 0)       # столбец Продано

# СЧЁТЕСЛИ:
=СЧЁТЕСЛИ(D3:D10; 5)             # сколько раз встречается 5

# СУММЕСЛИ:
=СУММЕСЛИ($E$3:$E$10; A2; $D$3:$D$10)  # сумма оценок по имени
```

## 4. Алгоритм решения



<div class="info-box">
<strong style="color:#1b91ff;">Рекомендуем 2-й путь:</strong> переносить данные через ВПР() и работать с понятными названиями (товар/район), а не с абстрактными артикулами/ID — надёжнее и меньше ошибок!
</div>

```
Шаг 1. Открыть файл, ознакомиться с тремя таблицами
Шаг 2. Добавить в «Движение товаров» столбцы:
        G: Товар     =ВПР(D2; Товар.$A$2:$F$61; 3; 0)
        H: Район     =ВПР(C2; Магазин.$A$2:$C$19; 2; 0)
        I: Цена      =ВПР(D2; Товар.$A$2:$F$61; 6; 0)
        J: Стоимость =E2*I2
Шаг 3. Автофильтры: Дата | Тип операции | Товар | Район
Шаг 4. =АГРЕГАТ(9; 1; J:J)  — сумма только видимых строк
Шаг 5 (Тип 2): добавить Поступило(I) и Продано(J) через ЕСЛИ()
               Ответ = АГРЕГАТ(I) − АГРЕГАТ(J)
```

## 5. Тип 1 — Итоговая сумма/вес/объём

<div class="info-box">
<strong style="color:#1b91ff;">Формулировка:</strong> «определите, на какую сумму...», «общий вес...», «общий объём...»<br>
<strong>Ответ</strong> = АГРЕГАТ(9; 1; столбец_Стоимости)
</div>

### Задание 313 — ответ: **133228**
Продажи варенца термостатного в Нагорном районе, 5–14 октября.

```
# ВПР: Товар(G), Район(H), Цена(I)
# Стоимость = E2 * I2  → столбец J
# Фильтры:
#   Дата: 07, 09, 10, 14 октября
#   Тип: Продажа
#   Товар: Варенец термостатный
#   Район: Нагорный
# =АГРЕГАТ(9;1; J1092:J3720) → 133228
```

### Задание 303 — ответ: **1501**
Объём (л) всех шампуней на Тургеневской улице, 3–22 сентября.

```
# Добавить: Товар, Улица, Ед.изм., Кол-во_в_упак
# Фильтры: Продажа | Тургеневская | 3 вида шампуня
# Объём(л) = Кол-во_в_упак * Кол-во_шт / 1000
# =АГРЕГАТ(9;1; K:K) → 1501.8 → целая часть: 1501
```

## 6. Тип 2 — Изменение количества

<div class="info-box">
<strong style="color:#1b91ff;">Формулировка:</strong> «на сколько увеличилось количество...», «определите прибыль...»<br>
<strong>Ответ</strong> = АГРЕГАТ(Поступило) − АГРЕГАТ(Продано)
</div>

### Задание 305 — ответ: **1440**
Увеличение запаса клюквы в сахаре, Прибрежный район, 8–17 августа.

```
# ВПР: Товар(G), Район(H)
# Фильтры: даты | Клюква в сахаре | Прибрежный
# Поступило(I) = ЕСЛИ(F="Поступление"; E; 0)
# Продано(J)   = ЕСЛИ(F="Продажа";    E; 0)
# АГРЕГАТ(поступило) = 2100
# АГРЕГАТ(продано)   = 660
# Ответ = 2100 - 660 = 1440
```

### Задание 307 — ответ: **6146**
Увеличение запаса йогуртов (3 вида) 1.5%, Нагорный, 1–15 октября.

```
# Те же шаги. Фильтр: 3 вида йогурта | Нагорный
# АГРЕГАТ(поступило) = 11400
# АГРЕГАТ(продано)   = 5254
# Ответ = 11400 - 5254 = 6146
```
