## 2. Функция ВПР()

```
=ВПР( Искомое_значение ; Таблица ; Номер_столбца ; 0 )
```

| Аргумент | Что указывать |
|:---|:---|
| Искомое_значение | Ключ из текущей строки (Артикул или ID магазина) |
| Таблица | Весь диапазон таблицы с **$** (нажми F4!) |
| Номер_столбца | Номер нужного столбца (3=название, 6=цена) |
| 0 | Точный поиск — всегда 0! |

```
# Название товара по артикулу:
=ВПР(D2; Товар.$A$2:$F$61; 3; 0)

# Район магазина по ID:
=ВПР(C2; Магазин.$A$2:$C$19; 2; 0)

# Цена товара:
=ВПР(D2; Товар.$A$2:$F$61; 6; 0)
```

> **⚠️ Важно:** фильтр скрывает строки, но `СУММ()` всё равно их считает! Используй `АГРЕГАТ(9;1;диапазон)`
