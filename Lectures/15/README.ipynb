## Элементы математической логики

### Основные операции

| Операция | Обозначение | Python | Описание |
|:---:|:---:|:---:|:---|
| Конъюнкция (И) | `A ∧ B` | `A and B` | истина, если оба истинны |
| Дизъюнкция (ИЛИ) | `A ∨ B` | `A or B` | истина, если хотя бы один истинен |
| Инверсия (НЕ) | `¬A` | `not A` | меняет значение на противоположное |
| Импликация (ЕСЛИ…ТО) | `A → B` | `A <= B` | ложна только при A=1, B=0 |
| Эквивалентность | `A ↔ B` | `A == B` | истина, если оба одинаковы |

> **Запись импликации в Python:** `A → B` эквивалентно `(not A) or B`, что также записывается как `A <= B` (поскольку значения `True`/`False` равны `1`/`0`).

### Законы алгебры логики

| Закон | Формула |
|:---|:---|
| Двойное отрицание | `¬(¬A) = A` |
| Де Морган | `¬(A ∧ B) = ¬A ∨ ¬B` и `¬(A ∨ B) = ¬A ∧ ¬B` |
| Дистрибутивность | `A ∧ (B ∨ C) = (A ∧ B) ∨ (A ∧ C)` |
| Идемпотентность | `A ∧ A = A`, `A ∨ A = A` |
| Дополнение | `A ∧ ¬A = 0`, `A ∨ ¬A = 1` |
| Поглощение | `A ∨ (A ∧ B) = A`, `A ∧ (A ∨ B) = A` |

## Поразрядная конъюнкция (`&`)

**Поразрядная конъюнкция** — операция над двоичными представлениями чисел: конъюнкция применяется к каждой паре соответствующих битов.

**Алгоритм:**
1. Перевести оба числа в двоичную систему
2. Выровнять по правому краю (добавить ведущие нули)
3. Применить конъюнкцию поразрядно
4. Перевести результат обратно в десятичную

**Пример:** `12 & 10`
```
12 = 1100₂
10 = 1010₂
     ────
   = 1000₂ = 8
```

В Python оператор `&` выполняет это автоматически:

```python
print(12 & 10)   # → 8
print(14 & 5)    # → 4  (1110 & 0101 = 0100)
```

## Тип 1 — Два неизвестных `x`, `y`

Логическое выражение зависит от двух переменных `x` и `y`. Нужно найти `A`, при котором выражение тождественно истинно при **любых** `x` и `y` из заданного диапазона.

Используем `all()` с вложенным генератором по обеим переменным.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1500 — Тип 1 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для какого <strong>наименьшего</strong> целого неотрицательного числа A выражение<br><br><strong>(x − 3y &lt; A) ∨ (y &gt; 400) ∨ (x &gt; 56)</strong><br><br>тождественно истинно при любых целых положительных x и y?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 54</div></div></div>

````python
# Задание 1500 — ищем наименьшее A, перебор от 0 вверх
def f(x, y, A):
    return (x - 3 * y < A) or (y > 400) or (x > 56)

for A in range(0, 100):
    if all(f(x, y, A)
           for x in range(1, 100)
           for y in range(1, 100)):
        print(A)   # Ответ: 54
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1501 — Тип 1 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для какого <strong>наибольшего</strong> целого неотрицательного числа A выражение<br><br><strong>(x + y ≤ 30) ∨ (y ≤ x + 2) ∨ (y ≥ A)</strong><br><br>тождественно истинно при любых целых положительных x и y?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 17</div></div></div>

````python
# Задание 1501 — ищем наибольшее A => [::-1]
def f(x, y, A):
    return (x + y <= 30) or (y <= x + 2) or (y >= A)

for A in range(0, 50)[::-1]:
    if all(f(x, y, A)
           for x in range(1, 50)
           for y in range(1, 50)):
        print(A)   # Ответ: 17
        break
````

## Тип 2 — Функция ДЕЛ(n, m)

В условии появляется функция **ДЕЛ(n, m)** — «натуральное число n делится без остатка на m».

```python
ДЕЛ(n, m)   →   n % m == 0
¬ДЕЛ(n, m)  →   n % m != 0
```

Импликацию `ДЕЛ(a, b) → ДЕЛ(a, c)` записываем как `(a % b == 0) <= (a % c == 0)`.


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1502 — Тип 2 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для какого <strong>наименьшего</strong> натурального числа A логическое выражение<br><br><strong>(ДЕЛ(x, 9) → ¬ДЕЛ(x, 6)) ∨ (x + A ≥ 100)</strong><br><br>истинно при любом целом положительном значении x?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 82</div></div></div>

````python
# Задание 1502
# ДЕЛ(x,9) → ¬ДЕЛ(x,6) = (x%9==0) → (x%6!=0) = (x%9==0) <= (x%6!=0)
def f(x, A):
    return ((x % 9 == 0) <= (x % 6 != 0)) or (x + A >= 100)

for A in range(1, 500):
    if all(f(x, A) for x in range(1, 500)):
        print(A)   # Ответ: 82
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1503 — Тип 2 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для какого <strong>наибольшего</strong> натурального числа A логическое выражение<br><br><strong>¬ДЕЛ(x, A) → (ДЕЛ(x, 28) → ¬ДЕЛ(x, 49))</strong><br><br>истинно при любом натуральном значении x?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 196</div></div></div>

````python
# Задание 1503 — наибольшее A => [::-1]
# ¬ДЕЛ(x,A) → (...) = (x%A!=0) <= (...)
def f(x, A):
    return (x % A != 0) <= ((x % 28 == 0) <= (x % 49 != 0))

for A in range(1, 200)[::-1]:
    if all(f(x, A) for x in range(1, 1000)):
        print(A)   # Ответ: 196
        break
````

## Тип 3 — Поразрядная конъюнкция `&`

В условии явно упоминается операция `m & n`. В Python это встроенный оператор, переписывается напрямую. Алгоритм решения не меняется.

```python
# m & n = 0  →  m & n == 0
# ¬(m & n = 0)  →  not (m & n == 0)  или  m & n != 0
```


<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1504 — Тип 3 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для какого <strong>наибольшего</strong> целого положительного числа A выражение<br><br><strong>(x &amp; A = 0) ∨ ¬(x &amp; 37 = 0) ∨ ¬(x &amp; 12 = 0)</strong><br><br>тождественно истинно при любом целом значении x?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 45</div></div></div>

````python
# Задание 1504 — поразрядная конъюнкция, наибольшее A => [::-1]
def f(x, A):
    return (x & A == 0) or (not (x & 37 == 0)) or (not (x & 12 == 0))

for A in range(1, 100)[::-1]:
    if all(f(x, A) for x in range(1, 100)):
        print(A)   # Ответ: 45
        break
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1505 — Тип 3 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Для какого <strong>наименьшего</strong> неотрицательного целого числа A выражение<br><br><strong>x &amp; 39 = 0 ∨ (x &amp; 11 = 0 → ¬(x &amp; A = 0))</strong><br><br>тождественно истинно при любом неотрицательном целом значении x?<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 36</div></div></div>

````python
# Задание 1505 — наименьшее A, x неотрицательные => range(0, ...)
# (x&11=0 → ¬(x&A=0)) = (x&11==0) <= (x&A!=0)
def f(x, A):
    return (x & 39 == 0) or ((x & 11 == 0) <= (not (x & A == 0)))

for A in range(0, 100):
    if all(f(x, A) for x in range(0, 100)):
        print(A)   # Ответ: 36
        break
````

## Тип 4 — Задание с отрезками

Даны два отрезка (B и C), нужно найти **длину** отрезка A, при которой логическое выражение тождественно истинно.

### Ключевые идеи

**1. Отрезки — как `frozenset`**

```python
B = frozenset(range(24, 91))   # [24; 90]
C = frozenset(range(47, 116))  # [47; 115]
```

Принадлежность `x ∈ B` проверяется как `x in B`. Это быстро и читаемо.

**2. Отрезок A задаётся двумя границами A1, A2**

```python
def f(x, A1, A2):
    A = A1 <= x <= A2          # x ∈ A
    return (логическое выражение)
```

**3. Выколотые точки — ε-окрестность**

Граница отрезка может попадать на «выколотую» точку. Проверяем окрестность каждого целого числа с шагом ε = 0.1:

```python
arr = [x + eps for x in range(min_val, max_val) for eps in (0, 0.1, 0.9)]
# Даёт: 24, 24.1, 24.9, 25, 25.1, 25.9, ...
```

**4. Перебор пар границ через `combinations`**

```python
from itertools import combinations

ans = []
for A1, A2 in combinations(arr, 2):
    if all(f(x, A1, A2) for x in arr):
        ans.append(A2 - A1)

print(min(ans))   # или max(ans)
```


<div style="background:#1c1700;border-left:3px solid #f59e0b;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#f59e0b;">Как правильно задать диапазон arr?</strong><br><br>Левую границу берём от <strong>минимального</strong> значения среди всех отрезков B и C.<br>Правую — от <strong>максимального</strong> значения.<br><br>Например, для B=[24;90] и C=[47;115]:<br><code>arr = [x + eps for x in range(24, 115) for eps in (0, 0.1, 0.9)]</code></div>

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1506 — Тип 4 (разбор)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Даны отрезки <strong>B = [24; 90]</strong> и <strong>C = [47; 115]</strong>.<br>Укажите <strong>наименьшую</strong> возможную длину отрезка A, для которого выражение<br><br><strong>(x ∈ C) → ((¬(x ∈ A) ∧ (x ∈ B)) → ¬(x ∈ C))</strong><br><br>истинно при любом значении x.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 43</div></div></div>

````python
# Задание 1506
from itertools import combinations

B = frozenset(range(24, 91))
C = frozenset(range(47, 116))

def f(x, A1, A2):
    A = A1 <= x <= A2
    return (x in C) <= (((not A) and (x in B)) <= (x not in C))

arr = [x + eps for x in range(24, 115) for eps in (0, 0.1, 0.9)]

ans = []
for A1, A2 in combinations(arr, 2):
    if all(f(x, A1, A2) for x in arr):
        ans.append(A2 - A1)

print(min(ans))   # Ответ: 43
````

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Задание 1507 — Тип 4 (пример)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Даны отрезки <strong>P = [15; 40]</strong> и <strong>Q = [21; 63]</strong>.<br>Укажите <strong>наименьшую</strong> возможную длину отрезка A, для которого выражение<br><br><strong>(x ∈ P) → (((x ∈ Q) ∧ ¬(x ∈ A)) → ¬(x ∈ P))</strong><br><br>истинно при любом значении x.<div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-radius:6px;font-family:monospace;font-size:13px;color:#10b981;">✅ Ответ: 19</div></div></div>

````python
# Задание 1507
from itertools import combinations

P = frozenset(range(15, 41))
Q = frozenset(range(21, 64))

def f(x, A1, A2):
    A = A1 <= x <= A2
    return (x in P) <= (((x in Q) and (not A)) <= (x not in P))

arr = [x + eps for x in range(15, 63) for eps in (0, 0.1, 0.9)]

ans = []
for A1, A2 in combinations(arr, 2):
    if all(f(x, A1, A2) for x in arr):
        ans.append(A2 - A1)

print(min(ans))   # Ответ: 19
````

## Шпаргалка

### Перевод логики в Python

| Математика | Python |
|:---|:---|
| `A ∧ B` | `A and B` |
| `A ∨ B` | `A or B` |
| `¬A` | `not A` |
| `A → B` | `A <= B` |
| `ДЕЛ(n, m)` | `n % m == 0` |
| `¬ДЕЛ(n, m)` | `n % m != 0` |
| `m & n = 0` | `m & n == 0` |
| `x ∈ [a; b]` | `x in frozenset(range(a, b+1))` |

### Шаблон типов 1–3
