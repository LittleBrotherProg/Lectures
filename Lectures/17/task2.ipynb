**Важный подготовительный этап:**

Перед выполнением всех пунктов нам необходимо перевести все строчные объекты в списке `file` в числовые объекты.

**Пункт 1:**

```python
mn = min(file)
print(mn)
```

Минимальное число в последовательности: 8

**2 пункт:** Создадим пары элементов. Пары элементов можно составить разными комбинациями. Поэтому обратимся к последнему предложению задачи.

Там сказано: **"В данной задаче под парой подразумевается два идущих подряд элемента последовательности."** Реализуем данные пары с помощью индексов списка.

```python
for i in range(len(file)-1):  # Цикл по индексам от 0 до предпоследнего элемента
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    print(num1,num2)         # Выводим пару соседних элементов
```

`for i in range(len(file)-1):` - цикл проходит по всем индексам от 0 до длины списка минус 2. Минус 1 нужен потому, что мы обращаемся к элементу i+1, иначе на последней итерации получим выход за границы списка.

`num1 = file[i]` - берём элемент с текущим индексом i. Например, если file = [10, 20, 30, 40], то при i=0 получим num1=10.

`num2 = file[i+1]` - берём следующий элемент. Для того же примера при i=0 получим num2=20.

**3 пункт:** Проверим на деление

Проверим условие, что хотя бы одно из чисел пары при делении на 117 даёт остаток, равный минимальному элементу последовательности (8).

```python
c = 0  # Инициализация счетчика пар, удовлетворяющих условию
for i in range(len(file)-1):  # Цикл по индексам от 0 до предпоследнего элемента
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    if num1 % 117 == mn or num2 % 117 == mn:  # Проверяем условие для хотя бы одного из двух чисел
        c += 1               # Увеличиваем счетчик, если условие выполняется
print(c)  # Выводим общее количество пар, удовлетворяющих условию
```

`num1 % 117 == mn` - проверяет, даёт ли текущее число при делении на 117 остаток, равный значению переменной `mn`

`num2 % 117 == mn` - аналогичная проверка для следующего числа

`or` - условие выполняется, если **ХОТЯ БЫ ОДНО** из двух чисел удовлетворяет критерию

Отлично, первый ответ мы получили, но в задаче от нас просят 2 числа.

Посмотрим на вопрос задачи: **"В ответе запишите количество найденных пар, затем максимальную из сумм элементов таких пар."**

Для того чтобы найти максимальную сумму элементов подходящих пар, нам нужна переменная, в которую мы будем запоминать максимальную сумму.

```python
mx = 0  # Инициализация переменной для хранения максимальной суммы пары
c = 0  # Инициализация счетчика пар, удовлетворяющих условию
for i in range(len(file)-1):  # Цикл по индексам от 0 до предпоследнего элемента
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    if num1 % 117 == mn or num2 % 117 == mn:  # Проверяем условие для хотя бы одного из двух чисел
        c += 1               # Увеличиваем счетчик, если условие выполняется
        mx = max(mx, num1+num2)  # Обновляем максимальную сумму, если текущая сумма больше предыдущего максимума
print(c, mx)  # Выводим количество подходящих пар и максимальную сумму среди них
```

`mx = 0` - инициализация переменной для хранения максимальной суммы пары (начинаем с 0, так как суммы чисел всегда ≥ 0)

`mx = max(mx, num1+num2)` - функция `max()` сравнивает текущее значение `mx` с суммой текущей пары `num1+num2` и сохраняет большее значение обратно в `mx`

В итоге мы получаем максимальную сумму пар: **173738**

**Итоговые ответы в нашей задаче:** 175 173738

**Итоговый код:**

```python
# Открываем файл "17_4597.txt" и читаем все строки в список
file = open("17_4597.txt").readlines()

# Преобразуем все строки в список целых чисел
file = list(map(int,file))

# Находим минимальное значение в списке чисел
mn = min(file)

mx = 0  # Инициализация переменной для хранения максимальной суммы пары
c = 0  # Инициализация счетчика пар, удовлетворяющих условию

for i in range(len(file)-1):  # Цикл по индексам от 0 до предпоследнего элемента
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    
    # Проверяем условие: хотя бы одно число должно давать остаток mn при делении на 117
    if num1 % 117 == mn or num2 % 117 == mn:
        c += 1               # Увеличиваем счетчик, если условие выполняется
        mx = max(mx, num1+num2)  # Обновляем максимальную сумму пар

# Выводим количество подходящих пар и максимальную сумму среди них
print(c, mx)
```
