В файле содержится последовательность целых чисел. Элементы последовательности могут принимать целые значения от –10 000 до 10 000 включительно. Определите количество пар последовательности, в которых только одно число оканчивается на 3, а сумма квадратов элементов пары не меньше квадрата максимального элемента последовательности, оканчивающегося на 3. В ответе запишите два числа: сначала количество найденных пар, затем максимальную из сумм квадратов элементов таких пар. В этот раз я покажу новый способ открытия файла и преобразования строчных объектов в числовые.

```python
file = [int(x) for x in open('17_4705.txt').readlines()]
print(file)
```

Теперь обратим внимание на основные условия поиска пар:

**"Определите количество пар последовательности, в которых только одно число оканчивается на 3, а сумма квадратов элементов пары не меньше квадрата максимального элемента последовательности, оканчивающегося на 3."**

**Найдём максимальный элемент последовательности, оканчивающийся на 3:**

```python
mx = max(filter(lambda x: x % 10 == 3, file))
print(mx)
```

Максимальный элемент последовательности, оканчивающийся на 3, найден: **9973**

Теперь возведём его в квадрат, как говорится в условии выше.

Теперь посмотрим, как формируются пары элементов. В данной задаче под парой подразумевается два идущих подряд элемента последовательности.

**Сформируем их:**

```python
for i in range(len(file)-1):
    num1 = file[i]
    num2 = file[i+1]
    print(num1,num2)
```

Теперь определим пары, в которых **только один элемент** заканчивается на 3.

```python
for i in range(len(file)-1):  # Цикл по всем соседним парам элементов в списке
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    bool_num1 = num1 % 10 == 3  # Проверяем, оканчивается ли первое число на 3
    bool_num2 = num2 % 10 == 3  # Проверяем, оканчивается ли второе число на 3
    if bool_num1 + bool_num2 == 1:  # Проверяем, что ровно одно число оканчивается на 3
        print(num1,num2)
```

`num % 10` - получает последнюю цифру числа (остаток от деления на 10)

`num % 10 == 3` - проверяет, равна ли последняя цифра 3

`bool_num1` и `bool_num2` - логические переменные (True/False), которые становятся True, если число оканчивается на 3

---

`bool_num1 + bool_num2 == 1` - в Python True = 1, False = 0, поэтому:

- `0 + 0 = 0` - оба числа НЕ оканчиваются на 3 ❌
- `1 + 0 = 1` - первое оканчивается на 3, второе нет ✅  
- `0 + 1 = 1` - первое не оканчивается, второе оканчивается на 3 ✅
- `1 + 1 = 2` - оба числа оканчиваются на 3 ❌

**Обратите внимание на 5 пару**, которая получилась: `-4140 -4607`. Как видно, ни одно число не заканчивается на 3. Но почему же они тут?

В Python есть проблемы с проверкой на окончание отрицательных чисел.

```python
print(-4140 % 10)
print(-4607 % 10)
```

Как видно, есть проблемы со вторым числом. Почему же так происходит?

**Математическое объяснение:**
В Python оператор `%` возвращает остаток от деления, а не математический модуль. Для отрицательных чисел это работает по формуле:
`a % b = остаток`, где `0 ≤ остаток < b`

**Разберем наш пример:**
`-4607 % 10`

**Шаг 1:** Находим, сколько раз 10 помещается в -4607
`-4607 ÷ 10 = -460.7`

Берем целую часть в меньшую сторону: `-461` (это важно!)
Округление в меньшую сторону гарантирует, что остаток всегда будет неотрицательным и меньше делителя по модулю, что соответствует математическому определению операции взятия остатка в Python.

**Шаг 2:** Вычисляем по формуле:
`остаток = делимое - (делитель * целая_часть)`
`остаток = -4607 - (10 * -461)`
`остаток = -4607 - (-4610)`
`остаток = -4607 + 4610`
`остаток = 3`

**Для получения адекватного остатка при делении** добавим `abs()` для получения модуля числа.

```python
for i in range(len(file)-1):  # Цикл по всем соседним парам элементов в списке
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    bool_num1 = abs(num1) % 10 == 3  # Проверяем, оканчивается ли ПОСЛЕДНЯЯ ЦИФРА ЧИСЛА на 3 (учитывая модуль)
    bool_num2 = abs(num2) % 10 == 3  # Проверяем, оканчивается ли ПОСЛЕДНЯЯ ЦИФРА ЧИСЛА на 3 (учитывая модуль)
    if bool_num1 + bool_num2 == 1:  # Проверяем, что ровно одно число оканчивается на 3
        print(num1, num2)
```

`num % 10` - возвращает остаток от деления на 10, который всегда равен последней цифре числа

`abs(num) % 10` - возвращает последнюю цифру абсолютного значения числа

Теперь каждая наша пара имеет ровно одно число, оканчивающееся на 3. Продолжим выполнять условия, следующие -> **"а сумма квадратов элементов пары не меньше квадрата максимального элемента последовательности"**. Получим сумму квадратов элементов пары и найдём их количество.

```python
c = 0  # Счетчик пар, удовлетворяющих условиям
for i in range(len(file)-1):  # Цикл по всем соседним парам элементов
    num1 = file[i]           # Текущий элемент последовательности
    num2 = file[i+1]         # Следующий элемент последовательности
    bool_num1 = abs(num1) % 10 == 3  # Проверяет, оканчивается ли последняя цифра num1 на 3
    bool_num2 = abs(num2) % 10 == 3  # Проверяет, оканчивается ли последняя цифра num2 на 3
    sm2 = num1**2 + num2**2  # Вычисляет сумму квадратов двух чисел
    if bool_num1 + bool_num2 == 1 and sm2 >= mx:  # Проверяет два условия:
        # 1. Ровно одно число оканчивается на 3
        # 2. Сумма квадратов не меньше значения переменной mx
        c += 1  # Увеличиваем счетчик подходящих пар
print(c)  # Выводим количество найденных пар
```

Перейдём ко второму вопросу. **"В ответе запишите два числа: сначала количество найденных пар, затем максимальную из сумм квадратов элементов таких пар."**

```python
c = 0  # Счетчик пар, удовлетворяющих условиям
sum_mx = 0  # Переменная для хранения максимальной суммы квадратов среди подходящих пар
for i in range(len(file)-1):  # Цикл по всем соседним парам элементов списка
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    bool_num1 = abs(num1) % 10 == 3  # Проверяем, оканчивается ли num1 на 3 (с учетом модуля)
    bool_num2 = abs(num2) % 10 == 3  # Проверяем, оканчивается ли num2 на 3 (с учетом модуля)
    sm2 = num1**2 + num2**2  # Вычисляем сумму квадратов двух чисел
    if bool_num1 + bool_num2 == 1 and sm2 >= mx:  # Проверяем условия:
        # 1. Ровно одно число оканчивается на 3
        # 2. Сумма квадратов не меньше значения переменной mx
        c += 1               # Увеличиваем счетчик подходящих пар
        sum_mx = max(sum_mx, sm2)  # Обновляем максимальную сумму квадратов
print(c, sum_mx)  # Выводим количество пар и максимальную сумму квадратов
```

`sum_mx = max(sum_mx, sm2)` - находит максимальную сумму квадратов среди всех пар, удовлетворяющих условиям

**Полный код:**

```python
# Читаем файл и преобразуем все строки в список целых чисел
file = [int(x) for x in open('17_4705.txt').readlines()]

# Находим максимальное число в файле, которое оканчивается на 3
mx = max(filter(lambda x: x % 10 == 3, file))

# Возводим это максимальное число в квадрат
mx = mx ** 2

c = 0  # Счетчик пар, удовлетворяющих условиям
sum_mx = 0  # Переменная для хранения максимальной суммы квадратов среди подходящих пар

for i in range(len(file)-1):  # Цикл по всем соседним парам элементов списка
    num1 = file[i]           # Получаем текущий элемент последовательности
    num2 = file[i+1]         # Получаем следующий элемент последовательности
    bool_num1 = abs(num1) % 10 == 3  # Проверяем, оканчивается ли num1 на 3 (с учетом модуля)
    bool_num2 = abs(num2) % 10 == 3  # Проверяем, оканчивается ли num2 на 3 (с учетом модуля)
    sm2 = num1**2 + num2**2  # Вычисляем сумму квадратов двух чисел
    
    # Проверяем условия:
    # 1. Ровно одно число оканчивается на 3
    # 2. Сумма квадратов не меньше квадрата максимального числа, оканчивающегося на 3
    if bool_num1 + bool_num2 == 1 and sm2 >= mx:
        c += 1               # Увеличиваем счетчик подходящих пар
        sum_mx = max(sum_mx, sm2)  # Обновляем максимальную сумму квадратов

# Выводим количество пар и максимальную сумму квадратов
print(c, sum_mx)
```
