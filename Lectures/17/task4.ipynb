В файле содержится последовательность целых чисел. Её элементы по модулю не превышают 100000 включительно. Определите количество троек элементов последовательности, в которых хотя бы один элемент оканчивается на 3 и является трёхзначным числом, а сумма всех элементов меньше максимального элемента последовательности, оканчивающегося на 3 и являющегося трёхзначным числом. В ответе запишите количество найденных троек, затем максимальную из сумм элементов таких троек.

В данной задаче под тройкой подразумевается три идущих подряд элемента последовательности.

Файлы к заданию: https://kompege.ru/files/TPT3iXQw0.txt

**1) Открываем файл:**

```python
file = [int(x) for x in open('17_17636.txt').readlines()]
print(file)
```

**"Определите количество троек элементов последовательности, в которых хотя бы один элемент оканчивается на 3 и является трёхзначным числом, а сумма всех элементов меньше максимального элемента последовательности, оканчивающегося на 3 и являющегося трёхзначным числом."**

**1) Находим максимальный элемент последовательности, оканчивающийся на 3 и являющийся трёхзначным числом:**

```python
mx = max(filter(lambda x: abs(x)%10 ==3 and 100 < abs(x) < 1000,file))
print(mx)
```

**2) Определяем количество троек элементов последовательности, в которых хотя бы один элемент оканчивается на 3 и является трёхзначным числом:**

```python
c = 0  # Счетчик троек, удовлетворяющих условию
for i in range(len(file)-2):  # Цикл по всем тройкам подряд идущих элементов
    num1 = file[i]           # Первый элемент тройки
    num2 = file[i+1]         # Второй элемент тройки  
    num3 = file[i+2]         # Третий элемент тройки
    
    # Проверяем, является ли num1 трехзначным числом, оканчивающимся на 3
    bool_num1 = abs(num1) % 10 == 3 and 100 < abs(num1) < 1000
    
    # Проверяем, является ли num2 трехзначным числом, оканчивающимся на 3
    bool_num2 = abs(num2) % 10 == 3 and 100 < abs(num2) < 1000
    
    # Проверяем, является ли num3 трехзначным числом, оканчивающимся на 3
    bool_num3 = abs(num3) % 10 == 3 and 100 < abs(num3) < 1000
    
    # Если хотя бы одно число в тройке является трехзначным и оканчивается на 3
    if bool_num1 + bool_num2 + bool_num3 >= 1:
        print(num1, num2, num3)  # Выводим всю тройку чисел
```

**3) Сумма всех элементов меньше максимального элемента последовательности, оканчивающегося на 3 и являющегося трёхзначным числом:**

```python
c = 0
for i in range(len(file)-2):
    num1 = file[i]
    num2 = file[i+1]
    num3 = file[i+2]
    bool_num1 = abs(num1)%10==3 and 100 < abs(num1) < 1000
    bool_num2 = abs(num2)%10==3 and 100 < abs(num2) < 1000
    bool_num3 = abs(num3)%10==3 and 100 < abs(num3) < 1000
    sum3 = num1 + num2 + num3
    if bool_num1+bool_num2+bool_num3 >= 1 and sum3  < mx:
        c+=1
print(c)
```

**4) В ответе запишите количество найденных троек, затем максимальную из сумм элементов таких троек:**

```python
c = 0  # Счетчик троек, удовлетворяющих условиям
for i in range(len(file)-2):  # Цикл по всем тройкам подряд идущих элементов
    num1 = file[i]           # Первый элемент тройки
    num2 = file[i+1]         # Второй элемент тройки  
    num3 = file[i+2]         # Третий элемент тройки
    
    # Проверяем, является ли num1 трехзначным числом, оканчивающимся на 3
    bool_num1 = abs(num1) % 10 == 3 and 100 < abs(num1) < 1000
    
    # Проверяем, является ли num2 трехзначным числом, оканчивающимся на 3
    bool_num2 = abs(num2) % 10 == 3 and 100 < abs(num2) < 1000
    
    # Проверяем, является ли num3 трехзначным числом, оканчивающимся на 3
    bool_num3 = abs(num3) % 10 == 3 and 100 < abs(num3) < 1000
    
    sum3 = num1 + num2 + num3  # Вычисляем сумму трех чисел
    
    # Если выполняется два условия:
    # 1. Хотя бы одно число в тройке является трехзначным и оканчивается на 3
    # 2. Сумма трех чисел меньше значения переменной mx
    if bool_num1 + bool_num2 + bool_num3 >= 1 and sum3 < mx:
        c += 1  # Увеличиваем счетчик подходящих троек

print(c)  # Выводим общее количество найденных троек
```

**Полный код:**

```python
# Читаем файл и преобразуем все строки в список целых чисел
file = [int(x) for x in open('17_17636.txt').readlines()]

# Находим максимальное трехзначное число, оканчивающееся на 3
mx = max(filter(lambda x: abs(x) % 10 == 3 and 100 < abs(x) < 1000, file))

sum_mx = 0  # Переменная для хранения максимальной суммы среди подходящих троек
c = 0       # Счетчик троек, удовлетворяющих условиям

for i in range(len(file)-2):  # Цикл по всем тройкам подряд идущих элементов
    num1 = file[i]           # Первый элемент тройки
    num2 = file[i+1]         # Второй элемент тройки  
    num3 = file[i+2]         # Третий элемент тройки
    
    # Проверяем, является ли num1 трехзначным числом, оканчивающимся на 3
    bool_num1 = abs(num1) % 10 == 3 and 100 < abs(num1) < 1000
    
    # Проверяем, является ли num2 трехзначным числом, оканчивающимся на 3
    bool_num2 = abs(num2) % 10 == 3 and 100 < abs(num2) < 1000
    
    # Проверяем, является ли num3 трехзначным числом, оканчивающимся на 3
    bool_num3 = abs(num3) % 10 == 3 and 100 < abs(num3) < 1000
    
    sum3 = num1 + num2 + num3  # Вычисляем сумму трех чисел
    
    # Если выполняется два условия:
    # 1. Хотя бы одно число в тройке является трехзначным и оканчивается на 3
    # 2. Сумма трех чисел меньше максимального трехзначного числа, оканчивающегося на 3
    if bool_num1 + bool_num2 + bool_num3 >= 1 and sum3 < mx:
        c += 1               # Увеличиваем счетчик подходящих троек
        sum_mx = max(sum3, sum_mx)  # Обновляем максимальную сумму

# Выводим количество троек и максимальную сумму среди них
print(c, sum_mx)
```

Как видно, условие для максимального числа последовательности и для элементов троек одинаковы, поэтому можем переделать решение под одну функцию.

```python
# Читаем файл и преобразуем все строки в список целых чисел
file = [int(x) for x in open('17_17636.txt').readlines()]

# Создаем функцию для проверки условия: трехзначное число, оканчивающееся на 3
def f(x):
    return abs(x) % 10 == 3 and 100 < abs(x) < 1000

# Находим максимальное число, удовлетворяющее условию функции f
mx = max([x for x in file if f(x) == 1])

# Создаем список для хранения сумм подходящих троек
c = []

for i in range(len(file)-2):  # Цикл по всем тройкам подряд идущих элементов
    # Более компактное присваивание трех переменных
    num1, num2, num3 = file[i], file[i+1], file[i+2]
    
    sum3 = num1 + num2 + num3  # Вычисляем сумму трех чисел
    
    # Проверяем условия:
    # 1. Хотя бы одно число удовлетворяет функции f (трехзначное и оканчивается на 3)
    # 2. Сумма трех чисел меньше максимального числа, найденного ранее
    if f(num1) + f(num2) + f(num3) >= 1 and sum3 < mx:
        c.append(sum3)  # Добавляем сумму в список

# Выводим количество троек и максимальную сумму среди них
print(len(c), max(c))
```

**Преимущества данного кода:**

**1. Использование функции f(x)**
- Устраняет дублирование кода - условие проверки записано один раз
- Повышает читаемость - понятное название функции вместо сложного условия  
- Упрощает изменение - если нужно изменить условие, делаем это в одном месте

**2. Более компактное присваивание**

```python
# Было:
num1 = file[i]
num2 = file[i+1]
num3 = file[i+2]

# Стало:
num1, num2, num3 = file[i], file[i+1], file[i+2]
```

**3. Использование списка вместо счетчика и переменной**

**Было:** отдельные переменные `c` и `sum_mx`

**Стало:** один список `c`, который хранит все суммы

**Преимущество:** сохраняется вся информация о суммах, а не только максимум

**4. Более эффективное вычисление максимума**

**Было:** `sum_mx = max(sum3, sum_mx)` - обновление на каждой итерации

**Стало:** `max(c)` - один раз в конце

**Преимущество:** код проще, а производительность практически не страдает

**5. Улучшенная фильтрация для mx**

```python
# Было: filter с lambda
mx = max(filter(lambda x: abs(x)%10 ==3 and 100 < abs(x) < 1000, file))

# Стало: list comprehension с функцией
mx = max([x for x in file if f(x)==1])
```

**Преимущество:** более читаемо и использует уже созданную функцию

**Итог:**
Код стал более читаемым, поддерживаемым и компактным без потери функциональности.

Функция `f(x)` позволяет легко понять логику проверки и при необходимости модифицировать её.
