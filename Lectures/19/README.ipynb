## 3. Тип 1 — Одна куча камней

<div class="tip-box">
<strong style="color:#2fca72;">Признак:</strong> «Петя и Ваня добавляют (или убирают) камни из <strong>одной кучи</strong>»
</div>

### Структура функции

````python
def f(s, m):
    # ШАГ 1: Условие победы (подставь свой предел вместо 51)
    if s >= 51:
        return m % 2 == 0  # чётный ход = ход Вани = победа Вани

    # ШАГ 2: Ходы закончились, победителя нет
    if m == 0:
        return False

    # ШАГ 3: Список всех возможных ходов (подставь свои операции)
    h = [f(s + 1, m - 1),   # операция 1: +1 камень
         f(s + 4, m - 1),   # операция 2: +4 камня
         f(s * 2, m - 1)]   # операция 3: умножить на 2

    # ШАГ 4: any() для хода Вани, all() для хода Пети
    return any(h) if (m + 1) % 2 == 0 else all(h)


# Задание 19: минимальное S, где Ваня выигрывает первым ходом
print('19:', *[s for s in range(1, 51) if f(s, 2)])

# Задание 20: Петя выигрывает вторым ходом, но НЕ первым
print('20:', *[s for s in range(1, 51) if f(s, 3) and not f(s, 1)])

# Задание 21: Ваня выигрывает первым или вторым, но НЕ только первым
print('21:', min([s for s in range(1, 51) if f(s, 4) and not f(s, 2)]))
````

### Разбор каждой строки

| Строка | Что делает | Почему так |
|:---|:---|:---|
| `if s >= 51` | Проверка победы | Достигли предела — игра окончена |
| `m % 2 == 0` | Определяем победителя | `m=2` → чётный → ход Вани → он победил |
| `if m == 0` | Ходы исчерпаны | Никто не успел выиграть |
| `h = [f(..., m-1)]` | Все возможные ходы | Рекурсия для каждой операции |
| `any(h)` | Ход Вани | Ему хватит одного победного исхода |
| `all(h)` | Ход Пети | Все его ходы должны вести к поражению |
| `(m+1) % 2 == 0` | Чей сейчас ход | Определяем по чётности текущего m |

### Пример: камни убываются (-2, -5, //3)

````python
def f(s, m):
    if s <= 19:              # победа при s <= 19 (убываем, не прибавляем)
        return m % 2 == 0
    if m == 0:
        return False
    h = [f(s - 2, m - 1),   # убрать 2 камня
         f(s - 5, m - 1),   # убрать 5 камней
         f(s // 3, m - 1)]  # разделить на 3 (целочисленно)
    return any(h) if (m + 1) % 2 == 0 else all(h)

print('19:', min([s for s in range(20, 100) if f(s, 2)]))
print('20:', *[s for s in range(20, 100) if f(s, 3) and not f(s, 1)])
print('21:', *[s for s in range(20, 100) if f(s, 4) and not f(s, 2)]))
````

## 4. Тип 2 — Две кучи камней

<div class="tip-box">
<strong style="color:#2fca72;">Признак:</strong> «Игроки применяют операцию к <strong>одной из двух куч</strong> по своему выбору»
</div>

<div class="warn-box">
<strong style="color:#f0c040;">⚠️ Особенность задания 19 с двумя кучами:</strong><br><br>
Формулировка «Ваня выиграл после <strong>неудачного</strong> хода Пети» означает — достаточно <strong>хотя бы одного</strong> неудачного хода.<br>
Для задания 19 используем только <code>any()</code>, убираем <code>all()</code>.<br>
Для заданий 20–21 возвращаем стандартную проверку.
</div>

### Задание 19 (две кучи, неудачный ход)

````python
def f(s1, s2, m):
    # Победа: СУММА камней в двух кучах >= предела
    if s1 + s2 >= 81:
        return m % 2 == 0
    if m == 0:
        return False

    # Каждую операцию применяем к каждой из двух куч (4 варианта)
    h = [f(s1 + 1, s2, m - 1),  # +1 к первой куче
         f(s1, s2 + 1, m - 1),  # +1 ко второй куче
         f(s1 * 2, s2, m - 1),  # ×2 первую кучу
         f(s1, s2 * 2, m - 1)]  # ×2 вторую кучу

    # Для задания 19: только any() — достаточно одного неудачного хода Пети
    return any(h)


# Задание 19 (первая куча фиксирована = 7 камней)
print('19:', min([s for s in range(1, 74) if f(7, s, 2)]))
````

### Задания 20–21 (две кучи, стандартная проверка)

````python
def f(s1, s2, m):
    if s1 + s2 >= 81:
        return m % 2 == 0
    if m == 0:
        return False

    h = [f(s1 + 1, s2, m - 1),
         f(s1, s2 + 1, m - 1),
         f(s1 * 2, s2, m - 1),
         f(s1, s2 * 2, m - 1)]

    # Возвращаем стандартную проверку: any для Вани, all для Пети
    return any(h) if (m + 1) % 2 == 0 else all(h)


# Задание 20
print('20:', *[s for s in range(1, 74) if f(7, s, 3) and not f(7, s, 1)])

# Задание 21
print('21:', min([s for s in range(1, 74) if f(7, s, 4) and not f(7, s, 2)]))
````
