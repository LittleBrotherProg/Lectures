## Теория

Задания 19–21 — игра с двумя игроками (Ваня и Петя). Нужно определить, выиграет ли первый игрок при правильной стратегии.

**Ключевые правила:**
- `m % 2 == 0` → чётный ход → ход Вани (первого)
- Победа при `s >= limit` (или `s <= limit` при убывании)
- Рекурсия перебирает все возможные ходы

```python
def f(s, m):
    if s >= 51:              # условие победы
        return m % 2 == 0   # чётный ход = ход Вани = победа Вани
    if m == 0:
        return False         # ходы кончились без победителя
    # all() — Ваня выигрывает при любом ходе противника
    # any() — Ваня выигрывает хотя бы при одном своём ходе
    ходы = [f(s+3, m-1), f(s*2, m-1)]
    if m % 2 == 0:           # ход Вани → он выбирает выгодный
        return any(ходы)
    else:                    # ход Пети → он выбирает невыгодный для Вани
        return all(ходы)
```

<div class="task-title">📋 Задание Тип 19 — Пример: убывание, запрещённые ходы | Ответ: <strong style="color:#2fca72;">Перебором</strong></div>
<div style="color:#c8d8e8;line-height:1.7;">Два игрока делают ходы поочерёдно. На столе s камней. Можно убрать 2 или 5 камней, или разделить нацело на 3. Побеждает тот, кто делает ход при s ≤ 19. Всего ходов не более m.<br>
<strong>При каких начальных значениях s Ваня (первый) выигрывает при m=2?</strong></div>
</div>

````python
def f(s, m):
    if s <= 19:              # победа при малом числе камней
        return m % 2 == 0
    if m == 0:
        return False
    moves = [f(s-2, m-1),   # убрать 2
             f(s-5, m-1),   # убрать 5
             f(s//3, m-1)]  # разделить на 3 (только если делится)
    # Фильтруем невалидные ходы:
    valid = []
    if s >= 2:  valid.append(f(s-2, m-1))
    if s >= 5:  valid.append(f(s-5, m-1))
    if s % 3 == 0 and s > 19: valid.append(f(s//3, m-1))
    if not valid: return m % 2 != 0   # нет ходов = проигрыш
    return any(valid) if m % 2 == 0 else all(valid)

# Задание 19: f(s, 2)
winners = [s for s in range(1, 100) if f(s, 2)]
print("Ваня выигрывает при s:", winners)
````
