## О задании

Задания 19, 20, 21 — **повышенный** уровень, 2 балла. В каждом задании описана игра двух игроков, Ваня и Петя делают ходы поочерёдно. Ваня ходит первым. Нужно написать программу и определить, при каких начальных значениях Ваня выигрывает при **правильной игре обеих сторон**.

| Задание | Что найти |
|:---:|:---|
| 19 | При каких `s` Ваня выигрывает (`f(s, 2)`) |
| 20 | При каких `s` Ваня выигрывает за N ходов, но **не** за N-2 |
| 21 | Аналогично 20, но для другого N |

## Ключевая идея: рекурсия + any/all

Функция `f(s, m)` возвращает `True`, если текущий игрок (тот, кто ходит при количестве оставшихся ходов `m`) выигрывает при правильной игре обеих сторон.

- **Чётный ход (`m % 2 == 0`)** — ходит **Ваня** → ему достаточно найти **хотя бы один** выигрышный ход → `any()`
- **Нечётный ход (`m % 2 != 0`)** — ходит **Петя** → он выбирает **наихудший** для Вани ход → `all()` (Ваня выигрывает только если **все** ходы Пети ведут к победе Вани)

<div style="background:#0d1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;padding:12px 18px;margin:12px 0;font-size:14px;color:#94a3b8;font-family:sans-serif;"><strong style="color:#10b981;">Запомните:</strong> <code>any()</code> = ходит Ваня (оптимист), <code>all()</code> = ходит Петя (пессимист для Вани).</div>

## Универсальный шаблон функции

```python
def f(s, m):
    # 1. Условие победы
    if s >= LIMIT:
        return m % 2 == 0   # True, если ход Вани (чётный) — он победил

    # 2. Ходы кончились
    if m == 0:
        return False

    # 3. Все возможные ходы
    moves = [f(s + op, m - 1) for op in ОПЕРАЦИИ]

    # 4. Стратегия
    if m % 2 == 0:           # ход Вани → ищем хотя бы одну победу
        return any(moves)
    else:                    # ход Пети → Ваня побеждает только если всё хорошо
        return all(moves)

# Задание 19:
winners = [s for s in range(1, 100) if f(s, 2)]
print(winners)

# Задание 20 (выигрывает за 3 хода, но не за 1):
result20 = [s for s in range(1, 100) if f(s, 3) and not f(s, 1)]

# Задание 21 (за 4 хода, но не за 2):
result21 = [s for s in range(1, 100) if f(s, 4) and not f(s, 2)]
```

## Пример 1 — камни убываются (−2, −5, //3)

Ваня и Петя делают ходы поочерёдно. На столе `s` камней. Игрок может убрать 2 или 5 камней, или разделить количество нацело на 3. Победа наступает при `s ≤ 19`.

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Пример 1 — Убывание (−2, −5, //3)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Два игрока делают ходы поочерёдно, начиная с Вани. На столе s камней. Разрешённые ходы:<br>
<strong>A. Убрать 2 камня<br>
B. Убрать 5 камней<br>
С. Разделить количество камней нацело на 3 (только если делится)</strong><br><br>
Победа наступает при s ≤ 19 — тот, кто делает ход при таком числе камней, выигрывает.<br>
Ваня и Петя играют с максимальным числом ходов m = 2.<br><br>
<strong>При каких начальных значениях s (от 1 до 60) Ваня гарантированно выигрывает?</strong></div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: Список значений</div></div>

````python
def f(s, m):
    if s <= 19:             # победа при маленьком числе камней
        return m % 2 == 0  # чётный ход = ход Вани
    if m == 0:
        return False

    moves = []
    if s >= 2: moves.append(f(s - 2, m - 1))   # убрать 2
    if s >= 5: moves.append(f(s - 5, m - 1))   # убрать 5
    if s % 3 == 0 and s > 19:                   # разделить на 3
        moves.append(f(s // 3, m - 1))

    if not moves:
        return m % 2 != 0  # нет ходов → текущий проигрывает

    return any(moves) if m % 2 == 0 else all(moves)

# Задание 19: f(s, 2)
winners_19 = [s for s in range(1, 61) if f(s, 2)]
print("Задание 19:", winners_19)

# Задание 20: f(s, 3) and not f(s, 1)
winners_20 = [s for s in range(1, 61) if f(s, 3) and not f(s, 1)]
print("Задание 20:", winners_20)

# Задание 21: f(s, 4) and not f(s, 2)
winners_21 = [s for s in range(1, 61) if f(s, 4) and not f(s, 2)]
print("Задание 21:", winners_21)
````

## Пример 2 — две кучи камней

Если в задании **два параметра** (две кучи s1 и s2), функция принимает три аргумента: `f(s1, s2, m)`.

Победа обычно определяется по **сумме** камней.

<div style="background:#1a2235;border:1px solid #2a3a5c;border-left:3px solid #f59e0b;border-radius:8px;padding:18px 22px;margin:16px 0;"><div style="color:#f59e0b;font-size:12px;font-weight:700;letter-spacing:1px;text-transform:uppercase;margin-bottom:10px;font-family:sans-serif;">📋 Пример 2 — Две кучи (+1, ×2 к любой)</div><div style="color:#e2e8f0;line-height:1.8;font-size:14px;font-family:sans-serif;">Ваня и Петя делают ходы поочерёдно. Перед игроками две кучи: в первой s₁ камней, во второй s₂ камней. За один ход игрок может:<br>
<strong>A. Добавить 1 камень в любую кучу<br>
B. Удвоить любую кучу</strong><br><br>
Победа наступает, когда сумма камней в обеих кучах ≥ 81.<br>
Ваня ходит первым, всего ходов m.<br><br>
<strong>При каких начальных (s₁, s₂) Ваня выигрывает при m=2?</strong></div><div style="margin-top:12px;padding:8px 14px;background:#0f1f12;border-left:3px solid #10b981;border-radius:0 6px 6px 0;color:#10b981;font-weight:600;font-size:14px;font-family:sans-serif;">✅ Ответ: Список пар</div></div>

````python
from functools import lru_cache

@lru_cache(maxsize=None)
def f(s1, s2, m):
    if s1 + s2 >= 81:
        return m % 2 == 0      # победа
    if m == 0:
        return False

    moves = [
        f(s1 + 1, s2, m - 1),  # A к первой
        f(s1, s2 + 1, m - 1),  # A ко второй
        f(s1 * 2, s2, m - 1),  # B к первой
        f(s1, s2 * 2, m - 1),  # B ко второй
    ]

    return any(moves) if m % 2 == 0 else all(moves)

# Задание 19: при каких (s1, s2) Ваня выигрывает с 2 ходами
winners = [(s1, s2) for s1 in range(1, 50) for s2 in range(1, 50)
           if f(s1, s2, 2)]
print(f"Ваня выигрывает при {len(winners)} начальных позициях")
print("Примеры:", winners[:5])
````

## Шпаргалка

### Структура функции
```python
def f(s, m):
    if ПОБЕДА:       return m % 2 == 0
    if m == 0:       return False
    moves = [f(s+op, m-1) for op in операции]
    return any(moves) if m % 2 == 0 else all(moves)
```

### Какой вызов использовать
| Задание | Вызов |
|:---:|:---|
| 19 | `f(s, 2)` |
| 20 | `f(s, 3) and not f(s, 1)` |
| 21 | `f(s, 4) and not f(s, 2)` |

### Правило any/all
| Ход | Функция | Смысл |
|:---:|:---:|:---|
| Чётный (Ваня) | `any()` | Хотя бы один ход выигрышный |
| Нечётный (Петя) | `all()` | Все ходы Пети ведут к победе Вани |
