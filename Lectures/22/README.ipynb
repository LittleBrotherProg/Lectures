**Алгоритм:**
1. Открыть файл — несколько листов с процессами
2. Вычислить время завершения каждого процесса: `=C2+D2`
3. ВПР найти, когда освободится нужный поток: `=ВПР(C2;$A:$G;7;0)`

5. Подсчитать ответ по условию

## Полный алгоритм в LibreOffice Calc

```
Шаг 1. Добавить столбец "Время завершения":
   E2 = C2 + D2
   (растянуть вниз)

Шаг 2. Отсортировать по приоритету (убывание), потом по времени начала (возрастание):
   Данные → Сортировка

Шаг 3. Для каждого процесса найти когда поток свободен:
   ВПР(C2; $A:$G; 7; 0)

Шаг 4. Фактическое начало = МАКС(C2; ВПР-результат)

Шаг 5. Фильтр по фактическому началу >= 6
   СЧЁТЕСЛИ(диапазон; ">=6")
```

<div class="warn-box">
<strong style="color:#f0c040;">Важно:</strong> нумерация мс начинается с 1, поэтому условие «не ранее 6-й мс» означает фактическое время начала ≥ 6.
</div>

## Шпаргалка

```
# Время завершения:
=C2 + D2

# ВПР для поиска завершения потока:
=ВПР(C2; $A:$G; 7; 0)

# Фактическое начало:
=МАКС(C2; ВПР(C2; $A:$G; 7; 0))

# Подсчёт процессов с начала >= 6:
=СЧЁТЕСЛИ(F2:F100; ">=6")
```
