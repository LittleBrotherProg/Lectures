## О задании

Задание 22 — **повышенный** уровень. Даётся файл с таблицей процессов. Нужно смоделировать работу планировщика задач с двумя (или более) параллельными потоками.

**Что дано:**
- Таблица: номер процесса, приоритет, время начала (мс), длительность (мс)
- Процессы назначаются потокам в порядке убывания приоритета
- При равных приоритетах — по возрастанию времени начала

**Что найти:** количество процессов, начавшихся не ранее заданного момента.

## Ключевые понятия

| Термин | Описание |
|:---|:---|
| **Поток** | Параллельная линия выполнения |
| **Время завершения** | Время начала + длительность |
| **Ожидание** | Процесс ждёт, пока поток освободится |
| **Фактическое начало** | `MAX(заявленное время; освобождение потока)` |

## Алгоритм решения в LibreOffice Calc

```
Шаг 1. Открыть файл, изучить таблицу процессов

Шаг 2. Добавить столбец "Время завершения":
        =C2 + D2
        (растянуть вниз)

Шаг 3. Отсортировать по приоритету (убывание),
        затем по времени начала (возрастание):
        Данные → Сортировка

Шаг 4. Для каждого процесса найти, когда освободится
        нужный поток через ВПР:
        =ВПР(C2; $A:$G; 7; 0)

Шаг 5. Фактическое начало процесса:
        =МАКС(C2; ВПР(C2; $A:$G; 7; 0))

Шаг 6. Подсчитать результат:
        =СЧЁТЕСЛИ(F2:F100; ">=6")
```

## Полные формулы для Calc

```
E2: =C2 + D2                           # Время завершения

# После сортировки по приоритету ↓ и времени ↑:
F2: =МАКС(C2; ВПР(C2; $A:$G; 7; 0))   # Фактическое начало

# Ответ:
=СЧЁТЕСЛИ(F2:F100; ">=6")
```
